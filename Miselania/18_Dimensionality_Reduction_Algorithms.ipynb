# **18. Dimensionality Reduction Algorithms**

**Algoritmos de Reducci贸n de Dimensionalidad:**

La reducci贸n de dimensionalidad es el proceso de reducir la cantidad de variables (dimensiones) en un conjunto de datos mientras se mantiene la mayor parte de su informaci贸n relevante. Esto es 煤til para abordar problemas como la "maldici贸n de la dimensionalidad", mejorar la eficiencia computacional y visualizar datos en espacios de menor dimensi贸n. Algunos algoritmos comunes incluyen PCA (An谩lisis de Componentes Principales), t-SNE (t-distributed Stochastic Neighbor Embedding) y UMAP (Uniform Manifold Approximation and Projection).

**Terminolog铆a:**

- **Componentes Principales:** Nuevas variables obtenidas al proyectar los datos originales en la direcci贸n donde la variabilidad es m谩xima.
- **Embedding (Incrustaci贸n):** Proceso de mapear datos de alta dimensi贸n a un espacio de menor dimensi贸n.
- **Autovalores y Autovectores:** En PCA, los autovalores representan la cantidad de varianza explicada por cada componente principal, y los autovectores representan las direcciones de estos componentes.

**PCA (An谩lisis de Componentes Principales):**

- **F贸rmula Matem谩tica:**

  - Dada una matriz de datos $(X)$ de dimensiones $(m \times n)$ (m observaciones, n variables), la matriz de covarianza es $C = \frac{1}{m} X^T X$.
  - Los autovalores $(\lambda_i)$ y autovectores $(v_i)$ de $(C)$ son calculados.
  - Los autovectores ordenados por autovalores proporcionan las direcciones de los componentes principales.
  - La proyecci贸n de $(X)$ en el espacio de los componentes principales es $X_{\text{pca}} = X V$, donde $(V)$ es la matriz de autovectores.

**Ejemplo en Python:**

```python
import numpy as np
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

# Cargar el conjunto de datos Iris
iris = load_iris()
X = iris.data
y = iris.target

# Aplicar PCA para reducir a 2 dimensiones
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Visualizar los resultados con personalizaci贸n
plt.figure(figsize=(8, 6))

# Scatter plot con etiquetas y colores personalizados
scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis', edgecolors='k', marker='o', s=70)

# Etiquetas de ejes
plt.xlabel('Primera Componente Principal', fontsize=12)
plt.ylabel('Segunda Componente Principal', fontsize=12)

# T铆tulo de la gr谩fica
plt.title('PCA: Reducci贸n de Dimensionalidad con Personalizaci贸n', fontsize=16)

# A帽adir barra de color
colorbar = plt.colorbar(scatter)
colorbar.set_label('Clases', fontsize=12)

# Mostrar la gr谩fica
plt.show()
```

**t-SNE (t-distributed Stochastic Neighbor Embedding):**

- **F贸rmula Matem谩tica:**

  - Calcula probabilidades condicionales en el espacio de alta dimensi贸n y baja dimensi贸n.
  - Minimiza la divergencia de Kullback-Leibler entre las distribuciones de probabilidad condicional.

- **Ejemplo en Python:**

```python
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

# Aplicar t-SNE para reducir a 2 dimensiones
tsne = TSNE(n_components=2, perplexity=30, random_state=42)
X_tsne = tsne.fit_transform(X)

# Visualizar los resultados con personalizaci贸n
plt.figure(figsize=(8, 6))

# Scatter plot con etiquetas y colores personalizados
scatter = plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, cmap='viridis', edgecolors='k', marker='o', s=70)

# Etiquetas de ejes
plt.xlabel('t-SNE Dimensi贸n 1', fontsize=12)
plt.ylabel('t-SNE Dimensi贸n 2', fontsize=12)

# T铆tulo de la gr谩fica
plt.title('t-SNE: Reducci贸n de Dimensionalidad con Personalizaci贸n', fontsize=16)

# A帽adir barra de color
colorbar = plt.colorbar(scatter)
colorbar.set_label('Clases', fontsize=12)

# Mostrar la gr谩fica
plt.show()
```

**UMAP (Uniform Manifold Approximation and Projection):**

- **F贸rmula Matem谩tica:**

  - Utiliza t茅cnicas de optimizaci贸n algebraica y geom茅trica para aproximar la estructura geom茅trica subyacente.
  - Minimiza la energ铆a potencial.

- **Ejemplo en Python:**

```python
import umap
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore", category=UserWarning)

# Aplicar UMAP para reducir a 2 dimensiones
umap_model = umap.UMAP(n_components=2, random_state=42)
X_umap = umap_model.fit_transform(X)

# Visualizar los resultados con personalizaci贸n
plt.figure(figsize=(8, 6))

# Scatter plot con etiquetas y colores personalizados
scatter = plt.scatter(X_umap[:, 0], X_umap[:, 1], c=y, cmap='viridis', edgecolors='k', marker='o', s=70)

# Etiquetas de ejes
plt.xlabel('UMAP Dimensi贸n 1', fontsize=12)
plt.ylabel('UMAP Dimensi贸n 2', fontsize=12)

# T铆tulo de la gr谩fica
plt.title('UMAP: Reducci贸n de Dimensionalidad con Personalizaci贸n', fontsize=16)

# A帽adir barra de color
colorbar = plt.colorbar(scatter)
colorbar.set_label('Clases', fontsize=12)

# Mostrar la gr谩fica
plt.show()
```
