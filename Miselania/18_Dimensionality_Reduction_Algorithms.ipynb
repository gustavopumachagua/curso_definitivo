# **18. Dimensionality Reduction Algorithms**

**Algoritmos de Reducción de Dimensionalidad:**

La reducción de dimensionalidad es el proceso de reducir la cantidad de variables (dimensiones) en un conjunto de datos mientras se mantiene la mayor parte de su información relevante. Esto es útil para abordar problemas como la "maldición de la dimensionalidad", mejorar la eficiencia computacional y visualizar datos en espacios de menor dimensión. Algunos algoritmos comunes incluyen PCA (Análisis de Componentes Principales), t-SNE (t-distributed Stochastic Neighbor Embedding) y UMAP (Uniform Manifold Approximation and Projection).

**Terminología:**

- **Componentes Principales:** Nuevas variables obtenidas al proyectar los datos originales en la dirección donde la variabilidad es máxima.
- **Embedding (Incrustación):** Proceso de mapear datos de alta dimensión a un espacio de menor dimensión.
- **Autovalores y Autovectores:** En PCA, los autovalores representan la cantidad de varianza explicada por cada componente principal, y los autovectores representan las direcciones de estos componentes.

**PCA (Análisis de Componentes Principales):**

- **Fórmula Matemática:**

  - Dada una matriz de datos $(X)$ de dimensiones $(m \times n)$ (m observaciones, n variables), la matriz de covarianza es $C = \frac{1}{m} X^T X$.
  - Los autovalores $(\lambda_i)$ y autovectores $(v_i)$ de $(C)$ son calculados.
  - Los autovectores ordenados por autovalores proporcionan las direcciones de los componentes principales.
  - La proyección de $(X)$ en el espacio de los componentes principales es $X_{\text{pca}} = X V$, donde $(V)$ es la matriz de autovectores.

**Ejemplo en Python:**

```python
import numpy as np
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

# Cargar el conjunto de datos Iris
iris = load_iris()
X = iris.data
y = iris.target

# Aplicar PCA para reducir a 2 dimensiones
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Visualizar los resultados con personalización
plt.figure(figsize=(8, 6))

# Scatter plot con etiquetas y colores personalizados
scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis', edgecolors='k', marker='o', s=70)

# Etiquetas de ejes
plt.xlabel('Primera Componente Principal', fontsize=12)
plt.ylabel('Segunda Componente Principal', fontsize=12)

# Título de la gráfica
plt.title('PCA: Reducción de Dimensionalidad con Personalización', fontsize=16)

# Añadir barra de color
colorbar = plt.colorbar(scatter)
colorbar.set_label('Clases', fontsize=12)

# Mostrar la gráfica
plt.show()
```

**t-SNE (t-distributed Stochastic Neighbor Embedding):**

- **Fórmula Matemática:**

  - Calcula probabilidades condicionales en el espacio de alta dimensión y baja dimensión.
  - Minimiza la divergencia de Kullback-Leibler entre las distribuciones de probabilidad condicional.

- **Ejemplo en Python:**

```python
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

# Aplicar t-SNE para reducir a 2 dimensiones
tsne = TSNE(n_components=2, perplexity=30, random_state=42)
X_tsne = tsne.fit_transform(X)

# Visualizar los resultados con personalización
plt.figure(figsize=(8, 6))

# Scatter plot con etiquetas y colores personalizados
scatter = plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, cmap='viridis', edgecolors='k', marker='o', s=70)

# Etiquetas de ejes
plt.xlabel('t-SNE Dimensión 1', fontsize=12)
plt.ylabel('t-SNE Dimensión 2', fontsize=12)

# Título de la gráfica
plt.title('t-SNE: Reducción de Dimensionalidad con Personalización', fontsize=16)

# Añadir barra de color
colorbar = plt.colorbar(scatter)
colorbar.set_label('Clases', fontsize=12)

# Mostrar la gráfica
plt.show()
```

**UMAP (Uniform Manifold Approximation and Projection):**

- **Fórmula Matemática:**

  - Utiliza técnicas de optimización algebraica y geométrica para aproximar la estructura geométrica subyacente.
  - Minimiza la energía potencial.

- **Ejemplo en Python:**

```python
import umap
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore", category=UserWarning)

# Aplicar UMAP para reducir a 2 dimensiones
umap_model = umap.UMAP(n_components=2, random_state=42)
X_umap = umap_model.fit_transform(X)

# Visualizar los resultados con personalización
plt.figure(figsize=(8, 6))

# Scatter plot con etiquetas y colores personalizados
scatter = plt.scatter(X_umap[:, 0], X_umap[:, 1], c=y, cmap='viridis', edgecolors='k', marker='o', s=70)

# Etiquetas de ejes
plt.xlabel('UMAP Dimensión 1', fontsize=12)
plt.ylabel('UMAP Dimensión 2', fontsize=12)

# Título de la gráfica
plt.title('UMAP: Reducción de Dimensionalidad con Personalización', fontsize=16)

# Añadir barra de color
colorbar = plt.colorbar(scatter)
colorbar.set_label('Clases', fontsize=12)

# Mostrar la gráfica
plt.show()
```

Estos ejemplos ilustran cómo aplicar PCA, t-SNE y UMAP para reducir la dimensionalidad de un conjunto de datos y visualizarlo en un espacio de menor dimensión.
