**Teorema del L铆mite Central (CLT):**

El Teorema del L铆mite Central es un principio fundamental en estad铆sticas que describe el comportamiento de las medias de muestras grandes, independientemente de la distribuci贸n original de los datos. Establece que, a medida que el tama帽o de la muestra aumenta, la distribuci贸n de las medias de las muestras se aproxima a una distribuci贸n normal, independientemente de la forma de la distribuci贸n original.

**Terminolog铆a:**

- **Muestra Aleatoria:** Un conjunto de observaciones seleccionadas de una poblaci贸n de manera que cada miembro tenga la misma probabilidad de ser incluido.

- **Media de la Muestra $(\bar{X})$:** La media calculada a partir de una muestra espec铆fica.

- **Media Poblacional $(\mu)$:** La media de toda la poblaci贸n.

- **Desviaci贸n Est谩ndar de la Poblaci贸n $(\sigma)$:** La medida de dispersi贸n de toda la poblaci贸n.

- **Desviaci贸n Est谩ndar de la Muestra $(s)$:** La medida de dispersi贸n de una muestra espec铆fica.

**F贸rmula Matem谩tica para el Teorema del L铆mite Central:**

$Z = \frac{\bar{X} - \mu}{\frac{\sigma}{\sqrt{n}}}$

Donde:

- $( Z )$ es la puntuaci贸n z que sigue una distribuci贸n normal est谩ndar.
  
- $( \bar{X} )$ es la media de la muestra.
  
- $( \mu )$ es la media poblacional.
  
- $( \sigma )$ es la desviaci贸n est谩ndar poblacional.
  
- $( n )$ es el tama帽o de la muestra.

**Ejemplo con Gr谩fico en Python:**

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos de una distribuci贸n no normal (por ejemplo, una distribuci贸n exponencial)
np.random.seed(42)
population_data = np.random.exponential(scale=2, size=10000)

# Funci贸n para calcular la media de la muestra
def sample_mean(sample_size):
    return np.mean(np.random.choice(population_data, size=sample_size, replace=True))

# Realizar muestreo y calcular las medias de la muestra
sample_sizes = [5, 20, 50, 100]
num_samples = 1000

plt.figure(figsize=(12, 8))

for i, size in enumerate(sample_sizes, 1):
    sample_means = [sample_mean(size) for _ in range(num_samples)]
    
    # Normalizar las medias para comparaci贸n
    normalized_means = (sample_means - np.mean(population_data)) / (np.std(population_data) / np.sqrt(size))
    
    # Histograma de las medias normalizadas
    plt.subplot(2, 2, i)
    plt.hist(normalized_means, bins=30, density=True, alpha=0.7, color='skyblue', edgecolor='black')
    plt.title(f'Tama帽o de Muestra: {size}')
    plt.xlabel('Medias Normalizadas')
    plt.ylabel('Frecuencia')

plt.tight_layout()
plt.show()
```

En este ejemplo:
- Generamos datos de una distribuci贸n exponencial para representar la poblaci贸n.
- Realizamos muestreos de diferentes tama帽os y calculamos las medias de las muestras.
- Normalizamos las medias para compararlas y observamos c贸mo la distribuci贸n se aproxima a una normal a medida que aumenta el tama帽o de la muestra.
