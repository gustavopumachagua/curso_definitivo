# **52. Central Limit Theorem**

**Teorema del Límite Central (CLT): Explicación Detallada**

**Teorema del Límite Central (CLT):**

El Teorema del Límite Central es un principio fundamental en estadísticas que describe el comportamiento de las medias de muestras grandes, independientemente de la distribución original de los datos. Establece que, a medida que el tamaño de la muestra aumenta, la distribución de las medias de las muestras se aproxima a una distribución normal, independientemente de la forma de la distribución original.

**Terminología:**

- **Muestra Aleatoria:** Un conjunto de observaciones seleccionadas de una población de manera que cada miembro tenga la misma probabilidad de ser incluido.

- **Media de la Muestra $(\bar{X})$:** La media calculada a partir de una muestra específica.

- **Media Poblacional $(\mu)$:** La media de toda la población.

- **Desviación Estándar de la Población $(\sigma)$:** La medida de dispersión de toda la población.

- **Desviación Estándar de la Muestra $(s)$:** La medida de dispersión de una muestra específica.

**Fórmula Matemática para el Teorema del Límite Central:**

$Z = \frac{\bar{X} - \mu}{\frac{\sigma}{\sqrt{n}}}$

Donde:

- $( Z )$ es la puntuación z que sigue una distribución normal estándar.
  
- $( \bar{X} )$ es la media de la muestra.
  
- $( \mu )$ es la media poblacional.
  
- $( \sigma )$ es la desviación estándar poblacional.
  
- $( n )$ es el tamaño de la muestra.

**Ejemplo con Gráfico en Python:**

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos de una distribución no normal (por ejemplo, una distribución exponencial)
np.random.seed(42)
population_data = np.random.exponential(scale=2, size=10000)

# Función para calcular la media de la muestra
def sample_mean(sample_size):
    return np.mean(np.random.choice(population_data, size=sample_size, replace=True))

# Realizar muestreo y calcular las medias de la muestra
sample_sizes = [5, 20, 50, 100]
num_samples = 1000

plt.figure(figsize=(12, 8))

for i, size in enumerate(sample_sizes, 1):
    sample_means = [sample_mean(size) for _ in range(num_samples)]
    
    # Normalizar las medias para comparación
    normalized_means = (sample_means - np.mean(population_data)) / (np.std(population_data) / np.sqrt(size))
    
    # Histograma de las medias normalizadas
    plt.subplot(2, 2, i)
    plt.hist(normalized_means, bins=30, density=True, alpha=0.7, color='skyblue', edgecolor='black')
    plt.title(f'Tamaño de Muestra: {size}')
    plt.xlabel('Medias Normalizadas')
    plt.ylabel('Frecuencia')

plt.tight_layout()
plt.show()
```

En este ejemplo:
- Generamos datos de una distribución exponencial para representar la población.
- Realizamos muestreos de diferentes tamaños y calculamos las medias de las muestras.
- Normalizamos las medias para compararlas y observamos cómo la distribución se aproxima a una normal a medida que aumenta el tamaño de la muestra.
