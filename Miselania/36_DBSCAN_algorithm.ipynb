# **36. DBSCAN algorithm**

**DBSCAN (Density-Based Spatial Clustering of Applications with Noise):**

**Explicaci√≥n:**

DBSCAN es un algoritmo de agrupamiento basado en la densidad que divide el conjunto de datos en clusters bas√°ndose en la densidad de los puntos. A diferencia de los m√©todos de agrupamiento basados en la distancia, DBSCAN no requiere especificar el n√∫mero de clusters de antemano y puede identificar clusters de formas arbitrarias.

**Terminolog√≠a:**

1. **√âpsilon (Œµ):** Es el par√°metro que especifica la distancia m√°xima entre dos puntos para que uno sea considerado en el vecindario del otro.

2. **MinPts:** Es el n√∫mero m√≠nimo de puntos requeridos para formar un cluster. Un punto se considera parte de un cluster si hay al menos MinPts puntos dentro de una distancia Œµ de √©l.

3. **Core Points:** Son los puntos que tienen al menos MinPts puntos dentro de Œµ de su vecindario.

4. **Border Points:** Son puntos que tienen menos de MinPts puntos en su vecindario pero est√°n dentro de Œµ de un Core Point.

5. **Noise (Outliers):** Son puntos que no son Core Points ni Border Points.

**F√≥rmula Matem√°tica:**

El algoritmo DBSCAN no tiene una f√≥rmula matem√°tica √∫nica, pero la clasificaci√≥n de puntos como Core, Border o Noise se basa en la definici√≥n del vecindario Œµ de cada punto.

**Ejemplo con Gr√°fico en Python:**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons

# Generar datos de ejemplo (dos lunas)
data, _ = make_moons(n_samples=200, noise=0.05, random_state=42)

# Aplicar DBSCAN
dbscan = DBSCAN(eps=0.3, min_samples=5)
labels = dbscan.fit_predict(data)

# Personalizar la apariencia de la gr√°fica
plt.figure(figsize=(8, 6))
scatter = plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='viridis', edgecolors='k', marker='o')

# Agregar leyenda para los puntos ruidosos (-1)
legend_labels = list(set(labels) - {-1})
legend_labels.append(-1)
legend_handles = [plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=scatter.cmap(scatter.norm(label)), markersize=10) for label in legend_labels]
plt.legend(legend_handles, legend_labels, title='Etiquetas de Cluster')

# Personalizar etiquetas y t√≠tulo
plt.xlabel('Caracter√≠stica 1')
plt.ylabel('Caracter√≠stica 2')
plt.title('DBSCAN: Detecci√≥n de Clusters')

plt.show()
```

En este ejemplo, se generan datos en forma de dos lunas, y DBSCAN se aplica para detectar los clusters. El par√°metro Œµ y MinPts se ajustan seg√∫n la densidad esperada de los clusters.
