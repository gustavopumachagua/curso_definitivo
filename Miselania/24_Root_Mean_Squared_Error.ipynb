# **Root Mean Squared Error**

El Root Mean Squared Error (RMSE) es una métrica comúnmente utilizada para evaluar la precisión de un modelo de regresión. Mide la raíz cuadrada de la media de los errores al cuadrado entre las predicciones del modelo y los valores reales. Cuanto menor sea el RMSE, mejor será la capacidad predictiva del modelo.

**Terminología:**

- **Error Residual (o simplemente "Error"):** La diferencia entre el valor predicho por el modelo y el valor real observado.
- **N:** El número total de observaciones (instancias).
- **$(y_i)$:** El valor real de la variable dependiente para la observación $(i)$.
- **$(\hat{y}_i)$:** El valor predicho por el modelo para la observación $(i)$.

**Fórmula Matemática:**

La fórmula para el RMSE es la siguiente:

$RMSE = \sqrt{\frac{1}{N} \sum_{i=1}^{N} (y_i - \hat{y}_i)^2}$

**Ejemplo con Gráfico en Python:**

Vamos a calcular y visualizar el RMSE usando Python:

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
from math import sqrt

# Generar datos de ejemplo
np.random.seed(42)
y_true = np.random.rand(100) * 10  # Valores reales
y_pred = y_true + np.random.randn(100)  # Valores predichos con error aleatorio

# Calcular el RMSE
rmse = sqrt(mean_squared_error(y_true, y_pred))

# Visualizar resultados
print(f'RMSE: {rmse:.4f}')

# Gráfico de dispersión entre valores reales y predichos
plt.figure(figsize=(8, 6))
plt.scatter(y_true, y_pred, edgecolor='k', alpha=0.7, label='Puntos de Datos')
plt.plot([min(y_true), max(y_true)], [min(y_true), max(y_true)], '--k', lw=2, label='Línea de 45 Grados')

# Personalizaciones adicionales
plt.xlabel('Valor Real')
plt.ylabel('Valor Predicho')
plt.title('Gráfico de Dispersión y Línea de 45 Grados')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

Este código utiliza la biblioteca scikit-learn para calcular el RMSE y matplotlib para visualizar un gráfico de dispersión entre los valores reales y los predichos, junto con una línea de 45 grados que representa una predicción perfecta. Puedes ajustar los datos de ejemplo para explorar el impacto del RMSE en diferentes conjuntos de datos y modelos de regresión.
