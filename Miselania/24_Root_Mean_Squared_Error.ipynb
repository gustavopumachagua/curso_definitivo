# **Root Mean Squared Error**

El Root Mean Squared Error (RMSE) es una m茅trica com煤nmente utilizada para evaluar la precisi贸n de un modelo de regresi贸n. Mide la ra铆z cuadrada de la media de los errores al cuadrado entre las predicciones del modelo y los valores reales. Cuanto menor sea el RMSE, mejor ser谩 la capacidad predictiva del modelo.

**Terminolog铆a:**

- **Error Residual (o simplemente "Error"):** La diferencia entre el valor predicho por el modelo y el valor real observado.
- **N:** El n煤mero total de observaciones (instancias).
- **$(y_i)$:** El valor real de la variable dependiente para la observaci贸n $(i)$.
- **$(\hat{y}_i)$:** El valor predicho por el modelo para la observaci贸n $(i)$.

**F贸rmula Matem谩tica:**

La f贸rmula para el RMSE es la siguiente:

$RMSE = \sqrt{\frac{1}{N} \sum_{i=1}^{N} (y_i - \hat{y}_i)^2}$

**Ejemplo con Gr谩fico en Python:**

Vamos a calcular y visualizar el RMSE usando Python:

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
from math import sqrt

# Generar datos de ejemplo
np.random.seed(42)
y_true = np.random.rand(100) * 10  # Valores reales
y_pred = y_true + np.random.randn(100)  # Valores predichos con error aleatorio

# Calcular el RMSE
rmse = sqrt(mean_squared_error(y_true, y_pred))

# Visualizar resultados
print(f'RMSE: {rmse:.4f}')

# Gr谩fico de dispersi贸n entre valores reales y predichos
plt.figure(figsize=(8, 6))
plt.scatter(y_true, y_pred, edgecolor='k', alpha=0.7, label='Puntos de Datos')
plt.plot([min(y_true), max(y_true)], [min(y_true), max(y_true)], '--k', lw=2, label='L铆nea de 45 Grados')

# Personalizaciones adicionales
plt.xlabel('Valor Real')
plt.ylabel('Valor Predicho')
plt.title('Gr谩fico de Dispersi贸n y L铆nea de 45 Grados')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

Este c贸digo utiliza la biblioteca scikit-learn para calcular el RMSE y matplotlib para visualizar un gr谩fico de dispersi贸n entre los valores reales y los predichos, junto con una l铆nea de 45 grados que representa una predicci贸n perfecta. Puedes ajustar los datos de ejemplo para explorar el impacto del RMSE en diferentes conjuntos de datos y modelos de regresi贸n.
