# **50. t-test**

**Prueba t de Student (t-test): Explicaci√≥n Detallada**

**Introducci√≥n:**

La prueba t de Student, o simplemente t-test, es una t√©cnica estad√≠stica utilizada para comparar las medias de dos grupos y determinar si hay una diferencia significativa entre ellas. Es especialmente √∫til cuando se trabaja con muestras peque√±as y se asume que las poblaciones de las que provienen los grupos son aproximadamente normales.

**Terminolog√≠a:**

- **Variable Dependiente:** La caracter√≠stica que estamos midiendo en ambos grupos.

- **Variable Independiente:** La caracter√≠stica que define los grupos.

- **Hip√≥tesis Nula $(H_0)$:** Afirma que no hay diferencia significativa entre las medias de los dos grupos.

- **Hip√≥tesis Alternativa $(H_1)$ o $(H_a)$:** Afirma que hay una diferencia significativa entre las medias de los dos grupos.

**F√≥rmula Matem√°tica:**

La f√≥rmula de la prueba t para muestras independientes es:

$t = \frac{\bar{X}_1 - \bar{X}_2}{\sqrt{\frac{s_1^2}{n_1} + \frac{s_2^2}{n_2}}}$

Donde:

- $(\bar{X}_1)$ y $(\bar{X}_2)$ son las medias de los dos grupos.
  
- $(s_1)$ y $(s_2)$ son las desviaciones est√°ndar de los dos grupos.
  
- $(n_1)$ y $(n_2)$ son los tama√±os de las muestras de los dos grupos.

**Prueba de dos colas:**

En la prueba de dos colas, se busca si hay una diferencia significativa en ambas direcciones. La regi√≥n cr√≠tica se encuentra en ambos extremos de la distribuci√≥n t.

**Ejemplo con Gr√°fico en Python:**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Generar datos para dos grupos
np.random.seed(42)
group1 = np.random.normal(loc=5, scale=2, size=30)
group2 = np.random.normal(loc=7, scale=2, size=30)

# Realizar la prueba t de dos colas
t_statistic, p_value = stats.ttest_ind(group1, group2)

# Personalizar la apariencia de la gr√°fica
plt.figure(figsize=(10, 6))

# Histograma para el Grupo 1
plt.hist(group1, alpha=0.5, label='Grupo 1', color='blue', edgecolor='black', bins=15)

# Histograma para el Grupo 2
plt.hist(group2, alpha=0.5, label='Grupo 2', color='orange', edgecolor='black', bins=15)

# Personalizar etiquetas y t√≠tulo
plt.title('Distribuciones de los Dos Grupos', fontsize=16)
plt.xlabel('Valores', fontsize=14)
plt.ylabel('Frecuencia', fontsize=14)

# A√±adir leyenda
plt.legend(fontsize=12)

# Mostrar cuadr√≠cula
plt.grid(True, linestyle='--', alpha=0.5)

# A√±adir l√≠nea vertical para la prueba t
plt.axvline(x=np.mean(group1), color='blue', linestyle='--', linewidth=2, label='Media Grupo 1')
plt.axvline(x=np.mean(group2), color='orange', linestyle='--', linewidth=2, label='Media Grupo 2')

# Imprimir el resultado de la prueba t
print(f'Test Statistic: {t_statistic:.4f}')
print(f'P-value: {p_value:.4f}')

# Interpretar el resultado
if p_value < 0.05:
    plt.text(9, 5, 'Diferencia significativa', fontsize=12, color='red')
else:
    plt.text(9, 5, 'No hay diferencia significativa', fontsize=12, color='green')

# Mostrar la gr√°fica
plt.show()
```

Este ejemplo utiliza la biblioteca `scipy.stats` para realizar una prueba t de dos colas entre dos grupos generados aleatoriamente. La prueba t se realiza para determinar si hay una diferencia significativa entre las medias de los dos grupos. El resultado de la prueba y las distribuciones de los grupos se muestran en un gr√°fico.
