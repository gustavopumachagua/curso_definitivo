# **41. Affinity Propagation**

**Affinity Propagation: Explicaci√≥n Detallada**

**Introducci√≥n:**

Affinity Propagation es un algoritmo de agrupamiento (clustering) que se utiliza para identificar patrones o grupos en un conjunto de datos. A diferencia de algunos algoritmos de agrupamiento que requieren especificar el n√∫mero de cl√∫steres de antemano, Affinity Propagation determina autom√°ticamente el n√∫mero de cl√∫steres.

**Terminolog√≠a:**

1. **Similaridad (Affinity):** La medida de cu√°nto se prefiere que un punto sea el "exemplar" de otro. Se puede expresar en t√©rminos de similitud o distancia.

2. **Responsabilidad (Responsibility):** La medida de cu√°nto un punto quiere que otro sea su "exemplar". Calculado en funci√≥n de las similitudes.

3. **Disponibilidad (Availability):** La medida de la idoneidad de un punto para convertirse en el "exemplar" de otro. Se calcula en funci√≥n de las responsabilidades y las similitudes.

**F√≥rmula Matem√°tica:**

El algoritmo de Affinity Propagation actualiza iterativamente las responsabilidades y disponibilidades para converger a un conjunto de puntos "exemplares". Las f√≥rmulas para las actualizaciones son las siguientes:

$r(i, k) = s(i, k) - \max_{k' \neq k}\{a(i, k') + s(i, k')\}$

$a(i, k) = \min(0, r(k, k) + \sum_{i' \neq i, i' \neq k} \max(0, r(i', k)))$

Donde:

- $r(i, k)$ es la responsabilidad del punto $(i)$ hacia el punto $(k)$.
- $a(i, k)$ es la disponibilidad del punto $(i)$ para convertirse en el "exemplar" del punto $(k)$.
- $s(i, k)$ es la similitud entre los puntos $(i)$ y $(k)$.

**Ejemplo en Python:**

```python
from sklearn.cluster import AffinityPropagation
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

# Generar datos de ejemplo
X, _ = make_blobs(n_samples=300, centers=3, random_state=42)

# Aplicar Affinity Propagation
af = AffinityPropagation().fit(X)

# Visualizar los resultados
cluster_centers_indices = af.cluster_centers_indices_
labels = af.labels_

# Plot
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', edgecolors='k')
plt.title('Affinity Propagation: Clustering')
plt.show()
```

En este ejemplo, se generan datos de tres grupos utilizando `make_blobs`. Luego, se aplica Affinity Propagation y se visualizan los resultados con colores diferentes para cada cl√∫ster.
