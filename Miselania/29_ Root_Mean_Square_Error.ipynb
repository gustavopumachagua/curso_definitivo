# **29. Root Mean Square Error**

El Root Mean Square Error (Error Cuadr√°tico Medio) es una m√©trica com√∫nmente utilizada para evaluar la precisi√≥n de un modelo de regresi√≥n. Mide la diferencia entre los valores predichos por el modelo y los valores reales de la variable dependiente. La ra√≠z cuadrada del error cuadr√°tico medio proporciona una medida de la dispersi√≥n de los errores en unidades originales.

**Terminolog√≠a:**

1. **Error Cuadr√°tico:** La diferencia al cuadrado entre la predicci√≥n del modelo y el valor real.

2. **Error Cuadr√°tico Medio (MSE):** El promedio de los errores cuadr√°ticos.

3. **Root Mean Square Error (RMSE):** La ra√≠z cuadrada del MSE. Representa la desviaci√≥n est√°ndar de los errores.

**F√≥rmula Matem√°tica:**

La f√≥rmula del RMSE se expresa matem√°ticamente de la siguiente manera:

$RMSE = \sqrt{\frac{1}{n}\sum_{i=1}^{n}(y_i - \hat{y}_i)^2}$

- $( n )$ es el n√∫mero total de observaciones.
- $( y_i )$ es el valor real de la observaci√≥n $( i )$.
- $\hat{y}_i$ es el valor predicho para la observaci√≥n $( i )$.

**Interpretaci√≥n:**

El RMSE cuantifica la magnitud promedio de los errores de predicci√≥n. Un RMSE m√°s bajo indica una mejor concordancia entre las predicciones del modelo y los valores reales.

**Ejemplo con Gr√°fico en Python:**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error

# Datos de ejemplo
y_true = np.array([2, 4, 5, 4, 5])
y_pred = np.array([1.5, 3.5, 4.5, 5, 5.5])

# Calcular el Error Cuadr√°tico Medio (MSE)
mse = mean_squared_error(y_true, y_pred)

# Calcular el RMSE
rmse = np.sqrt(mse)

# Visualizar los datos y las predicciones con l√≠neas de conexi√≥n
plt.figure(figsize=(8, 6))
plt.plot(range(len(y_true)), y_true, marker='o', linestyle='-', label='Real', color='blue')
plt.plot(range(len(y_pred)), y_pred, marker='o', linestyle='-', label='Predicci√≥n', color='orange')

# L√≠neas conectando los puntos reales y predichos
for i in range(len(y_true)):
    plt.plot([i, i], [y_true[i], y_pred[i]], color='gray', linestyle='--', linewidth=1)

plt.legend()
plt.xlabel('Observaciones')
plt.ylabel('Valores')
plt.title(f'RMSE: {rmse:.2f}')
plt.show()
```

Este ejemplo ilustra c√≥mo calcular el RMSE utilizando la biblioteca scikit-learn en Python. Se comparan los valores reales con las predicciones y se muestra un gr√°fico que indica visualmente la diferencia entre ellos. La m√©trica RMSE se calcula para evaluar la calidad del modelo de regresi√≥n.
