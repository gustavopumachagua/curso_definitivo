# **32. Time Series**

Las series temporales son conjuntos de datos donde las observaciones se recopilan, registran o registran en función del tiempo. Este tipo de datos se encuentra comúnmente en diversas áreas, como finanzas, economía, meteorología y muchas otras disciplinas. Las series temporales suelen presentar patrones temporales que pueden ser explorados y modelados.

**Terminología:**

1. **Punto de Datos Temporal (Observación):** Cada entrada en la serie temporal, asociada con un momento específico en el tiempo.

2. **Frecuencia:** El intervalo de tiempo entre dos puntos de datos consecutivos (por ejemplo, diario, mensual, anual).

3. **Tendencia:** La dirección general de la serie temporal, es decir, si los valores tienden a aumentar, disminuir o permanecer constantes con el tiempo.

4. **Estacionalidad:** Patrones recurrentes o ciclos en la serie temporal que se repiten a intervalos regulares.

5. **Ruido:** Variabilidad aleatoria o irregularidad en la serie temporal que no sigue un patrón discernible.

**Modelos de Series Temporales:**

1. **Modelo Aditivo:** Una serie temporal se puede descomponer en componentes aditivos, como tendencia, estacionalidad y términos residuales.

2. **Modelo Multiplicativo:** Similar al modelo aditivo, pero los componentes se multiplican entre sí en lugar de sumarse.

**Fórmulas Matemáticas:**

1. **Media Móvil (Moving Average):**

   $MA_t = \frac{1}{k} \sum_{i=1}^{k} X_{t-i}$

2. **Suavización Exponencial (Exponential Smoothing):**

   $\hat{y}_{t+1} = \alpha X_t + (1-\alpha) \hat{y}_t$

   donde:

    $( \hat{y}_{t+1} )$ es la predicción para el siguiente punto de datos, $( X_t )$ es el valor actual y $( \hat{y}_t )$ es la predicción anterior.

**Ejemplo con Gráfico en Python:**

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.dates import date2num

# Crear una serie temporal ficticia
date_rng = pd.date_range(start='2023-01-01', end='2023-12-31', freq='D')
time_series = pd.Series(range(len(date_rng)), index=date_rng)

# Personalizar la gráfica
plt.figure(figsize=(10, 6))

# Plotear la serie temporal con un estilo de línea y color específicos
plt.plot(date2num(time_series.index), time_series, linestyle='-', color='b', label='Datos de la serie temporal')

# Personalizar el título y las etiquetas de los ejes
plt.title('Ejemplo de Serie Temporal', fontsize=16)
plt.xlabel('Fecha', fontsize=14)
plt.ylabel('Valor', fontsize=14)

# Mostrar la leyenda
plt.legend()

# Personalizar la cuadrícula
plt.grid(True, linestyle='--', alpha=0.7)

# Rotar las fechas en el eje x para mayor legibilidad
plt.xticks(rotation=45)

# Añadir líneas de referencia, por ejemplo, una línea horizontal en y=500
plt.axhline(y=500, color='r', linestyle='--', label='Línea de referencia')

# Añadir anotaciones
plt.annotate('Punto importante', xy=(date2num(pd.to_datetime('2023-06-15')), 300),
             xytext=(date2num(pd.to_datetime('2023-03-15')), 450),
             arrowprops=dict(facecolor='black', shrink=0.05), fontsize=12)

# Mostrar la gráfica
plt.show()
```

Este código en Python utiliza la biblioteca pandas para crear una serie temporal ficticia y la biblioteca matplotlib para visualizarla en un gráfico. En la práctica, el análisis de series temporales y la modelización implican técnicas más avanzadas, como modelos ARIMA, SARIMA, suavización exponencial y otros enfoques.
