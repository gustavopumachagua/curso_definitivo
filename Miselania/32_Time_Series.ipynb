# **32. Time Series**

Las series temporales son conjuntos de datos donde las observaciones se recopilan, registran o registran en funci贸n del tiempo. Este tipo de datos se encuentra com煤nmente en diversas 谩reas, como finanzas, econom铆a, meteorolog铆a y muchas otras disciplinas. Las series temporales suelen presentar patrones temporales que pueden ser explorados y modelados.

**Terminolog铆a:**

1. **Punto de Datos Temporal (Observaci贸n):** Cada entrada en la serie temporal, asociada con un momento espec铆fico en el tiempo.

2. **Frecuencia:** El intervalo de tiempo entre dos puntos de datos consecutivos (por ejemplo, diario, mensual, anual).

3. **Tendencia:** La direcci贸n general de la serie temporal, es decir, si los valores tienden a aumentar, disminuir o permanecer constantes con el tiempo.

4. **Estacionalidad:** Patrones recurrentes o ciclos en la serie temporal que se repiten a intervalos regulares.

5. **Ruido:** Variabilidad aleatoria o irregularidad en la serie temporal que no sigue un patr贸n discernible.

**Modelos de Series Temporales:**

1. **Modelo Aditivo:** Una serie temporal se puede descomponer en componentes aditivos, como tendencia, estacionalidad y t茅rminos residuales.

2. **Modelo Multiplicativo:** Similar al modelo aditivo, pero los componentes se multiplican entre s铆 en lugar de sumarse.

**F贸rmulas Matem谩ticas:**

1. **Media M贸vil (Moving Average):**

   $MA_t = \frac{1}{k} \sum_{i=1}^{k} X_{t-i}$

2. **Suavizaci贸n Exponencial (Exponential Smoothing):**

   $\hat{y}_{t+1} = \alpha X_t + (1-\alpha) \hat{y}_t$

   donde:

    $( \hat{y}_{t+1} )$ es la predicci贸n para el siguiente punto de datos, $( X_t )$ es el valor actual y $( \hat{y}_t )$ es la predicci贸n anterior.

**Ejemplo con Gr谩fico en Python:**

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.dates import date2num

# Crear una serie temporal ficticia
date_rng = pd.date_range(start='2023-01-01', end='2023-12-31', freq='D')
time_series = pd.Series(range(len(date_rng)), index=date_rng)

# Personalizar la gr谩fica
plt.figure(figsize=(10, 6))

# Plotear la serie temporal con un estilo de l铆nea y color espec铆ficos
plt.plot(date2num(time_series.index), time_series, linestyle='-', color='b', label='Datos de la serie temporal')

# Personalizar el t铆tulo y las etiquetas de los ejes
plt.title('Ejemplo de Serie Temporal', fontsize=16)
plt.xlabel('Fecha', fontsize=14)
plt.ylabel('Valor', fontsize=14)

# Mostrar la leyenda
plt.legend()

# Personalizar la cuadr铆cula
plt.grid(True, linestyle='--', alpha=0.7)

# Rotar las fechas en el eje x para mayor legibilidad
plt.xticks(rotation=45)

# A帽adir l铆neas de referencia, por ejemplo, una l铆nea horizontal en y=500
plt.axhline(y=500, color='r', linestyle='--', label='L铆nea de referencia')

# A帽adir anotaciones
plt.annotate('Punto importante', xy=(date2num(pd.to_datetime('2023-06-15')), 300),
             xytext=(date2num(pd.to_datetime('2023-03-15')), 450),
             arrowprops=dict(facecolor='black', shrink=0.05), fontsize=12)

# Mostrar la gr谩fica
plt.show()
```
