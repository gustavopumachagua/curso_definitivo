# **46. Binomial Distribution**

**Distribuci√≥n Binomial: Explicaci√≥n Detallada**

**Terminolog√≠a:**

- **Ensayo Bernoulli:** Es un experimento con dos resultados posibles, generalmente denominados √©xito y fracaso.
- **N√∫mero de Ensayos $(n)$:** Representa la cantidad total de ensayos en una secuencia de ensayos Bernoulli.
- **Probabilidad de √âxito $(p)$:** Es la probabilidad de que ocurra un √©xito en un solo ensayo Bernoulli.
- **Variable Aleatoria Binomial $(X)$:** Representa el n√∫mero total de √©xitos en una secuencia de $(n)$ ensayos Bernoulli.

**F√≥rmula Matem√°tica:**

La funci√≥n de masa de probabilidad (PMF) de la distribuci√≥n binomial se define como:

$P(X = k) = \binom{n}{k} \cdot p^k \cdot (1 - p)^{n - k}$

Donde:

- $P(X = k)$ es la probabilidad de tener $(k)$ √©xitos en $(n)$ ensayos.
- $\binom{n}{k}$ es el coeficiente binomial, que representa el n√∫mero de formas de elegir $(k)$ √©xitos de $(n)$ ensayos.
- $( p )$ es la probabilidad de √©xito en un solo ensayo.
- $( k )$ es el n√∫mero de √©xitos deseados.
- $( n )$ es el n√∫mero total de ensayos.

**Ejemplo con C√≥digo Python:**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Par√°metros de la distribuci√≥n binomial
n = 10  # N√∫mero de ensayos
p = 0.5  # Probabilidad de √©xito en un solo ensayo

# Crear una secuencia de valores para la variable aleatoria binomial
k_values = np.arange(0, n+1)

# Calcular las probabilidades para cada valor de k
probabilities = binom.pmf(k_values, n, p)

# Visualizar la distribuci√≥n binomial
plt.bar(k_values, probabilities, align='center', alpha=0.7, color='blue', edgecolor='k')
plt.title('Distribuci√≥n Binomial')
plt.xlabel('N√∫mero de √âxitos (k)')
plt.ylabel('Probabilidad')
plt.show()
```

Este c√≥digo utiliza la funci√≥n de masa de probabilidad (PMF) de la distribuci√≥n binomial para visualizar la probabilidad de cada n√∫mero de √©xitos $(k)$ en una secuencia de $(n)$ ensayos Bernoulli. El gr√°fico resultante es una distribuci√≥n binomial.
