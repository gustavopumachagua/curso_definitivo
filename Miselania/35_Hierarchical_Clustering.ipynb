# **35. Hierarchical Clustering**

**Clustering Jer√°rquico:**

**Explicaci√≥n:**

El clustering jer√°rquico es un m√©todo de agrupamiento que construye una jerarqu√≠a de clusters. En este enfoque, los datos se agrupan en varios niveles, formando un √°rbol o dendrograma. Puede ser divisivo (top-down), comenzando con todos los datos en un solo cluster y dividi√©ndolos en clusters m√°s peque√±os, o aglomerativo (bottom-up), comenzando con clusters individuales y fusion√°ndolos para formar clusters m√°s grandes.

**Terminolog√≠a:**

1. **Enlace (Linkage):** Es el criterio utilizado para medir la distancia entre dos clusters. Algunos m√©todos comunes de enlace incluyen:
   - **Enlace Simple:** Mide la distancia entre los elementos m√°s cercanos de los dos clusters.
   - **Enlace Completo:** Mide la distancia entre los elementos m√°s lejanos de los dos clusters.
   - **Enlace Promedio:** Mide la distancia promedio entre todos los pares de elementos de los dos clusters.

2. **Dendrograma:** Es una representaci√≥n visual de la jerarqu√≠a de clusters. Cada nodo en el dendrograma representa un cluster, y las hojas del √°rbol representan los elementos individuales.

**F√≥rmula Matem√°tica:**

La f√≥rmula exacta puede variar seg√∫n el m√©todo de enlace utilizado, pero en general, la distancia entre dos clusters se calcula de acuerdo con alg√∫n criterio de enlace.

**Ejemplo con Gr√°fico en Python:**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage

# Datos de ejemplo
np.random.seed(42)
data = np.random.rand(10, 2)

# M√©todo de enlace y construcci√≥n del dendrograma
linkage_matrix = linkage(data, method='ward')

# Personalizar la apariencia del dendrograma
plt.figure(figsize=(10, 6))
dendrogram(linkage_matrix, 
           color_threshold=1.5,  # Ajusta este umbral seg√∫n tus necesidades
           leaf_font_size=10,    # Tama√±o de fuente para etiquetas de hojas
           above_threshold_color='gray',  # Color para las l√≠neas por encima del umbral
           truncate_mode='lastp',  # Puedes ajustar el modo de truncado seg√∫n tus necesidades
           p=10,                  # N√∫mero de nodos a mostrar en el √∫ltimo p nivel del dendrograma
           show_leaf_counts=True, # Muestra el n√∫mero de observaciones en cada hoja
           orientation='top',     # Puedes ajustar la orientaci√≥n del dendrograma
           labels=list(range(1, 11)))  # Puedes proporcionar etiquetas personalizadas

plt.title('Dendrograma Personalizado')
plt.xlabel('√çndice de Datos')
plt.ylabel('Distancia')

# Puedes agregar l√≠neas horizontales para indicar el umbral
plt.axhline(y=1.5, color='r', linestyle='--', linewidth=2, label='Umbral')

# Mostrar la leyenda
plt.legend()

plt.show()
```

En este ejemplo, se generan datos aleatorios y se utiliza el m√©todo de enlace "ward" para construir el dendrograma. El dendrograma muestra c√≥mo los datos se agrupan jer√°rquicamente, y la altura en la que se fusionan los clusters indica la distancia a la que se agruparon.
