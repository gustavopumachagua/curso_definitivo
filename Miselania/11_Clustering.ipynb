# **11. Clustering**

![clustering](../img/clustering.png "clustering")

El clustering (o agrupamiento) es una t√©cnica en el campo del aprendizaje no supervisado que consiste en dividir un conjunto de datos en grupos o cl√∫steres basados en la similitud entre los elementos. El objetivo principal es reunir objetos similares y separar aquellos que son diferentes. Esta t√©cnica es ampliamente utilizada en diversos campos, como miner√≠a de datos, an√°lisis de patrones, reconocimiento de patrones, entre otros.

**Terminolog√≠a:**

1. **Cl√∫ster:** Un grupo de objetos similares que comparten caracter√≠sticas comunes.
2. **Centroide:** El punto central de un cl√∫ster, que representa el centro de ese grupo.
3. **Similitud/Distancia:** Una medida que cuantifica la similitud o distancia entre dos puntos en funci√≥n de ciertos atributos.
4. **Algoritmo de Clustering:** Un conjunto de reglas o procedimientos para dividir un conjunto de datos en cl√∫steres.

**Formula Matem√°tica:**

En el clustering, el concepto clave es la medida de similitud o distancia entre los puntos. Algunas medidas comunes incluyen la distancia euclidiana o la correlaci√≥n. La f√≥rmula general de la distancia euclidiana entre dos puntos $(p)$ y $(q)$ en un espacio bidimensional ser√≠a:

$\text{Distancia Euclidiana}(p, q) = \sqrt{(q_1 - p_1)^2 + (q_2 - p_2)^2}$

**Ejemplo con K-Means en Python:**

Vamos a utilizar el algoritmo K-Means, uno de los algoritmos de clustering m√°s populares, para agrupar datos de manera sencilla en Python:

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

# Crear datos de ejemplo
X, _ = make_blobs(n_samples=300, centers=4, random_state=42, cluster_std=1.0)

# Aplicar K-Means con 4 cl√∫steres
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)  # Establece el valor expl√≠citamente
kmeans.fit(X)

# Obtener las etiquetas de los cl√∫steres y los centroides
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

# Visualizar los cl√∫steres y los centroides
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', edgecolors='k', s=50)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='X', s=200, label='Centroides')
plt.title('Ejemplo de Clustering con K-Means')
plt.xlabel('Caracter√≠stica 1')
plt.ylabel('Caracter√≠stica 2')
plt.legend()
plt.show()
```

Este c√≥digo utiliza K-Means para agrupar datos en 4 cl√∫steres y luego visualiza los puntos de datos coloreados seg√∫n sus asignaciones de cl√∫ster y marca los centroides de cada cl√∫ster en rojo.

**Resumen:**

El clustering es una herramienta poderosa para explorar y entender la estructura de datos no etiquetados. El K-Means es un ejemplo com√∫n, pero hay muchos otros algoritmos de clustering, y la elecci√≥n del algoritmo depende del tipo de datos y el problema espec√≠fico.

### **Ejemplo**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

# Crear un conjunto de datos sint√©tico con clusters
X, y = make_blobs(n_samples=300, centers=4, random_state=42)

# Crear un DataFrame
df = pd.DataFrame(data=X, columns=['Feature1', 'Feature2'])
df
```

```python
# Inicializar el modelo K-Means
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)

# Entrenar el modelo
kmeans.fit(X)

# Obtener las etiquetas de cluster y los centros
labels = kmeans.labels_
centers = kmeans.cluster_centers_

# A√±adir las etiquetas de cluster al DataFrame
df['Cluster'] = labels

# Visualizar los datos y los clusters
plt.figure(figsize=(10, 6))

# Scatter plot para los datos coloreados por cluster
plt.scatter(df['Feature1'], df['Feature2'], c=df['Cluster'], cmap='viridis', edgecolors='k', s=50)
plt.scatter(centers[:, 0], centers[:, 1], c='red', marker='X', s=200, label='Centroids')

plt.title('Clustering con K-Means')
plt.xlabel('Feature1')
plt.ylabel('Feature2')
plt.legend()
plt.show()
```
