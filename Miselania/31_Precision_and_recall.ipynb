# **31. Precision and recall**

**Precision y Recall (Precisión y Sensibilidad):**

La precisión y la sensibilidad (recall) son métricas comúnmente utilizadas para evaluar el rendimiento de un modelo de clasificación, especialmente cuando se trata con clases desbalanceadas.

**Terminología:**

1. **Verdaderos Positivos (TP):** Son los casos en los que el modelo predice correctamente la clase positiva.

2. **Falsos Positivos (FP):** Son los casos en los que el modelo predice incorrectamente la clase positiva cuando la clase real es negativa.

3. **Falsos Negativos (FN):** Son los casos en los que el modelo predice incorrectamente la clase negativa cuando la clase real es positiva.

4. **Precisión (Precision):** Mide la proporción de predicciones positivas del modelo que fueron realmente positivas. Se calcula como $\frac{TP}{TP + FP}$.

5. **Sensibilidad o Recall:** Mide la proporción de casos positivos reales que fueron identificados correctamente por el modelo. Se calcula como $\frac{TP}{TP + FN}$.

**Fórmulas Matemáticas:**

1. **Precisión (Precision):**

$Precision = \frac{TP}{TP + FP}$

2. **Sensibilidad o Recall:**

$Recall = \frac{TP}{TP + FN}$

**Ejemplo con Gráfico en Python:**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score

# Datos de ejemplo
y_true = np.array([2, 4, 5, 4, 5])
y_pred = np.array([1.5, 3.5, 4.5, 5, 5.5])

# Calcular el R2 Score
r2 = r2_score(y_true, y_pred)

# Visualizar los datos y las predicciones con líneas de conexión
plt.figure(figsize=(8, 6))
plt.plot(range(len(y_true)), y_true, marker='o', linestyle='-', label='Real', color='blue')
plt.plot(range(len(y_pred)), y_pred, marker='o', linestyle='-', label='Predicción', color='orange')

# Líneas conectando los puntos reales y predichos
for i in range(len(y_true)):
    plt.plot([i, i], [y_true[i], y_pred[i]], color='gray', linestyle='--', linewidth=1)

# Resaltar visualmente el R2 Score
plt.title(f'R2 Score: {r2:.2f}', color='green', fontsize=16)

plt.legend()
plt.xlabel('Observaciones')
plt.ylabel('Valores')
plt.show()
```

En este ejemplo, se utiliza la biblioteca scikit-learn en Python para calcular la matriz de confusión, la precisión y la sensibilidad. La matriz de confusión se visualiza con seaborn, y se muestran las métricas de precisión y sensibilidad en el título del gráfico. Estas métricas proporcionan información detallada sobre el rendimiento del modelo de clasificación.
