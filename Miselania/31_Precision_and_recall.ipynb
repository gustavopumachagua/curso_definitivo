# **31. Precision and recall**

**Precision y Recall (Precisi贸n y Sensibilidad):**

La precisi贸n y la sensibilidad (recall) son m茅tricas com煤nmente utilizadas para evaluar el rendimiento de un modelo de clasificaci贸n, especialmente cuando se trata con clases desbalanceadas.

**Terminolog铆a:**

1. **Verdaderos Positivos (TP):** Son los casos en los que el modelo predice correctamente la clase positiva.

2. **Falsos Positivos (FP):** Son los casos en los que el modelo predice incorrectamente la clase positiva cuando la clase real es negativa.

3. **Falsos Negativos (FN):** Son los casos en los que el modelo predice incorrectamente la clase negativa cuando la clase real es positiva.

4. **Precisi贸n (Precision):** Mide la proporci贸n de predicciones positivas del modelo que fueron realmente positivas. Se calcula como $\frac{TP}{TP + FP}$.

5. **Sensibilidad o Recall:** Mide la proporci贸n de casos positivos reales que fueron identificados correctamente por el modelo. Se calcula como $\frac{TP}{TP + FN}$.

**F贸rmulas Matem谩ticas:**

1. **Precisi贸n (Precision):**

$Precision = \frac{TP}{TP + FP}$

2. **Sensibilidad o Recall:**

$Recall = \frac{TP}{TP + FN}$

**Ejemplo con Gr谩fico en Python:**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score

# Datos de ejemplo
y_true = np.array([2, 4, 5, 4, 5])
y_pred = np.array([1.5, 3.5, 4.5, 5, 5.5])

# Calcular el R2 Score
r2 = r2_score(y_true, y_pred)

# Visualizar los datos y las predicciones con l铆neas de conexi贸n
plt.figure(figsize=(8, 6))
plt.plot(range(len(y_true)), y_true, marker='o', linestyle='-', label='Real', color='blue')
plt.plot(range(len(y_pred)), y_pred, marker='o', linestyle='-', label='Predicci贸n', color='orange')

# L铆neas conectando los puntos reales y predichos
for i in range(len(y_true)):
    plt.plot([i, i], [y_true[i], y_pred[i]], color='gray', linestyle='--', linewidth=1)

# Resaltar visualmente el R2 Score
plt.title(f'R2 Score: {r2:.2f}', color='green', fontsize=16)

plt.legend()
plt.xlabel('Observaciones')
plt.ylabel('Valores')
plt.show()
```

En este ejemplo, se utiliza la biblioteca scikit-learn en Python para calcular la matriz de confusi贸n, la precisi贸n y la sensibilidad. La matriz de confusi贸n se visualiza con seaborn, y se muestran las m茅tricas de precisi贸n y sensibilidad en el t铆tulo del gr谩fico. Estas m茅tricas proporcionan informaci贸n detallada sobre el rendimiento del modelo de clasificaci贸n.
