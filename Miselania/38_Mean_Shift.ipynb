# **38. Mean Shift**

**Mean Shift: Explicaci√≥n Detallada**

**Introducci√≥n:**
Mean Shift es un algoritmo de clustering no param√©trico utilizado para la segmentaci√≥n de datos. La idea principal detr√°s del Mean Shift es encontrar los modos o picos de densidad en el espacio de caracter√≠sticas, agrupando puntos de datos en regiones de alta densidad. Es especialmente √∫til en problemas de clustering donde la forma y el n√∫mero de clusters no est√°n predefinidos.

**Terminolog√≠a:**

1. **Kernel:** En Mean Shift, se utiliza un kernel para estimar la densidad de probabilidad en el espacio de caracter√≠sticas. El kernel determina la influencia de los puntos circundantes en la estimaci√≥n de la densidad en un punto espec√≠fico.

2. **Banda de B√∫squeda (Bandwidth):** Es un par√°metro crucial que define la distancia alrededor de cada punto que se considera al calcular la media ponderada.

3. **Moda:** Representa la posici√≥n en el espacio de caracter√≠sticas donde la densidad de puntos es m√°xima.

**F√≥rmula Matem√°tica:**
La f√≥rmula para el Mean Shift en 2D se puede expresar de la siguiente manera:

$m(x) = \frac{\sum_{i=1}^{N} K\left(\frac{x - x_i}{h}\right) \cdot x_i}{\sum_{i=1}^{N} K\left(\frac{x - x_i}{h}\right)}$

- $m(x)$: Nueva posici√≥n calculada para el punto $(x)$.
- $(K)$: Kernel, que especifica la influencia de los puntos vecinos.
- $(h)$: Bandwidth, controla la distancia para considerar en la b√∫squeda.

**Ejemplo en Python:**

```python
import numpy as np
from sklearn.cluster import MeanShift, estimate_bandwidth
import matplotlib.pyplot as plt

# Generar datos de ejemplo
np.random.seed(0)
X = np.concatenate([np.random.normal(loc=i, scale=0.5, size=(100, 2)) for i in range(4)])

# Estimar la banda de b√∫squeda
bandwidth = estimate_bandwidth(X, quantile=0.2, n_samples=500)

# Crear y ajustar el modelo Mean Shift
mean_shift_model = MeanShift(bandwidth=bandwidth)
mean_shift_model.fit(X)

# Obtener etiquetas y centros de los clusters
labels = mean_shift_model.labels_
cluster_centers = mean_shift_model.cluster_centers_

# Personalizar la apariencia de la gr√°fica
plt.figure(figsize=(8, 6))
scatter = plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', edgecolors='k', marker='o')
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], c='red', marker='x', s=200, label='Centros de Clusters')

# Personalizar leyenda
legend_handles = [plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=scatter.cmap(scatter.norm(label)), markersize=10) for label in set(labels)]
legend_handles.append(plt.Line2D([0], [0], marker='x', color='red', markerfacecolor='red', markersize=10, label='Centros de Clusters'))
plt.legend(legend_handles, list(set(labels)) + ['Centros'], title='Etiquetas de Cluster')

# Personalizar etiquetas y t√≠tulo
plt.title('Mean Shift Clustering')
plt.xlabel('Caracter√≠stica 1')
plt.ylabel('Caracter√≠stica 2')

plt.show()
```

Este c√≥digo en Python utiliza la biblioteca scikit-learn para implementar Mean Shift en datos de ejemplo bidimensionales y visualiza los resultados.
