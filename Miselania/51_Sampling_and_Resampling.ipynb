# **51. Sampling & Resampling**

**Muestreo y Remuestreo: Explicaci√≥n Detallada**

**Muestreo:**

En estad√≠sticas, el muestreo se refiere al proceso de seleccionar un subconjunto de elementos de una poblaci√≥n con el objetivo de realizar inferencias sobre toda la poblaci√≥n. El muestreo es crucial cuando trabajar con grandes conjuntos de datos o cuando es impr√°ctico o costoso analizar toda la poblaci√≥n.

**Terminolog√≠a:**

- **Poblaci√≥n:** El conjunto completo de elementos o individuos que tienen una caracter√≠stica com√∫n y son el objeto de estudio.

- **Muestra:** Un subconjunto representativo de la poblaci√≥n que se selecciona para el an√°lisis.

- **Muestreo Aleatorio Simple:** Un m√©todo en el que cada individuo de la poblaci√≥n tiene la misma probabilidad de ser seleccionado.

- **Muestreo Estratificado:** Divide la poblaci√≥n en subgrupos (estratos) y luego selecciona muestras de cada estrato.

**F√≥rmula Matem√°tica para la Media Muestral $(\bar{X})$:**

$\bar{X} = \frac{\sum_{i=1}^{n} X_i}{n}$

Donde:

- $\bar{X}$ es la media muestral.
  
- $( X_i )$ son los valores individuales en la muestra.
  
- $( n )$ es el tama√±o de la muestra.

**Remuestreo:**

El remuestreo es una t√©cnica que implica tomar repetidamente muestras de un conjunto de datos existente con el prop√≥sito de realizar inferencias adicionales sobre la poblaci√≥n subyacente.

**Terminolog√≠a:**

- **Bootstrap:** Un m√©todo de remuestreo que implica tomar muestras con reemplazo de la muestra original.

- **Jackknife:** Un m√©todo de remuestreo que implica omitir un solo elemento de la muestra original a la vez para crear conjuntos de datos m√°s peque√±os.

**Ejemplo con Gr√°fico en Python:**

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos de una distribuci√≥n normal
np.random.seed(42)
population_data = np.random.normal(loc=50, scale=10, size=1000)

# Muestreo aleatorio simple
sample = np.random.choice(population_data, size=100, replace=False)

# Bootstrap (remuestreo con reemplazo)
bootstrap_samples = [np.random.choice(sample, size=len(sample), replace=True) for _ in range(1000)]
bootstrap_means = np.mean(bootstrap_samples, axis=1)

# Personalizar la apariencia de la gr√°fica
plt.figure(figsize=(16, 6))

# Histograma de la poblaci√≥n
plt.subplot(1, 3, 1)
plt.hist(population_data, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Poblaci√≥n')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.grid(True, linestyle='--', alpha=0.5)

# Histograma de la muestra
plt.subplot(1, 3, 2)
plt.hist(sample, bins=30, color='lightgreen', edgecolor='black', alpha=0.7)
plt.title('Muestra')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.grid(True, linestyle='--', alpha=0.5)

# Histograma de las medias bootstrap
plt.subplot(1, 3, 3)
plt.hist(bootstrap_means, bins=30, color='coral', edgecolor='black', alpha=0.7)
plt.title('Bootstrap Means')
plt.xlabel('Valor Medio Bootstrap')
plt.ylabel('Frecuencia')
plt.grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

En este ejemplo:
- Generamos datos de una distribuci√≥n normal para representar la poblaci√≥n.
- Realizamos un muestreo aleatorio simple para obtener una muestra.
- Aplicamos el m√©todo bootstrap para generar m√∫ltiples muestras (con reemplazo) y calculamos las medias muestrales.
- Visualizamos la poblaci√≥n, la muestra y la distribuci√≥n de las medias bootstrap.
