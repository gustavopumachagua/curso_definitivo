# **40. Correlation**

**Correlaci√≥n: Explicaci√≥n Detallada**

**Introducci√≥n:**

La correlaci√≥n es una medida estad√≠stica que describe la relaci√≥n entre dos variables. Indica c√≥mo los cambios en una variable est√°n asociados con los cambios en otra. La correlaci√≥n no implica causalidad, pero simplemente cuantifica la fuerza y la direcci√≥n de la relaci√≥n lineal entre dos variables.

**Terminolog√≠a:**

1. **Coeficiente de Correlaci√≥n (r):** Representa la fuerza y direcci√≥n de la relaci√≥n lineal entre dos variables. Var√≠a entre -1 y 1.
   - **r = 1:** Correlaci√≥n positiva perfecta.
   - **r = -1:** Correlaci√≥n negativa perfecta.
   - **r ‚âà 0:** No hay correlaci√≥n.

2. **Matriz de Correlaci√≥n:** Una tabla que muestra los coeficientes de correlaci√≥n entre varias variables.

**F√≥rmula Matem√°tica:**

El coeficiente de correlaci√≥n (r) se calcula mediante la siguiente f√≥rmula:
$r = \frac{\sum{(X_i - \bar{X})(Y_i - \bar{Y})}}{\sqrt{\sum{(X_i - \bar{X})^2} \sum{(Y_i - \bar{Y})^2}}}$

Donde:

- $(X_i)$ y $(Y_i)$ son los valores individuales de las dos variables.
- $(\bar{X})$ y $(\bar{Y})$ son las medias de las dos variables.

**Ejemplo en Python:**

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos de ejemplo
np.random.seed(42)
X = np.random.rand(50)
Y = 2 * X + 1 + 0.1 * np.random.randn(50)  # Relaci√≥n lineal con ruido

# Calcular el coeficiente de correlaci√≥n
correlation_coef = np.corrcoef(X, Y)[0, 1]

# Personalizar la apariencia de la gr√°fica
plt.figure(figsize=(8, 6))
plt.scatter(X, Y, label=f'Correlaci√≥n = {correlation_coef:.2f}', color='blue', marker='o', edgecolors='black', alpha=0.7)

# Personalizar etiquetas y t√≠tulo
plt.xlabel('Variable X', fontsize=14)
plt.ylabel('Variable Y', fontsize=14)
plt.title('Correlaci√≥n entre X e Y', fontsize=16)

# Personalizar la leyenda
plt.legend(fontsize=12)

# Mostrar la cuadr√≠cula
plt.grid(True, linestyle='--', alpha=0.5)

plt.show()
```

En este ejemplo, se generan datos con una relaci√≥n lineal y se visualiza la dispersi√≥n de los puntos. El coeficiente de correlaci√≥n se calcula usando `np.corrcoef`.
