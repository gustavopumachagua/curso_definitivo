# **37. Distributions**

## **Uniform Distribution**

**Explicación:**

La distribución uniforme es un tipo de distribución de probabilidad donde cada valor en el rango posible tiene la misma probabilidad de ocurrir. En otras palabras, todos los eventos tienen igual probabilidad de suceder. Es útil en situaciones en las que no hay razón para favorecer un valor sobre otro dentro de un rango específico.

**Terminología:**

1. **Variable Aleatoria Continua (X):** Representa la variable que sigue una distribución uniforme.

2. **Parámetros a y b:** Son los parámetros que definen el rango de la distribución. La distribución uniforme se denota como U(a, b).

3. **Función de Densidad de Probabilidad (PDF):** Para una distribución uniforme, la PDF es constante en el intervalo [a, b] y es 0 fuera de ese intervalo.

**Fórmula Matemática:**

La función de densidad de probabilidad (PDF) para una distribución uniforme U(a, b) se define como:

$f(x) = \frac{1}{b - a} \text{ para } a \leq x \leq b$

$f(x) = 0 \text{ para } x < a \text{ o } x > b$

Donde:

- $( a )$ y $( b )$ son los parámetros que definen el intervalo.
- $f(x)$ es la probabilidad de que la variable aleatoria $( X )$ tome el valor $( x )$.

**Ejemplo con Gráfico en Python:**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform

# Definir parámetros a y b
a = 0
b = 10

# Generar datos de ejemplo de una distribución uniforme
data_uniform = uniform.rvs(size=10000, loc=a, scale=b-a)

# Visualizar el histograma
plt.hist(data_uniform, bins=50, density=True, alpha=0.7, color='g', edgecolor='k')
plt.plot([a, b], [1/(b-a), 1/(b-a)], 'r--', lw=2, label='PDF')
plt.title('Distribución Uniforme U(0, 10)')
plt.xlabel('Valor')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.show()
```

En este ejemplo, se generan datos de una distribución uniforme U(0, 10) y se visualiza el histograma. La línea roja representa la función de densidad de probabilidad (PDF), que es constante en el intervalo [0, 10].

## **Normal Distribution**

**Distribución Normal (Gaussiana):**

**Explicación:**

La distribución normal, también conocida como distribución gaussiana, es una de las distribuciones de probabilidad más importantes en estadística. Es una distribución continua que se caracteriza por ser simétrica alrededor de su media, con la mayor densidad de probabilidad en la media y una forma de campana.

**Terminología:**

1. **Variable Aleatoria Continua (X):** Representa la variable que sigue una distribución normal.

2. **Media (μ):** Es el valor central de la distribución y define su posición.

3. **Desviación Estándar (σ):** Indica la dispersión de los datos alrededor de la media.

4. **Función de Densidad de Probabilidad (PDF):** Describe la probabilidad de que la variable aleatoria tome un valor específico.

**Fórmula Matemática:**

La función de densidad de probabilidad (PDF) para una distribución normal con media $( \mu )$ y desviación estándar $( \sigma )$ se define como:

$f(x) = \frac{1}{\sigma \sqrt{2\pi}} e^{ -\frac{1}{2}\left(\frac{x-\mu}{\sigma}\right)^2}$

Donde:

- $( \mu )$ es la media.
- $( \sigma )$ es la desviación estándar.
- $( e )$ es la base del logaritmo natural.

**Ejemplo con Gráfico en Python:**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Definir parámetros de la distribución normal
mu = 0
sigma = 1

# Generar datos de ejemplo de una distribución normal
data_normal = np.random.normal(mu, sigma, 10000)

# Visualizar el histograma
plt.hist(data_normal, bins=50, density=True, alpha=0.7, color='b', edgecolor='k')

# Crear una línea para la función de densidad de probabilidad (PDF)
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, sigma)
plt.plot(x, p, 'k', linewidth=2)

plt.title('Distribución Normal N(0, 1)')
plt.xlabel('Valor')
plt.ylabel('Densidad de Probabilidad')
plt.show()
```

En este ejemplo, se generan datos de una distribución normal estándar (N(0, 1)) y se visualiza el histograma. La línea negra representa la función de densidad de probabilidad (PDF) de la distribución normal estándar.

## **Binomial Distribution**

**Explicación:**

La distribución binomial modela el número de éxitos en una secuencia de ensayos independientes, donde cada ensayo tiene dos resultados posibles: éxito o fracaso. Es una distribución discreta y se utiliza comúnmente en situaciones en las que se repiten experimentos con el mismo conjunto de condiciones y se busca la probabilidad de obtener un número específico de éxitos.

**Terminología:**

1. **Número de Ensayos (n):** Representa la cantidad total de experimentos o ensayos.

2. **Probabilidad de Éxito (p):** Es la probabilidad de que ocurra un éxito en un solo ensayo.

3. **Probabilidad de Fracaso (q):** Es la probabilidad de que ocurra un fracaso en un solo ensayo, donde $( q = 1 - p )$.

4. **Número de Éxitos (k):** Representa la cantidad de éxitos que se quiere lograr en los $( n )$ ensayos.

**Fórmula Matemática:**

La función de probabilidad de la distribución binomial se define como:

$P(X = k) = \binom{n}{k} p^k q^{(n-k)}$

Donde:

- $P(X = k)$ es la probabilidad de obtener $( k )$ éxitos en $( n )$ ensayos.
- $\binom{n}{k}$ es el coeficiente binomial, que se calcula como $\frac{n!}{k!(n-k)!}$.
- $( p )$ es la probabilidad de éxito en un solo ensayo.
- $( q )$ es la probabilidad de fracaso en un solo ensayo $( q = 1 - p )$.
- $( k )$ es el número de éxitos.

**Ejemplo con Gráfico en Python:**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Definir parámetros de la distribución binomial
n = 10  # Número de ensayos
p = 0.5  # Probabilidad de éxito en un solo ensayo

# Generar datos de ejemplo de una distribución binomial
data_binomial = np.random.binomial(n, p, 1000)

# Visualizar el histograma
plt.hist(data_binomial, bins=np.arange(0, n+2)-0.5, density=True, alpha=0.7, color='b', edgecolor='k')

# Crear una línea para la función de probabilidad de masa (PMF)
x = np.arange(0, n+1)
p = binom.pmf(x, n, p)
plt.vlines(x, 0, p, colors='r', lw=2, alpha=0.7)

plt.title('Distribución Binomial B(10, 0.5)')
plt.xlabel('Número de Éxitos')
plt.ylabel('Probabilidad')
plt.show()
```

En este ejemplo, se generan datos de una distribución binomial con $( n = 10 )$ ensayos y $( p = 0.5 )$ probabilidad de éxito en un solo ensayo. Se visualiza el histograma de los datos generados y se superponen las líneas rojas que representan la función de probabilidad de masa (PMF) de la distribución binomial.

## **Poisson Distribution**

**Distribución de Poisson:**

**Explicación:**

La distribución de Poisson modela la probabilidad de que ocurra un número específico de eventos en un intervalo de tiempo o espacio fijo, dado que los eventos ocurren con una tasa promedio constante y de manera independiente. Es una distribución discreta y se utiliza en situaciones donde los eventos son raros y la ocurrencia de un evento no afecta la ocurrencia de otro.

**Terminología:**

1. **Tasa de Ocurrencia (λ):** Representa la tasa promedio de eventos que ocurren en el intervalo especificado.

2. **Número de Eventos (k):** Representa el número de eventos que se quiere evaluar la probabilidad de que ocurran en el intervalo.

**Fórmula Matemática:**

La función de probabilidad de la distribución de Poisson se define como:

$P(X = k) = \frac{e^{-\lambda} \lambda^k}{k!}$

Donde:

- $P(X = k)$ es la probabilidad de que ocurran $( k )$ eventos en el intervalo.
- $( e )$ es la base del logaritmo natural (aproximadamente 2.71828).
- $( \lambda )$ es la tasa de ocurrencia promedio en el intervalo.
- $( k )$ es el número de eventos.

**Ejemplo con Gráfico en Python:**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Definir parámetros de la distribución de Poisson
lmbda = 3  # Tasa de ocurrencia promedio

# Generar datos de ejemplo de una distribución de Poisson
data_poisson = np.random.poisson(lmbda, 1000)

# Visualizar el histograma
plt.hist(data_poisson, bins=np.arange(0, 15)-0.5, density=True, alpha=0.7, color='b', edgecolor='k')

# Crear una línea para la función de probabilidad de masa (PMF)
x = np.arange(0, 15)
p = poisson.pmf(x, lmbda)
plt.vlines(x, 0, p, colors='r', lw=2, alpha=0.7)

plt.title('Distribución de Poisson (λ=3)')
plt.xlabel('Número de Eventos')
plt.ylabel('Probabilidad')
plt.show()
```

En este ejemplo, se generan datos de una distribución de Poisson con una tasa de ocurrencia promedio $( \lambda = 3 )$. Se visualiza el histograma de los datos generados y se superponen las líneas rojas que representan la función de probabilidad de masa (PMF) de la distribución de Poisson.

## **Bernoulli Distribution**

**Distribución de Bernoulli:**

**Explicación:**

La distribución de Bernoulli modela un experimento aleatorio con dos posibles resultados: éxito o fracaso. Se utiliza para describir situaciones en las que un evento tiene dos resultados mutuamente excluyentes. La variable aleatoria en una distribución de Bernoulli toma el valor de 1 para el éxito y 0 para el fracaso.

**Terminología:**

1. **Probabilidad de Éxito (p):** Representa la probabilidad de que ocurra el evento deseado (éxito).

**Fórmula Matemática:**

La función de masa de probabilidad (PMF) de la distribución de Bernoulli se define como:

$P(X = x) = \begin{cases} p & \text{si } x = 1 \\ q = 1 - p & \text{si } x = 0 \end{cases}$

Donde:

- $P(X = x)$ es la probabilidad de que la variable aleatoria tome el valor $( x )$.
- $( p )$ es la probabilidad de éxito.
- $( q )$ es la probabilidad de fracaso $( q = 1 - p )$.

**Ejemplo con Gráfico en Python:**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli

# Definir parámetros de la distribución de Bernoulli
p = 0.3  # Probabilidad de éxito

# Generar datos de ejemplo de una distribución de Bernoulli
data_bernoulli = np.random.choice([0, 1], p=[1-p, p], size=1000)

# Visualizar el histograma
plt.hist(data_bernoulli, bins=[-0.5, 0.5, 1.5], density=True, alpha=0.7, color='b', edgecolor='k')

# Crear una línea para la función de masa de probabilidad (PMF)
x = [0, 1]
p_values = [1-p, p]
plt.vlines(x, 0, p_values, colors='r', lw=2, alpha=0.7)

plt.title('Distribución de Bernoulli (p=0.3)')
plt.xlabel('Variable Aleatoria')
plt.ylabel('Probabilidad')
plt.xticks([0, 1], ['Fracaso (0)', 'Éxito (1)'])
plt.show()
```

En este ejemplo, se generan datos de una distribución de Bernoulli con una probabilidad de éxito $p = 0.3$. Se visualiza el histograma de los datos generados y se superponen las líneas rojas que representan la función de masa de probabilidad (PMF) de la distribución de Bernoulli.

## **Log Normal Distribution**

**Distribución Log-Normal:**

**Explicación:**

La distribución log-normal es una distribución de probabilidad continua de una variable aleatoria cuyo logaritmo es normalmente distribuido. Es útil para modelar variables que son producto de procesos multiplicativos, como los rendimientos de inversiones financieras.

**Terminología:**

1. **Media del Logaritmo $(\mu)$:** La media de la distribución logarítmica.
2. **Desviación Estándar del Logaritmo $(\sigma)$:** La desviación estándar de la distribución logarítmica.

**Fórmula Matemática:**

La función de densidad de probabilidad (PDF) de la distribución log-normal es dada por:

$f(x | \mu, \sigma) = \frac{1}{x \sigma \sqrt{2\pi}} e^{ -\frac{(\ln(x) - \mu)^2}{2\sigma^2} }$

donde:

- $( x > 0 )$ es la variable aleatoria.
- $( \mu )$ es la media del logaritmo.
- $( \sigma )$ es la desviación estándar del logaritmo.

**Ejemplo con Gráfico en Python:**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import lognorm

# Definir parámetros de la distribución log-normal
mu = 0
sigma = 0.1

# Generar datos de ejemplo de una distribución log-normal
data_lognorm = np.random.lognormal(mean=mu, sigma=sigma, size=1000)

# Visualizar el histograma
plt.hist(data_lognorm, bins=50, density=True, alpha=0.7, color='b', edgecolor='k')

# Crear una línea para la función de densidad de probabilidad (PDF)
x = np.linspace(0, 3, 100)
pdf_values = lognorm.pdf(x, sigma, scale=np.exp(mu))
plt.plot(x, pdf_values, 'r-', lw=2)

plt.title('Distribución Log-Normal')
plt.xlabel('Variable Aleatoria (x)')
plt.ylabel('Densidad de Probabilidad')
plt.show()
```

En este ejemplo, se generan datos de una distribución log-normal con parámetros $(\mu = 0)$ y $(\sigma = 0.1)$. Se visualiza el histograma de los datos generados y se superpone la línea roja que representa la función de densidad de probabilidad (PDF) de la distribución log-normal. La variable aleatoria $(x)$ es siempre positiva, ya que es el resultado de exponentiar el logaritmo de una variable normalmente distribuida.

## **Gamma Distribution**

**Distribución Gamma:**

**Explicación:**

La distribución gamma es una familia de distribuciones de probabilidad continuas parametrizadas por un parámetro de forma $(k)$ y un parámetro de escala $(\theta)$. Es una generalización de la distribución exponencial y se utiliza comúnmente para modelar tiempos de espera, entre otros fenómenos.

**Terminología:**

1. **Parámetro de Forma $(k)$:** Controla la forma de la distribución.
2. **Parámetro de Escala $(\theta)$:** Relacionado con la tasa de eventos.

**Fórmula Matemática:**

La función de densidad de probabilidad (PDF) de la distribución gamma es dada por:

$f(x | k, \theta) = \frac{x^{k-1} e^{-\frac{x}{\theta}}}{\theta^k \Gamma(k)}$

donde:

- $( x \geq 0 )$ es la variable aleatoria.
- $( k > 0 )$ es el parámetro de forma.
- $( \theta > 0 )$ es el parámetro de escala.
- $( \Gamma(k) )$ es la función gamma.

La función gamma $( \Gamma(k) )$ es una función que generaliza el factorial a números reales y complejos.

**Ejemplo con Gráfico en Python:**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma

# Definir parámetros de la distribución gamma
k = 2  # Parámetro de forma
theta = 1  # Parámetro de escala

# Generar datos de ejemplo de una distribución gamma
data_gamma = np.random.gamma(k, theta, size=1000)

# Visualizar el histograma
plt.hist(data_gamma, bins=50, density=True, alpha=0.7, color='b', edgecolor='k')

# Crear una línea para la función de densidad de probabilidad (PDF)
x = np.linspace(0, 10, 100)
pdf_values = gamma.pdf(x, k, scale=theta)
plt.plot(x, pdf_values, 'r-', lw=2)

plt.title('Distribución Gamma')
plt.xlabel('Variable Aleatoria (x)')
plt.ylabel('Densidad de Probabilidad')
plt.show()
```

En este ejemplo, se generan datos de una distribución gamma con parámetros $(k = 2)$ y $(\theta = 1)$. Se visualiza el histograma de los datos generados y se superpone la línea roja que representa la función de densidad de probabilidad (PDF) de la distribución gamma. La variable aleatoria $(x)$ es siempre no negativa.

## **Geometric Distribution**

**Distribución Geométrica:**

**Explicación:**

La distribución geométrica modela el número de ensayos de Bernoulli independientes necesarios para que ocurra un evento de éxito por primera vez. En otras palabras, representa el número de fallos antes de obtener el primer éxito en una secuencia de ensayos de Bernoulli.

**Terminología:**

1. **Probabilidad de Éxito $(p)$:** Probabilidad de que un ensayo de Bernoulli sea exitoso.
2. **Número de Ensayos $(X)$:** Número de ensayos necesarios para obtener el primer éxito.

**Fórmula Matemática:**

La función de probabilidad de masa (PMF) de la distribución geométrica es dada por:

$P(X = k) = (1 - p)^{k-1} \cdot p$

donde:

- $( k )$ es el número de ensayos.
- $( p )$ es la probabilidad de éxito.

**Ejemplo con Gráfico en Python:**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import geom

# Definir la probabilidad de éxito
p = 0.3

# Generar datos de ejemplo de una distribución geométrica
data_geom = np.random.geometric(p, size=1000)

# Calcular la probabilidad teórica para cada valor posible
k_values = np.arange(1, 11)  # Número de ensayos
pmf_values = geom.pmf(k_values, p)

# Visualizar el histograma
plt.hist(data_geom, bins=range(1, 12), density=True, alpha=0.7, color='b', edgecolor='k')

# Crear una línea para la función de masa de probabilidad (PMF)
plt.vlines(k_values, 0, pmf_values, colors='r', lw=2, label='PMF Teórica')

plt.title('Distribución Geométrica')
plt.xlabel('Número de Ensayos (k)')
plt.ylabel('Probabilidad')
plt.legend()
plt.show()
```

En este ejemplo, se generan datos de una distribución geométrica con una probabilidad de éxito de $(p = 0.3)$. Se visualiza el histograma de los datos generados y se superpone la línea roja que representa la función de masa de probabilidad (PMF) teórica de la distribución geométrica. El eje $(x)$ muestra el número de ensayos necesarios para obtener el primer éxito $(k)$.

## **Beta Distribution**

**Distribución Beta:**

**Explicación:**

La distribución Beta es una distribución de probabilidad continua definida en el intervalo $[0, 1]$. Es comúnmente utilizada para modelar la distribución de variables aleatorias que representan proporciones o probabilidades.

**Terminología:**

1. **Parámetros $(\alpha)$ y $(\beta)$:** Son parámetros de forma que determinan la forma específica de la distribución. Ambos deben ser mayores que cero.

**Función de Densidad de Probabilidad (PDF):**

La función de densidad de probabilidad (PDF) de la distribución Beta es dada por:

$f(x;\alpha,\beta) = \frac{x^{\alpha-1} \cdot (1-x)^{\beta-1}}{B(\alpha, \beta)}$

donde:

- $( x )$ es la variable aleatoria en el intervalo $[0, 1]$.
- $( \alpha )$ y $( \beta )$ son los parámetros de forma.
- $B(\alpha, \beta)$ es la función beta, que es una constante de normalización.

**Ejemplo con Gráfico en Python:**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta

# Definir parámetros de forma
alpha = 2
beta_val = 5

# Generar datos de ejemplo de una distribución Beta
data_beta = np.random.beta(alpha, beta_val, size=1000)

# Calcular la PDF teórica para cada valor posible
x_values = np.linspace(0, 1, 100)
pdf_values = beta.pdf(x_values, alpha, beta_val)

# Visualizar el histograma
plt.hist(data_beta, bins=50, density=True, alpha=0.7, color='b', edgecolor='k')

# Crear una línea para la función de densidad de probabilidad (PDF)
plt.plot(x_values, pdf_values, 'r-', lw=2, label='PDF Teórica')

plt.title('Distribución Beta')
plt.xlabel('Variable Aleatoria (x)')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.show()
```

En este ejemplo, se generan datos de una distribución Beta con parámetros de forma $(\alpha = 2)$ y $(\beta = 5)$. Se visualiza el histograma de los datos generados y se superpone la línea roja que representa la función de densidad de probabilidad (PDF) teórica de la distribución Beta. La PDF describe cómo se distribuyen las probabilidades a lo largo del intervalo $[0, 1]$.
