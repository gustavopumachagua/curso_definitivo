## **Uniform Distribution**

**Explicaci贸n:**

La distribuci贸n uniforme es un tipo de distribuci贸n de probabilidad donde cada valor en el rango posible tiene la misma probabilidad de ocurrir. En otras palabras, todos los eventos tienen igual probabilidad de suceder. Es 煤til en situaciones en las que no hay raz贸n para favorecer un valor sobre otro dentro de un rango espec铆fico.

**Terminolog铆a:**

1. **Variable Aleatoria Continua (X):** Representa la variable que sigue una distribuci贸n uniforme.

2. **Par谩metros a y b:** Son los par谩metros que definen el rango de la distribuci贸n. La distribuci贸n uniforme se denota como U(a, b).

3. **Funci贸n de Densidad de Probabilidad (PDF):** Para una distribuci贸n uniforme, la PDF es constante en el intervalo [a, b] y es 0 fuera de ese intervalo.

**F贸rmula Matem谩tica:**

La funci贸n de densidad de probabilidad (PDF) para una distribuci贸n uniforme U(a, b) se define como:

$f(x) = \frac{1}{b - a} \text{ para } a \leq x \leq b$

$f(x) = 0 \text{ para } x < a \text{ o } x > b$

Donde:

- $( a )$ y $( b )$ son los par谩metros que definen el intervalo.
- $f(x)$ es la probabilidad de que la variable aleatoria $( X )$ tome el valor $( x )$.

**Ejemplo con Gr谩fico en Python:**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform

# Definir par谩metros a y b
a = 0
b = 10

# Generar datos de ejemplo de una distribuci贸n uniforme
data_uniform = uniform.rvs(size=10000, loc=a, scale=b-a)

# Visualizar el histograma
plt.hist(data_uniform, bins=50, density=True, alpha=0.7, color='g', edgecolor='k')
plt.plot([a, b], [1/(b-a), 1/(b-a)], 'r--', lw=2, label='PDF')
plt.title('Distribuci贸n Uniforme U(0, 10)')
plt.xlabel('Valor')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.show()
```

En este ejemplo, se generan datos de una distribuci贸n uniforme U(0, 10) y se visualiza el histograma. La l铆nea roja representa la funci贸n de densidad de probabilidad (PDF), que es constante en el intervalo [0, 10].

## **Normal Distribution**

**Distribuci贸n Normal (Gaussiana):**

**Explicaci贸n:**

La distribuci贸n normal, tambi茅n conocida como distribuci贸n gaussiana, es una de las distribuciones de probabilidad m谩s importantes en estad铆stica. Es una distribuci贸n continua que se caracteriza por ser sim茅trica alrededor de su media, con la mayor densidad de probabilidad en la media y una forma de campana.

**Terminolog铆a:**

1. **Variable Aleatoria Continua (X):** Representa la variable que sigue una distribuci贸n normal.

2. **Media (渭):** Es el valor central de la distribuci贸n y define su posici贸n.

3. **Desviaci贸n Est谩ndar ():** Indica la dispersi贸n de los datos alrededor de la media.

4. **Funci贸n de Densidad de Probabilidad (PDF):** Describe la probabilidad de que la variable aleatoria tome un valor espec铆fico.

**F贸rmula Matem谩tica:**

La funci贸n de densidad de probabilidad (PDF) para una distribuci贸n normal con media $( \mu )$ y desviaci贸n est谩ndar $( \sigma )$ se define como:

$f(x) = \frac{1}{\sigma \sqrt{2\pi}} e^{ -\frac{1}{2}\left(\frac{x-\mu}{\sigma}\right)^2}$

Donde:

- $( \mu )$ es la media.
- $( \sigma )$ es la desviaci贸n est谩ndar.
- $( e )$ es la base del logaritmo natural.

**Ejemplo con Gr谩fico en Python:**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Definir par谩metros de la distribuci贸n normal
mu = 0
sigma = 1

# Generar datos de ejemplo de una distribuci贸n normal
data_normal = np.random.normal(mu, sigma, 10000)

# Visualizar el histograma
plt.hist(data_normal, bins=50, density=True, alpha=0.7, color='b', edgecolor='k')

# Crear una l铆nea para la funci贸n de densidad de probabilidad (PDF)
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, sigma)
plt.plot(x, p, 'k', linewidth=2)

plt.title('Distribuci贸n Normal N(0, 1)')
plt.xlabel('Valor')
plt.ylabel('Densidad de Probabilidad')
plt.show()
```

En este ejemplo, se generan datos de una distribuci贸n normal est谩ndar (N(0, 1)) y se visualiza el histograma. La l铆nea negra representa la funci贸n de densidad de probabilidad (PDF) de la distribuci贸n normal est谩ndar.

## **Binomial Distribution**

**Explicaci贸n:**

La distribuci贸n binomial modela el n煤mero de 茅xitos en una secuencia de ensayos independientes, donde cada ensayo tiene dos resultados posibles: 茅xito o fracaso. Es una distribuci贸n discreta y se utiliza com煤nmente en situaciones en las que se repiten experimentos con el mismo conjunto de condiciones y se busca la probabilidad de obtener un n煤mero espec铆fico de 茅xitos.

**Terminolog铆a:**

1. **N煤mero de Ensayos (n):** Representa la cantidad total de experimentos o ensayos.

2. **Probabilidad de xito (p):** Es la probabilidad de que ocurra un 茅xito en un solo ensayo.

3. **Probabilidad de Fracaso (q):** Es la probabilidad de que ocurra un fracaso en un solo ensayo, donde $( q = 1 - p )$.

4. **N煤mero de xitos (k):** Representa la cantidad de 茅xitos que se quiere lograr en los $( n )$ ensayos.

**F贸rmula Matem谩tica:**

La funci贸n de probabilidad de la distribuci贸n binomial se define como:

$P(X = k) = \binom{n}{k} p^k q^{(n-k)}$

Donde:

- $P(X = k)$ es la probabilidad de obtener $( k )$ 茅xitos en $( n )$ ensayos.
- $\binom{n}{k}$ es el coeficiente binomial, que se calcula como $\frac{n!}{k!(n-k)!}$.
- $( p )$ es la probabilidad de 茅xito en un solo ensayo.
- $( q )$ es la probabilidad de fracaso en un solo ensayo $( q = 1 - p )$.
- $( k )$ es el n煤mero de 茅xitos.

**Ejemplo con Gr谩fico en Python:**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Definir par谩metros de la distribuci贸n binomial
n = 10  # N煤mero de ensayos
p = 0.5  # Probabilidad de 茅xito en un solo ensayo

# Generar datos de ejemplo de una distribuci贸n binomial
data_binomial = np.random.binomial(n, p, 1000)

# Visualizar el histograma
plt.hist(data_binomial, bins=np.arange(0, n+2)-0.5, density=True, alpha=0.7, color='b', edgecolor='k')

# Crear una l铆nea para la funci贸n de probabilidad de masa (PMF)
x = np.arange(0, n+1)
p = binom.pmf(x, n, p)
plt.vlines(x, 0, p, colors='r', lw=2, alpha=0.7)

plt.title('Distribuci贸n Binomial B(10, 0.5)')
plt.xlabel('N煤mero de xitos')
plt.ylabel('Probabilidad')
plt.show()
```

En este ejemplo, se generan datos de una distribuci贸n binomial con $( n = 10 )$ ensayos y $( p = 0.5 )$ probabilidad de 茅xito en un solo ensayo. Se visualiza el histograma de los datos generados y se superponen las l铆neas rojas que representan la funci贸n de probabilidad de masa (PMF) de la distribuci贸n binomial.

## **Poisson Distribution**

**Distribuci贸n de Poisson:**

**Explicaci贸n:**

La distribuci贸n de Poisson modela la probabilidad de que ocurra un n煤mero espec铆fico de eventos en un intervalo de tiempo o espacio fijo, dado que los eventos ocurren con una tasa promedio constante y de manera independiente. Es una distribuci贸n discreta y se utiliza en situaciones donde los eventos son raros y la ocurrencia de un evento no afecta la ocurrencia de otro.

**Terminolog铆a:**

1. **Tasa de Ocurrencia (位):** Representa la tasa promedio de eventos que ocurren en el intervalo especificado.

2. **N煤mero de Eventos (k):** Representa el n煤mero de eventos que se quiere evaluar la probabilidad de que ocurran en el intervalo.

**F贸rmula Matem谩tica:**

La funci贸n de probabilidad de la distribuci贸n de Poisson se define como:

$P(X = k) = \frac{e^{-\lambda} \lambda^k}{k!}$

Donde:

- $P(X = k)$ es la probabilidad de que ocurran $( k )$ eventos en el intervalo.
- $( e )$ es la base del logaritmo natural (aproximadamente 2.71828).
- $( \lambda )$ es la tasa de ocurrencia promedio en el intervalo.
- $( k )$ es el n煤mero de eventos.

**Ejemplo con Gr谩fico en Python:**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Definir par谩metros de la distribuci贸n de Poisson
lmbda = 3  # Tasa de ocurrencia promedio

# Generar datos de ejemplo de una distribuci贸n de Poisson
data_poisson = np.random.poisson(lmbda, 1000)

# Visualizar el histograma
plt.hist(data_poisson, bins=np.arange(0, 15)-0.5, density=True, alpha=0.7, color='b', edgecolor='k')

# Crear una l铆nea para la funci贸n de probabilidad de masa (PMF)
x = np.arange(0, 15)
p = poisson.pmf(x, lmbda)
plt.vlines(x, 0, p, colors='r', lw=2, alpha=0.7)

plt.title('Distribuci贸n de Poisson (位=3)')
plt.xlabel('N煤mero de Eventos')
plt.ylabel('Probabilidad')
plt.show()
```

En este ejemplo, se generan datos de una distribuci贸n de Poisson con una tasa de ocurrencia promedio $( \lambda = 3 )$. Se visualiza el histograma de los datos generados y se superponen las l铆neas rojas que representan la funci贸n de probabilidad de masa (PMF) de la distribuci贸n de Poisson.

## **Bernoulli Distribution**

**Distribuci贸n de Bernoulli:**

**Explicaci贸n:**

La distribuci贸n de Bernoulli modela un experimento aleatorio con dos posibles resultados: 茅xito o fracaso. Se utiliza para describir situaciones en las que un evento tiene dos resultados mutuamente excluyentes. La variable aleatoria en una distribuci贸n de Bernoulli toma el valor de 1 para el 茅xito y 0 para el fracaso.

**Terminolog铆a:**

1. **Probabilidad de xito (p):** Representa la probabilidad de que ocurra el evento deseado (茅xito).

**F贸rmula Matem谩tica:**

La funci贸n de masa de probabilidad (PMF) de la distribuci贸n de Bernoulli se define como:

$P(X = x) = \begin{cases} p & \text{si } x = 1 \\ q = 1 - p & \text{si } x = 0 \end{cases}$

Donde:

- $P(X = x)$ es la probabilidad de que la variable aleatoria tome el valor $( x )$.
- $( p )$ es la probabilidad de 茅xito.
- $( q )$ es la probabilidad de fracaso $( q = 1 - p )$.

**Ejemplo con Gr谩fico en Python:**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli

# Definir par谩metros de la distribuci贸n de Bernoulli
p = 0.3  # Probabilidad de 茅xito

# Generar datos de ejemplo de una distribuci贸n de Bernoulli
data_bernoulli = np.random.choice([0, 1], p=[1-p, p], size=1000)

# Visualizar el histograma
plt.hist(data_bernoulli, bins=[-0.5, 0.5, 1.5], density=True, alpha=0.7, color='b', edgecolor='k')

# Crear una l铆nea para la funci贸n de masa de probabilidad (PMF)
x = [0, 1]
p_values = [1-p, p]
plt.vlines(x, 0, p_values, colors='r', lw=2, alpha=0.7)

plt.title('Distribuci贸n de Bernoulli (p=0.3)')
plt.xlabel('Variable Aleatoria')
plt.ylabel('Probabilidad')
plt.xticks([0, 1], ['Fracaso (0)', 'xito (1)'])
plt.show()
```

En este ejemplo, se generan datos de una distribuci贸n de Bernoulli con una probabilidad de 茅xito $p = 0.3$. Se visualiza el histograma de los datos generados y se superponen las l铆neas rojas que representan la funci贸n de masa de probabilidad (PMF) de la distribuci贸n de Bernoulli.

## **Log Normal Distribution**

**Distribuci贸n Log-Normal:**

**Explicaci贸n:**

La distribuci贸n log-normal es una distribuci贸n de probabilidad continua de una variable aleatoria cuyo logaritmo es normalmente distribuido. Es 煤til para modelar variables que son producto de procesos multiplicativos, como los rendimientos de inversiones financieras.

**Terminolog铆a:**

1. **Media del Logaritmo $(\mu)$:** La media de la distribuci贸n logar铆tmica.
2. **Desviaci贸n Est谩ndar del Logaritmo $(\sigma)$:** La desviaci贸n est谩ndar de la distribuci贸n logar铆tmica.

**F贸rmula Matem谩tica:**

La funci贸n de densidad de probabilidad (PDF) de la distribuci贸n log-normal es dada por:

$f(x | \mu, \sigma) = \frac{1}{x \sigma \sqrt{2\pi}} e^{ -\frac{(\ln(x) - \mu)^2}{2\sigma^2} }$

donde:

- $( x > 0 )$ es la variable aleatoria.
- $( \mu )$ es la media del logaritmo.
- $( \sigma )$ es la desviaci贸n est谩ndar del logaritmo.

**Ejemplo con Gr谩fico en Python:**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import lognorm

# Definir par谩metros de la distribuci贸n log-normal
mu = 0
sigma = 0.1

# Generar datos de ejemplo de una distribuci贸n log-normal
data_lognorm = np.random.lognormal(mean=mu, sigma=sigma, size=1000)

# Visualizar el histograma
plt.hist(data_lognorm, bins=50, density=True, alpha=0.7, color='b', edgecolor='k')

# Crear una l铆nea para la funci贸n de densidad de probabilidad (PDF)
x = np.linspace(0, 3, 100)
pdf_values = lognorm.pdf(x, sigma, scale=np.exp(mu))
plt.plot(x, pdf_values, 'r-', lw=2)

plt.title('Distribuci贸n Log-Normal')
plt.xlabel('Variable Aleatoria (x)')
plt.ylabel('Densidad de Probabilidad')
plt.show()
```

En este ejemplo, se generan datos de una distribuci贸n log-normal con par谩metros $(\mu = 0)$ y $(\sigma = 0.1)$. Se visualiza el histograma de los datos generados y se superpone la l铆nea roja que representa la funci贸n de densidad de probabilidad (PDF) de la distribuci贸n log-normal. La variable aleatoria $(x)$ es siempre positiva, ya que es el resultado de exponentiar el logaritmo de una variable normalmente distribuida.

## **Gamma Distribution**

**Distribuci贸n Gamma:**

**Explicaci贸n:**

La distribuci贸n gamma es una familia de distribuciones de probabilidad continuas parametrizadas por un par谩metro de forma $(k)$ y un par谩metro de escala $(\theta)$. Es una generalizaci贸n de la distribuci贸n exponencial y se utiliza com煤nmente para modelar tiempos de espera, entre otros fen贸menos.

**Terminolog铆a:**

1. **Par谩metro de Forma $(k)$:** Controla la forma de la distribuci贸n.
2. **Par谩metro de Escala $(\theta)$:** Relacionado con la tasa de eventos.

**F贸rmula Matem谩tica:**

La funci贸n de densidad de probabilidad (PDF) de la distribuci贸n gamma es dada por:

$f(x | k, \theta) = \frac{x^{k-1} e^{-\frac{x}{\theta}}}{\theta^k \Gamma(k)}$

donde:

- $( x \geq 0 )$ es la variable aleatoria.
- $( k > 0 )$ es el par谩metro de forma.
- $( \theta > 0 )$ es el par谩metro de escala.
- $( \Gamma(k) )$ es la funci贸n gamma.

La funci贸n gamma $( \Gamma(k) )$ es una funci贸n que generaliza el factorial a n煤meros reales y complejos.

**Ejemplo con Gr谩fico en Python:**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma

# Definir par谩metros de la distribuci贸n gamma
k = 2  # Par谩metro de forma
theta = 1  # Par谩metro de escala

# Generar datos de ejemplo de una distribuci贸n gamma
data_gamma = np.random.gamma(k, theta, size=1000)

# Visualizar el histograma
plt.hist(data_gamma, bins=50, density=True, alpha=0.7, color='b', edgecolor='k')

# Crear una l铆nea para la funci贸n de densidad de probabilidad (PDF)
x = np.linspace(0, 10, 100)
pdf_values = gamma.pdf(x, k, scale=theta)
plt.plot(x, pdf_values, 'r-', lw=2)

plt.title('Distribuci贸n Gamma')
plt.xlabel('Variable Aleatoria (x)')
plt.ylabel('Densidad de Probabilidad')
plt.show()
```

En este ejemplo, se generan datos de una distribuci贸n gamma con par谩metros $(k = 2)$ y $(\theta = 1)$. Se visualiza el histograma de los datos generados y se superpone la l铆nea roja que representa la funci贸n de densidad de probabilidad (PDF) de la distribuci贸n gamma. La variable aleatoria $(x)$ es siempre no negativa.

## **Geometric Distribution**

**Distribuci贸n Geom茅trica:**

**Explicaci贸n:**

La distribuci贸n geom茅trica modela el n煤mero de ensayos de Bernoulli independientes necesarios para que ocurra un evento de 茅xito por primera vez. En otras palabras, representa el n煤mero de fallos antes de obtener el primer 茅xito en una secuencia de ensayos de Bernoulli.

**Terminolog铆a:**

1. **Probabilidad de xito $(p)$:** Probabilidad de que un ensayo de Bernoulli sea exitoso.
2. **N煤mero de Ensayos $(X)$:** N煤mero de ensayos necesarios para obtener el primer 茅xito.

**F贸rmula Matem谩tica:**

La funci贸n de probabilidad de masa (PMF) de la distribuci贸n geom茅trica es dada por:

$P(X = k) = (1 - p)^{k-1} \cdot p$

donde:

- $( k )$ es el n煤mero de ensayos.
- $( p )$ es la probabilidad de 茅xito.

**Ejemplo con Gr谩fico en Python:**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import geom

# Definir la probabilidad de 茅xito
p = 0.3

# Generar datos de ejemplo de una distribuci贸n geom茅trica
data_geom = np.random.geometric(p, size=1000)

# Calcular la probabilidad te贸rica para cada valor posible
k_values = np.arange(1, 11)  # N煤mero de ensayos
pmf_values = geom.pmf(k_values, p)

# Visualizar el histograma
plt.hist(data_geom, bins=range(1, 12), density=True, alpha=0.7, color='b', edgecolor='k')

# Crear una l铆nea para la funci贸n de masa de probabilidad (PMF)
plt.vlines(k_values, 0, pmf_values, colors='r', lw=2, label='PMF Te贸rica')

plt.title('Distribuci贸n Geom茅trica')
plt.xlabel('N煤mero de Ensayos (k)')
plt.ylabel('Probabilidad')
plt.legend()
plt.show()
```

En este ejemplo, se generan datos de una distribuci贸n geom茅trica con una probabilidad de 茅xito de $(p = 0.3)$. Se visualiza el histograma de los datos generados y se superpone la l铆nea roja que representa la funci贸n de masa de probabilidad (PMF) te贸rica de la distribuci贸n geom茅trica. ## **Beta Distribution**

**Distribuci贸n Beta:**

**Explicaci贸n:**

La distribuci贸n Beta es una distribuci贸n de probabilidad continua definida en el intervalo $[0, 1]$. Es com煤nmente utilizada para modelar la distribuci贸n de variables aleatorias que representan proporciones o probabilidades.

**Terminolog铆a:**

1. **Par谩metros $(\alpha)$ y $(\beta)$:** Son par谩metros de forma que determinan la forma espec铆fica de la distribuci贸n. Ambos deben ser mayores que cero.

**Funci贸n de Densidad de Probabilidad (PDF):**

La funci贸n de densidad de probabilidad (PDF) de la distribuci贸n Beta es dada por:

$f(x;\alpha,\beta) = \frac{x^{\alpha-1} \cdot (1-x)^{\beta-1}}{B(\alpha, \beta)}$

donde:

- $( x )$ es la variable aleatoria en el intervalo $[0, 1]$.
- $( \alpha )$ y $( \beta )$ son los par谩metros de forma.
- $B(\alpha, \beta)$ es la funci贸n beta, que es una constante de normalizaci贸n.

**Ejemplo con Gr谩fico en Python:**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta

# Definir par谩metros de forma
alpha = 2
beta_val = 5

# Generar datos de ejemplo de una distribuci贸n Beta
data_beta = np.random.beta(alpha, beta_val, size=1000)

# Calcular la PDF te贸rica para cada valor posible
x_values = np.linspace(0, 1, 100)
pdf_values = beta.pdf(x_values, alpha, beta_val)

# Visualizar el histograma
plt.hist(data_beta, bins=50, density=True, alpha=0.7, color='b', edgecolor='k')

# Crear una l铆nea para la funci贸n de densidad de probabilidad (PDF)
plt.plot(x_values, pdf_values, 'r-', lw=2, label='PDF Te贸rica')

plt.title('Distribuci贸n Beta')
plt.xlabel('Variable Aleatoria (x)')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.show()
```

En este ejemplo, se generan datos de una distribuci贸n Beta con par谩metros de forma $(\alpha = 2)$ y $(\beta = 5)$. Se visualiza el histograma de los datos generados y se superpone la l铆nea roja que representa la funci贸n de densidad de probabilidad (PDF) te贸rica de la distribuci贸n Beta. La PDF describe c贸mo se distribuyen las probabilidades a lo largo del intervalo $[0, 1]$.
