# **1.  Matriz de Confusi贸n**

![Matriz de Confusi贸n](../img/matriz_confusion.avif "Matriz de Confusi贸n")

La matriz de confusi贸n es una herramienta que se utiliza en el campo de la clasificaci贸n de machine learning para evaluar el rendimiento de un modelo en la tarea de clasificaci贸n. Proporciona una visi贸n detallada de c贸mo un modelo clasifica las instancias de diferentes clases. La matriz de confusi贸n es especialmente 煤til cuando se trabaja con problemas de clasificaci贸n multiclase.

**Terminolog铆a de la matriz de confusi贸n:**

1. **Verdaderos Positivos (VP):** La cantidad de instancias que realmente pertenecen a una clase espec铆fica y que el modelo clasific贸 correctamente como pertenecientes a esa clase.

2. **Falsos Positivos (FP):** La cantidad de instancias que en realidad no pertenecen a una clase espec铆fica, pero el modelo las clasific贸 incorrectamente como pertenecientes a esa clase.

3. **Verdaderos Negativos (VN):** La cantidad de instancias que realmente no pertenecen a una clase espec铆fica y que el modelo clasific贸 correctamente como no pertenecientes a esa clase.

4. **Falsos Negativos (FN):** La cantidad de instancias que realmente pertenecen a una clase espec铆fica, pero el modelo las clasific贸 incorrectamente como no pertenecientes a esa clase.

La matriz de confusi贸n tiene la siguiente forma:

$\begin{matrix}
 & \text{Clase Positiva} & \text{Clase Negativa} \\
\text{Predicci贸n Positiva} & VP & FP \\
\text{Predicci贸n Negativa} & FN & VN \\
\end{matrix}$

La precisi贸n (accuracy) se calcula como la suma de los verdaderos positivos y verdaderos negativos dividida por el total de instancias:

$\text{Precisi贸n} = \frac{VP + VN}{VP + FP + FN + VN}$

A continuaci贸n, te mostrar茅 c贸mo crear y visualizar una matriz de confusi贸n en Python utilizando la biblioteca scikit-learn y matplotlib:

![Matrix Confusion](../img/matrix_confusion.png "Matrix Confusion")

```python
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Ejemplo de clasificaci贸n binaria
y_true = [1, 0, 1, 1, 0, 1, 0, 0, 1, 0]
y_pred = [1, 0, 1, 1, 0, 0, 1, 0, 1, 1]

# Crear matriz de confusi贸n
cm = confusion_matrix(y_true, y_pred)

# Personalizar la visualizaci贸n de la matriz de confusi贸n
plt.figure(figsize=(6, 4))
sns.heatmap(
    cm,
    annot=True,            # Muestra los valores en las celdas
    fmt="d",               # Formato de los valores (enteros)
    cmap="Blues",          # Colormap (escala de colores)
    xticklabels=['Negativo', 'Positivo'],
    yticklabels=['Negativo', 'Positivo'],
    linewidths=.5,          # Ancho de las l铆neas divisorias
    square=True,            # Mostrar celdas cuadradas
    cbar_kws={"shrink": 0.5} # Tama帽o de la barra de color
)

plt.xlabel('Predicciones')
plt.ylabel('Valores reales')
plt.title('Matriz de Confusi贸n')

# Agregar valores en cada celda
for i in range(cm.shape[0]):
    for j in range(cm.shape[1]):
        plt.text(j + 0.5, i + 0.5, str(cm[i, j]), ha='center', va='center', color='black')

plt.show()
```

En este ejemplo, `y_true` son las etiquetas reales y `y_pred` son las etiquetas predichas por tu modelo. La matriz de confusi贸n se visualiza como un mapa de calor con la ayuda de la biblioteca Seaborn. Este c贸digo es para un problema de clasificaci贸n binaria, pero puedes adaptarlo f谩cilmente para problemas de clasificaci贸n multiclase.

### **Ejemplos**

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Ejemplo de etiquetas reales y predicciones
y_true = np.array([1, 0, 1, 2, 1, 0, 1, 2, 2, 0])
y_pred = np.array([1, 0, 2, 2, 1, 0, 1, 2, 1, 0])

# Crear la matriz de confusi贸n
cm = confusion_matrix(y_true, y_pred)

# Calcular la exactitud del modelo
accuracy = accuracy_score(y_true, y_pred)

# Imprimir informe de clasificaci贸n
report = classification_report(y_true, y_pred)
print("Informe de clasificaci贸n:\n", report)

# Personalizar la visualizaci贸n de la matriz de confusi贸n
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="YlGnBu", linewidths=.5, cbar_kws={'label': 'Count'},
            xticklabels=np.unique(y_true), yticklabels=np.unique(y_true))
plt.xlabel('Clase Predicha')
plt.ylabel('Clase Real')
plt.title(f'Matriz de Confusi贸n\nExactitud: {accuracy:.2f}')
plt.show()
```

```python
import pandas as pd
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Supongamos que tienes un DataFrame llamado df con columnas 'y_true' y 'y_pred'
# Ejemplo de DataFrame
data = {'y_true': [1, 0, 1, 2, 1, 0, 1, 2, 2, 0, 2, 2, 0],
        'y_pred': [1, 0, 2, 2, 1, 0, 1, 2, 1, 0, 2, 2, 0]}
df = pd.DataFrame(data)

# Extraer las etiquetas reales y predicciones del DataFrame
y_true = df['y_true']
y_pred = df['y_pred']

# Crear la matriz de confusi贸n
cm = confusion_matrix(y_true, y_pred)

# Visualizar la matriz de confusi贸n como un mapa de calor
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=df['y_true'].unique(), yticklabels=df['y_true'].unique())
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Matriz de Confusi贸n (Multiclase)')
plt.show()
```
