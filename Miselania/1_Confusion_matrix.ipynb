# **1.  Matriz de Confusión**

![Matriz de Confusión](../img/matriz_confusion.avif "Matriz de Confusión")

La matriz de confusión es una herramienta que se utiliza en el campo de la clasificación de machine learning para evaluar el rendimiento de un modelo en la tarea de clasificación. Proporciona una visión detallada de cómo un modelo clasifica las instancias de diferentes clases. La matriz de confusión es especialmente útil cuando se trabaja con problemas de clasificación multiclase.

**Terminología de la matriz de confusión:**

1. **Verdaderos Positivos (VP):** La cantidad de instancias que realmente pertenecen a una clase específica y que el modelo clasificó correctamente como pertenecientes a esa clase.

2. **Falsos Positivos (FP):** La cantidad de instancias que en realidad no pertenecen a una clase específica, pero el modelo las clasificó incorrectamente como pertenecientes a esa clase.

3. **Verdaderos Negativos (VN):** La cantidad de instancias que realmente no pertenecen a una clase específica y que el modelo clasificó correctamente como no pertenecientes a esa clase.

4. **Falsos Negativos (FN):** La cantidad de instancias que realmente pertenecen a una clase específica, pero el modelo las clasificó incorrectamente como no pertenecientes a esa clase.

La matriz de confusión tiene la siguiente forma:

$\begin{matrix}
 & \text{Clase Positiva} & \text{Clase Negativa} \\
\text{Predicción Positiva} & VP & FP \\
\text{Predicción Negativa} & FN & VN \\
\end{matrix}$

La precisión (accuracy) se calcula como la suma de los verdaderos positivos y verdaderos negativos dividida por el total de instancias:

$\text{Precisión} = \frac{VP + VN}{VP + FP + FN + VN}$

A continuación, te mostraré cómo crear y visualizar una matriz de confusión en Python utilizando la biblioteca scikit-learn y matplotlib:

![Matrix Confusion](../img/matrix_confusion.png "Matrix Confusion")

```python
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Ejemplo de clasificación binaria
y_true = [1, 0, 1, 1, 0, 1, 0, 0, 1, 0]
y_pred = [1, 0, 1, 1, 0, 0, 1, 0, 1, 1]

# Crear matriz de confusión
cm = confusion_matrix(y_true, y_pred)

# Personalizar la visualización de la matriz de confusión
plt.figure(figsize=(6, 4))
sns.heatmap(
    cm,
    annot=True,            # Muestra los valores en las celdas
    fmt="d",               # Formato de los valores (enteros)
    cmap="Blues",          # Colormap (escala de colores)
    xticklabels=['Negativo', 'Positivo'],
    yticklabels=['Negativo', 'Positivo'],
    linewidths=.5,          # Ancho de las líneas divisorias
    square=True,            # Mostrar celdas cuadradas
    cbar_kws={"shrink": 0.5} # Tamaño de la barra de color
)

plt.xlabel('Predicciones')
plt.ylabel('Valores reales')
plt.title('Matriz de Confusión')

# Agregar valores en cada celda
for i in range(cm.shape[0]):
    for j in range(cm.shape[1]):
        plt.text(j + 0.5, i + 0.5, str(cm[i, j]), ha='center', va='center', color='black')

plt.show()
```

En este ejemplo, `y_true` son las etiquetas reales y `y_pred` son las etiquetas predichas por tu modelo. La matriz de confusión se visualiza como un mapa de calor con la ayuda de la biblioteca Seaborn. Este código es para un problema de clasificación binaria, pero puedes adaptarlo fácilmente para problemas de clasificación multiclase.

### **Ejemplos**

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Ejemplo de etiquetas reales y predicciones
y_true = np.array([1, 0, 1, 2, 1, 0, 1, 2, 2, 0])
y_pred = np.array([1, 0, 2, 2, 1, 0, 1, 2, 1, 0])

# Crear la matriz de confusión
cm = confusion_matrix(y_true, y_pred)

# Calcular la exactitud del modelo
accuracy = accuracy_score(y_true, y_pred)

# Imprimir informe de clasificación
report = classification_report(y_true, y_pred)
print("Informe de clasificación:\n", report)

# Personalizar la visualización de la matriz de confusión
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="YlGnBu", linewidths=.5, cbar_kws={'label': 'Count'},
            xticklabels=np.unique(y_true), yticklabels=np.unique(y_true))
plt.xlabel('Clase Predicha')
plt.ylabel('Clase Real')
plt.title(f'Matriz de Confusión\nExactitud: {accuracy:.2f}')
plt.show()
```

```python
import pandas as pd
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Supongamos que tienes un DataFrame llamado df con columnas 'y_true' y 'y_pred'
# Ejemplo de DataFrame
data = {'y_true': [1, 0, 1, 2, 1, 0, 1, 2, 2, 0, 2, 2, 0],
        'y_pred': [1, 0, 2, 2, 1, 0, 1, 2, 1, 0, 2, 2, 0]}
df = pd.DataFrame(data)

# Extraer las etiquetas reales y predicciones del DataFrame
y_true = df['y_true']
y_pred = df['y_pred']

# Crear la matriz de confusión
cm = confusion_matrix(y_true, y_pred)

# Visualizar la matriz de confusión como un mapa de calor
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=df['y_true'].unique(), yticklabels=df['y_true'].unique())
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Matriz de Confusión (Multiclase)')
plt.show()
```
