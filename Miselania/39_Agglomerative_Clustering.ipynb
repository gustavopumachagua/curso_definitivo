# **39. Agglomerative Clustering**

**Agglomerative Clustering: Explicaci√≥n Detallada**

**Introducci√≥n:**

El Agglomerative Clustering es un algoritmo de agrupamiento jer√°rquico que construye una jerarqu√≠a de grupos. Este enfoque comienza considerando cada punto de datos como un cl√∫ster individual y luego fusiona gradualmente los cl√∫steres m√°s cercanos para formar grupos m√°s grandes. Este proceso se repite hasta que todos los puntos de datos est√°n en un solo cl√∫ster o hasta que se cumple un criterio predefinido.

**Terminolog√≠a:**

1. **Matriz de Enlace (Linkage Matrix):** Matriz que registra las fusiones de cl√∫steres y las distancias entre ellos en cada paso.

2. **Enlace (Linkage):** Criterio utilizado para medir la distancia entre dos cl√∫steres y decidir cu√°les fusionar. Algunos m√©todos comunes son el enlace √∫nico, enlace completo y enlace promedio.

3. **Dendrograma:** Representaci√≥n gr√°fica de la jerarqu√≠a de cl√∫steres que muestra c√≥mo los cl√∫steres se fusionan a lo largo del tiempo.

**F√≥rmula Matem√°tica:**

La f√≥rmula matem√°tica espec√≠fica puede variar seg√∫n el m√©todo de enlace seleccionado. La distancia entre dos cl√∫steres se mide en funci√≥n de la proximidad entre sus puntos de datos. Por ejemplo, en el enlace √∫nico, la distancia entre dos cl√∫steres se define como la distancia m√°s corta entre cualquier par de puntos de los dos cl√∫steres.

**Ejemplo en Python:**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage

# Generar datos de ejemplo
np.random.seed(42)
X = np.random.rand(10, 2)

# Calcular la matriz de enlace utilizando enlace completo
Z = linkage(X, method='complete')

# Crear el dendrograma
dendrogram(Z)

# Mostrar el dendrograma
plt.title('Dendrograma')
plt.xlabel('√çndice de Datos')
plt.ylabel('Distancia')
plt.show()
```

Este ejemplo genera datos aleatorios bidimensionales y utiliza el m√©todo de enlace completo para construir un dendrograma que muestra la jerarqu√≠a de agrupamiento. Puedes ajustar el m√©todo de enlace y otros par√°metros seg√∫n tus necesidades espec√≠ficas.
