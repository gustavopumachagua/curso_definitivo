# **21. Uniform Distribution**

La Distribución Uniforme es una distribución de probabilidad continua en la que todos los valores posibles de una variable aleatoria tienen la misma probabilidad de ocurrir. En otras palabras, cada valor en el rango posible es igualmente probable. Esta distribución es fundamentalmente justa y se utiliza comúnmente en estadísticas y probabilidad.

**Terminología:**

- **$(a)$:** El valor mínimo en el rango de la distribución.
- **$(b)$:** El valor máximo en el rango de la distribución.
- **$(X)$:** Variable aleatoria que sigue una distribución uniforme en el intervalo $(a, b)$.
- **$f(x)$:** Función de densidad de probabilidad (PDF) que describe la probabilidad de que $(X)$ tome un valor específico.

**Fórmula Matemática:**

La función de densidad de probabilidad (PDF) de la distribución uniforme es constante en el intervalo $(a, b)$ y se define como:

$f(x) = \frac{1}{b - a}$

donde $(a)$ y $(b)$ son los extremos del intervalo.

**Ejemplo con Gráfico en Python:**

Vamos a generar un gráfico de la distribución uniforme en el intervalo $(a, b)$ usando Python:

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform

# Definir los parámetros a y b
a = 1
b = 5

# Crear una variable aleatoria uniforme
rv = uniform(loc=a, scale=b-a)

# Generar valores en el rango (a, b)
x = np.linspace(a, b, 1000)

# Calcular la función de densidad de probabilidad (PDF)
pdf = rv.pdf(x)

# Gráfico de la distribución uniforme
plt.plot(x, pdf, label='Distribución Uniforme')
plt.title('Distribución Uniforme en el Intervalo (1, 5)')
plt.xlabel('Valor')
plt.ylabel('Probabilidad')
plt.legend()
plt.show()
```

Este código utiliza la biblioteca SciPy para trabajar con la distribución uniforme y matplotlib para visualizarla. Puedes ajustar los valores de $(a)$ y $(b)$ según tus necesidades. La distribución es constante en el intervalo $(a, b)$, como se representa en el gráfico.
