# **21. Uniform Distribution**

La Distribuci贸n Uniforme es una distribuci贸n de probabilidad continua en la que todos los valores posibles de una variable aleatoria tienen la misma probabilidad de ocurrir. En otras palabras, cada valor en el rango posible es igualmente probable. Esta distribuci贸n es fundamentalmente justa y se utiliza com煤nmente en estad铆sticas y probabilidad.

**Terminolog铆a:**

- **$(a)$:** El valor m铆nimo en el rango de la distribuci贸n.
- **$(b)$:** El valor m谩ximo en el rango de la distribuci贸n.
- **$(X)$:** Variable aleatoria que sigue una distribuci贸n uniforme en el intervalo $(a, b)$.
- **$f(x)$:** Funci贸n de densidad de probabilidad (PDF) que describe la probabilidad de que $(X)$ tome un valor espec铆fico.

**F贸rmula Matem谩tica:**

La funci贸n de densidad de probabilidad (PDF) de la distribuci贸n uniforme es constante en el intervalo $(a, b)$ y se define como:

$f(x) = \frac{1}{b - a}$

donde $(a)$ y $(b)$ son los extremos del intervalo.

**Ejemplo con Gr谩fico en Python:**

Vamos a generar un gr谩fico de la distribuci贸n uniforme en el intervalo $(a, b)$ usando Python:

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform

# Definir los par谩metros a y b
a = 1
b = 5

# Crear una variable aleatoria uniforme
rv = uniform(loc=a, scale=b-a)

# Generar valores en el rango (a, b)
x = np.linspace(a, b, 1000)

# Calcular la funci贸n de densidad de probabilidad (PDF)
pdf = rv.pdf(x)

# Gr谩fico de la distribuci贸n uniforme
plt.plot(x, pdf, label='Distribuci贸n Uniforme')
plt.title('Distribuci贸n Uniforme en el Intervalo (1, 5)')
plt.xlabel('Valor')
plt.ylabel('Probabilidad')
plt.legend()
plt.show()
```
