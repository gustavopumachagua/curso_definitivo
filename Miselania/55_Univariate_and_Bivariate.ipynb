# **55. Univariate and Bivariate**

**An√°lisis Univariado y Bivariado: Explicaci√≥n Detallada**

**An√°lisis Univariado:**

- **Definici√≥n:** El an√°lisis univariado se centra en el estudio de una sola variable a la vez. Examina las propiedades, distribuciones y caracter√≠sticas de una √∫nica variable sin considerar la relaci√≥n con otras variables.

- **Terminolog√≠a:**

  - **Variable:** Una caracter√≠stica o propiedad que puede variar. En el an√°lisis univariado, se estudia una variable a la vez.
  - **Estad√≠sticas Descriptivas:** M√©tricas que resumen y describen las propiedades de una variable, como la media, la mediana, la desviaci√≥n est√°ndar, etc.
  - **Histograma:** Gr√°fico que muestra la distribuci√≥n de una variable a trav√©s de barras.

**Ejemplo de An√°lisis Univariado en Python:**

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos de ejemplo
data = np.random.randn(1000)

# Personalizar el histograma
plt.figure(figsize=(8, 6))

# Histograma con bordes m√°s oscuros
plt.hist(data, bins=30, color='blue', edgecolor='black', alpha=0.7)

# A√±adir una l√≠nea vertical para la media
mean_value = np.mean(data)
plt.axvline(mean_value, color='red', linestyle='dashed', linewidth=2, label=f'Media = {mean_value:.2f}')

# A√±adir una l√≠nea vertical para la mediana
median_value = np.median(data)
plt.axvline(median_value, color='green', linestyle='dashed', linewidth=2, label=f'Mediana = {median_value:.2f}')

# A√±adir una leyenda
plt.legend()

# A√±adir etiquetas y t√≠tulo
plt.title('Histograma: An√°lisis Univariado')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')

# Mostrar la cuadr√≠cula
plt.grid(True, linestyle='--', alpha=0.5)

# Ajustar el dise√±o
plt.tight_layout()

# Mostrar la gr√°fica
plt.show()
```

En este ejemplo, se genera un conjunto de datos aleatorios y se visualiza su distribuci√≥n mediante un histograma. Esto es un an√°lisis univariado, ya que estamos examinando las propiedades de una sola variable.

**An√°lisis Bivariado:**

- **Definici√≥n:** El an√°lisis bivariado implica el estudio de dos variables simult√°neamente para examinar la relaci√≥n entre ellas. Busca comprender c√≥mo cambia una variable cuando la otra var√≠a.

- **Terminolog√≠a:**

  - **Variables Bivariadas:** Dos variables que se estudian en conjunto.
  - **Diagrama de Dispersi√≥n:** Gr√°fico que representa puntos en un plano cartesiano, donde cada punto corresponde a los valores de ambas variables.
  - **Correlaci√≥n:** Medida estad√≠stica que cuantifica la relaci√≥n lineal entre dos variables.

**Ejemplo de An√°lisis Bivariado en Python:**

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos de ejemplo
x = np.random.randn(100)
y = 2 * x + np.random.randn(100)

# Personalizar el diagrama de dispersi√≥n
plt.figure(figsize=(8, 6))

# Diagrama de dispersi√≥n con puntos m√°s grandes y color m√°s claro
plt.scatter(x, y, color='skyblue', alpha=0.7, s=50)

# A√±adir una l√≠nea de regresi√≥n
plt.plot(x, 2 * x, color='red', linestyle='dashed', linewidth=2, label='Regresi√≥n Lineal')

# A√±adir etiquetas para los ejes y t√≠tulo
plt.title('Diagrama de Dispersi√≥n: An√°lisis Bivariado')
plt.xlabel('Variable X')
plt.ylabel('Variable Y')

# A√±adir una leyenda
plt.legend()

# Mostrar la cuadr√≠cula
plt.grid(True, linestyle='--', alpha=0.5)

# Ajustar el dise√±o
plt.tight_layout()

# Mostrar la gr√°fica
plt.show()
```

En este ejemplo, se generan dos conjuntos de datos y se visualizan en un diagrama de dispersi√≥n. La relaci√≥n lineal entre las dos variables se evidencia en el an√°lisis bivariado.

**Resumen:**

- El an√°lisis univariado se centra en una variable a la vez, mientras que el an√°lisis bivariado examina la relaci√≥n entre dos variables.
- En el an√°lisis univariado, se utilizan estad√≠sticas descriptivas y gr√°ficos como histogramas.
- En el an√°lisis bivariado, se emplean diagramas de dispersi√≥n y medidas de correlaci√≥n para entender la relaci√≥n entre dos variables.
