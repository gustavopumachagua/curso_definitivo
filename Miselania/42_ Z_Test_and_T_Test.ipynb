# **42. Z Test and T Test**

**Z-Test y T-Test: Explicación Detallada**

**Introducción:**

Z-Test y T-Test son métodos estadísticos utilizados para realizar pruebas de hipótesis sobre parámetros poblacionales, especialmente para comparar medias de dos grupos.

**Z-Test:**

El Z-Test se utiliza cuando la desviación estándar de la población es conocida. La fórmula general del Z-Test para comparar dos medias $(\mu_1)$ y $(\mu_2)$ es:

$Z = \frac{(\bar{x}_1 - \bar{x}_2)}{\sqrt{\frac{\sigma_1^2}{n_1} + \frac{\sigma_2^2}{n_2}}}$

Donde:

- $(\bar{x}_1)$ y $(\bar{x}_2)$ son las medias de las muestras.
- $(\sigma_1)$ y $(\sigma_2)$ son las desviaciones estándar de las poblaciones.
- $(n_1)$ y $(n_2)$ son los tamaños de las muestras.

**T-Test:**

El T-Test, por otro lado, se utiliza cuando la desviación estándar de la población es desconocida y se estima a partir de las muestras. Hay dos versiones comunes del T-Test: el T-Test para muestras independientes y el T-Test para muestras emparejadas.

- **T-Test para muestras independientes:**

  $t = \frac{(\bar{x}_1 - \bar{x}_2)}{\sqrt{\frac{s_1^2}{n_1} + \frac{s_2^2}{n_2}}}$

- **T-Test para muestras emparejadas:**

  $t = \frac{\bar{d}}{\frac{s_d}{\sqrt{n}}}$

Donde:

- $(\bar{d})$ es la media de las diferencias (solo para T-Test emparejado).
- $(s_1)$ y $(s_2)$ son las desviaciones estándar de las muestras.
- $(n_1)$ y $(n_2)$ son los tamaños de las muestras.
- $(s_d)$ es la desviación estándar de las diferencias (solo para T-Test emparejado).

**Ejemplo en Python:**

```python
import numpy as np
from scipy.stats import ttest_ind, ttest_rel
import matplotlib.pyplot as plt

# Generar datos de ejemplo
np.random.seed(42)
sample1 = np.random.normal(loc=5, scale=2, size=100)
sample2 = np.random.normal(loc=7, scale=2, size=100)

# T-Test para muestras independientes
t_stat, p_value = ttest_ind(sample1, sample2)
print(f"T-Test para muestras independientes: t = {t_stat}, p-value = {p_value}")

# T-Test para muestras emparejadas
sample_diff = sample1 - sample2
t_stat_pair, p_value_pair = ttest_rel(sample1, sample2)
print(f"T-Test para muestras emparejadas: t = {t_stat_pair}, p-value = {p_value_pair}")

# Personalizar la apariencia de la gráfica
plt.figure(figsize=(8, 6))
plt.hist(sample1, alpha=0.5, label='Sample 1', color='blue', edgecolor='black')
plt.hist(sample2, alpha=0.5, label='Sample 2', color='orange', edgecolor='black')

# Personalizar etiquetas y título
plt.xlabel('Valores', fontsize=14)
plt.ylabel('Frecuencia', fontsize=14)
plt.title('Distribución de Muestras', fontsize=16)

# Añadir leyenda
plt.legend(fontsize=12)

# Mostrar la cuadrícula
plt.grid(True, linestyle='--', alpha=0.5)

plt.show()
```

Este ejemplo de Python genera dos conjuntos de datos y realiza tanto un T-Test para muestras independientes como un T-Test para muestras emparejadas, mostrando el valor t y el valor p para cada caso.
