# **20. Bernoulli Distribution**

La Distribuci√≥n de Bernoulli es una distribuci√≥n de probabilidad discreta que describe un experimento aleatorio con dos posibles resultados: √©xito (1) o fracaso (0). Es especialmente √∫til cuando est√°s tratando con eventos binarios, donde solo hay dos resultados posibles.

**Terminolog√≠a:**

- **√âxito (1):** El resultado deseado del experimento.
- **Fracaso (0):** El resultado no deseado del experimento.
- **Probabilidad de √©xito (p):** La probabilidad de que ocurra un √©xito en un solo ensayo.

**F√≥rmula Matem√°tica:**

La funci√≥n de masa de probabilidad de la distribuci√≥n de Bernoulli se expresa como:

$P(X = k) = \begin{cases}
p & \text{si } k = 1 \\
q = 1 - p & \text{si } k = 0
\end{cases}$

donde:

- $P(X = k)$ es la probabilidad de que la variable aleatoria discreta $(X)$ tome el valor $(k)$,
- $( p )$ es la probabilidad de √©xito,
- $( q )$ es la probabilidad de fracaso $(1 - p)$.

**Ejemplo con Gr√°fico en Python:**

Vamos a generar un gr√°fico de la distribuci√≥n de Bernoulli usando Python:

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli

# Definir la probabilidad de √©xito
p = 0.3

# Crear una variable aleatoria de Bernoulli
rv = bernoulli(p)

# Valores posibles (0 y 1)
x = [0, 1]

# Probabilidades correspondientes
probabilities = [rv.pmf(value) for value in x]

# Gr√°fico de la distribuci√≥n de Bernoulli
plt.bar(x, probabilities, align='center', alpha=0.7)
plt.xticks(x, ['Fracaso (0)', '√âxito (1)'])
plt.title('Distribuci√≥n de Bernoulli (p = 0.3)')
plt.xlabel('Resultado')
plt.ylabel('Probabilidad')
plt.show()
```

Este c√≥digo utiliza la biblioteca SciPy para trabajar con la distribuci√≥n de Bernoulli y matplotlib para visualizarla. Ajusta el valor de $(p)$ seg√∫n tus necesidades.
