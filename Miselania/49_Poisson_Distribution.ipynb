# **49. Poisson Distribution**

**Distribución de Poisson: Explicación Detallada**

**Introducción:**

La distribución de Poisson es una distribución de probabilidad discreta que describe el número de eventos que ocurrirán en un intervalo de tiempo fijo o en un área espacial fija, bajo la condición de que estos eventos ocurren con una tasa constante y de manera independiente. Es especialmente útil para modelar eventos raros.

**Terminología:**

- **Variable Aleatoria (X):** Representa el número de eventos que ocurren en un intervalo o área específica.

- **Tasa (λ):** Representa la frecuencia promedio de eventos en el intervalo especificado.

**Fórmula Matemática:**

La probabilidad de observar exactamente $(k)$ eventos en un intervalo o área es dada por la fórmula de masa de probabilidad de Poisson:

$P(X = k) = \frac{e^{-\lambda} \cdot \lambda^k}{k!}$

Donde:

- $( e )$ es la base del logaritmo natural (aproximadamente 2.71828).
- $( \lambda )$ es la tasa de eventos.
- $( k! )$ es el factorial de $(k)$.

**Propiedades Importantes:**

- La media $( \mu )$ y la varianza $( \sigma^2 )$ de la distribución de Poisson son ambas igual a la tasa $( \lambda )$: $( \mu = \sigma^2 = \lambda )$.

**Ejemplo con Gráfico en Python:**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Parámetros de la distribución de Poisson
lambda_parameter = 3

# Crear datos para graficar la distribución
x_values = np.arange(0, 10)
poisson_distribution_values = poisson.pmf(x_values, lambda_parameter)

# Graficar la distribución de Poisson
plt.bar(x_values, poisson_distribution_values, color='blue', alpha=0.7, label=f'$\lambda={lambda_parameter}$')
plt.title('Distribución de Poisson')
plt.xlabel('Número de Eventos (k)')
plt.ylabel('Probabilidad')
plt.legend()
plt.show()
```

Este ejemplo utiliza la biblioteca `scipy` para generar y graficar la distribución de Poisson con un parámetro de tasa $( \lambda )$ establecido en 3. La gráfica muestra la probabilidad de observar diferentes cantidades de eventos. A medida que $( \lambda )$ aumenta, la distribución se sesga hacia la derecha y se vuelve más similar a una distribución normal.
