# **16. Function**

**Sigmoid Function:**

La funci贸n sigmoidal (o log铆stica) es una funci贸n de activaci贸n utilizada com煤nmente en redes neuronales y modelos de regresi贸n log铆stica. Su forma caracter铆stica es una curva S y transforma cualquier valor real en el rango de 0 a 1. Aqu铆 hay una explicaci贸n detallada:

**Terminolog铆a:**

1. **Funci贸n Sigmoidal (Log铆stica):**

 $\sigma(z) = \frac{1}{1 + e^{-z}}$

   - $\sigma(z)$: Salida de la funci贸n sigmoidal para una entrada $(z)$.
   - $( e )$: Base del logaritmo natural (aproximadamente 2.71828).

2. **Uso en Redes Neuronales:**

 La funci贸n sigmoidal se utiliza para introducir no linealidades en el modelo y mapear las entradas ponderadas a una probabilidad.

3. **Caracter铆sticas:**

   - Produce valores en el rango de 0 a 1.
   - til en problemas de clasificaci贸n binaria.

**Ejemplo en Python:**

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la funci贸n sigmoidal
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Generar valores de entrada
x_values = np.linspace(-7, 7, 200)

# Calcular valores de salida
sigmoid_values = sigmoid(x_values)

# Visualizar la funci贸n sigmoidal
plt.figure(figsize=(8, 6))
plt.plot(x_values, sigmoid_values, label='Sigmoid Function', color='blue', linewidth=2)
plt.title('Sigmoid Function')
plt.xlabel('Input (z)')
plt.ylabel('Output (sigmoid(z))')
plt.legend()
plt.grid(True)
plt.show()
```

**ReLU (Rectified Linear Unit):**

ReLU es otra funci贸n de activaci贸n ampliamente utilizada, especialmente en redes neuronales profundas. A diferencia de la funci贸n sigmoidal, ReLU es lineal para valores positivos y cero para valores negativos. Aqu铆 hay una explicaci贸n detallada:

**Terminolog铆a:**

1. **Funci贸n ReLU:**

 $f(x) = \max(0, x)$

   - $f(x)$: Salida de la funci贸n ReLU para una entrada $(x)$.

2. **Uso en Redes Neuronales:**

 ReLU introduce no linealidades en el modelo y es eficiente en t茅rminos computacionales.

3. **Caracter铆sticas:**

   - Activa la neurona solo si la entrada es positiva.
   - Proporciona convergencia m谩s r谩pida en el entrenamiento.

**Ejemplo en Python:**

```python
# Definir la funci贸n ReLU
def relu(x):
    return np.maximum(0, x)

# Calcular valores de salida para la funci贸n ReLU
relu_values = relu(x_values)

# Visualizar la funci贸n ReLU
plt.figure(figsize=(8, 6))
plt.plot(x_values, relu_values, label='ReLU Function', color='green', linewidth=2)
plt.title('ReLU Function')
plt.xlabel('Input (x)')
plt.ylabel('Output (ReLU(x))')
plt.legend()
plt.grid(True)
plt.show()
```

Ambos ejemplos muestran visualizaciones de las funciones sigmoidal y ReLU. Estas funciones son fundamentales en la construcci贸n y entrenamiento de modelos de aprendizaje profundo.
