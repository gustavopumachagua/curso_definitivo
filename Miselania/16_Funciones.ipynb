# **16. Function**

**Sigmoid Function:**

La función sigmoidal (o logística) es una función de activación utilizada comúnmente en redes neuronales y modelos de regresión logística. Su forma característica es una curva S y transforma cualquier valor real en el rango de 0 a 1. Aquí hay una explicación detallada:

**Terminología:**

1. **Función Sigmoidal (Logística):**

 $\sigma(z) = \frac{1}{1 + e^{-z}}$

   - $\sigma(z)$: Salida de la función sigmoidal para una entrada $(z)$.
   - $( e )$: Base del logaritmo natural (aproximadamente 2.71828).

2. **Uso en Redes Neuronales:**

 La función sigmoidal se utiliza para introducir no linealidades en el modelo y mapear las entradas ponderadas a una probabilidad.

3. **Características:**

   - Produce valores en el rango de 0 a 1.
   - Útil en problemas de clasificación binaria.

**Ejemplo en Python:**

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función sigmoidal
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Generar valores de entrada
x_values = np.linspace(-7, 7, 200)

# Calcular valores de salida
sigmoid_values = sigmoid(x_values)

# Visualizar la función sigmoidal
plt.figure(figsize=(8, 6))
plt.plot(x_values, sigmoid_values, label='Sigmoid Function', color='blue', linewidth=2)
plt.title('Sigmoid Function')
plt.xlabel('Input (z)')
plt.ylabel('Output (sigmoid(z))')
plt.legend()
plt.grid(True)
plt.show()
```

**ReLU (Rectified Linear Unit):**

ReLU es otra función de activación ampliamente utilizada, especialmente en redes neuronales profundas. A diferencia de la función sigmoidal, ReLU es lineal para valores positivos y cero para valores negativos. Aquí hay una explicación detallada:

**Terminología:**

1. **Función ReLU:**

 $f(x) = \max(0, x)$

   - $f(x)$: Salida de la función ReLU para una entrada $(x)$.

2. **Uso en Redes Neuronales:**

 ReLU introduce no linealidades en el modelo y es eficiente en términos computacionales.

3. **Características:**

   - Activa la neurona solo si la entrada es positiva.
   - Proporciona convergencia más rápida en el entrenamiento.

**Ejemplo en Python:**

```python
# Definir la función ReLU
def relu(x):
    return np.maximum(0, x)

# Calcular valores de salida para la función ReLU
relu_values = relu(x_values)

# Visualizar la función ReLU
plt.figure(figsize=(8, 6))
plt.plot(x_values, relu_values, label='ReLU Function', color='green', linewidth=2)
plt.title('ReLU Function')
plt.xlabel('Input (x)')
plt.ylabel('Output (ReLU(x))')
plt.legend()
plt.grid(True)
plt.show()
```

Ambos ejemplos muestran visualizaciones de las funciones sigmoidal y ReLU. Estas funciones son fundamentales en la construcción y entrenamiento de modelos de aprendizaje profundo.
