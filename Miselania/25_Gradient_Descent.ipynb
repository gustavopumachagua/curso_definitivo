# **25. Gradient Descent**

El descenso del gradiente es un algoritmo de optimización utilizado para encontrar el mínimo de una función. Es especialmente útil en problemas de optimización en los que se busca minimizar una función de pérdida en un espacio de parámetros.

**Terminología:**

1. **Función Objetivo (o Función de Pérdida):** La función que queremos minimizar. En el contexto del aprendizaje automático, esta función mide la diferencia entre las predicciones del modelo y los valores reales.

2. **Gradiente:** Un vector de derivadas parciales que indica la dirección y la magnitud del cambio más pronunciado de la función objetivo.

3. **Tasa de Aprendizaje (Learning Rate):** Un hiperparámetro que controla el tamaño de los pasos que damos durante la optimización.

4. **Parámetros:** Los valores ajustables de un modelo que queremos encontrar para minimizar la función objetivo.

**Fórmula Matemática:**

El descenso del gradiente sigue la dirección opuesta al gradiente de la función objetivo. La actualización de los parámetros se realiza mediante la siguiente fórmula:

$\theta = \theta - \alpha \cdot \nabla J(\theta)$

Donde:

- $(\theta)$ son los parámetros del modelo.
- $(\alpha)$ es la tasa de aprendizaje.
- $\nabla J(\theta)$ es el gradiente de la función de pérdida con respecto a los parámetros.

**Ejemplo con Gráfico en Python:**

Vamos a ilustrar el descenso del gradiente en un problema simple de regresión lineal:

```python
import numpy as np
import matplotlib.pyplot as plt

# Función objetivo (en este caso, una regresión lineal)
def objective_function(theta, x):
    return theta[0] + theta[1] * x

# Gradiente de la función objetivo (en este caso, derivadas parciales de la regresión lineal)
def gradient(theta, x, y):
    grad_0 = 2 * np.mean(objective_function(theta, x) - y)
    grad_1 = 2 * np.mean((objective_function(theta, x) - y) * x)
    return np.array([grad_0, grad_1])

# Descenso del gradiente
def gradient_descent(x, y, alpha=0.01, epochs=100):
    theta = np.zeros(2)  # Inicialización de parámetros
    history = [theta.copy()]  # Historial de parámetros para visualización
    for _ in range(epochs):
        grad = gradient(theta, x, y)
        theta = theta - alpha * grad
        history.append(theta.copy())
    return np.array(history)

# Datos de ejemplo
np.random.seed(42)
x = 2 * np.random.rand(100, 1)
y = 4 + 3 * x + np.random.randn(100, 1)

# Aplicar descenso del gradiente
history = gradient_descent(x, y, alpha=0.01, epochs=100)

# Visualizar el proceso de optimización
plt.scatter(x, y, label='Datos de entrenamiento')
for i in range(0, 100, 10):
    plt.plot(x, objective_function(history[i], x), label=f'Iteración {i}')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Descenso del Gradiente: Regresión Lineal')
plt.legend()
plt.show()
```

Este código implementa una regresión lineal simple y utiliza el descenso del gradiente para encontrar los parámetros (pendiente e intercepto) que minimizan la función de pérdida. La visualización muestra cómo la línea de regresión se ajusta a los datos con el paso de las iteraciones. La tasa de aprendizaje y el número de épocas son hiperparámetros que puedes ajustar para observar su impacto en el proceso de optimización.
