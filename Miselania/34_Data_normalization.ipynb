# **34. Data normalization**

**Terminolog√≠a:**

La normalizaci√≥n de datos, tambi√©n conocida como escalamiento, es un paso com√∫n en el preprocesamiento de datos en machine learning. Su objetivo es ajustar los valores de las caracter√≠sticas para que est√©n dentro de un rango espec√≠fico, lo que puede mejorar el rendimiento de ciertos algoritmos y facilitar la convergencia del modelo.

**M√©todos Comunes de Normalizaci√≥n:**

1. **Min-Max Scaling:**

   $X_{\text{norm}} = \frac{X - X_{\text{min}}}{X_{\text{max}} - X_{\text{min}}}$

   **Donde:**

   - $(X)$ es el valor original.
   - $(X_{\text{min}})$ es el valor m√≠nimo en la caracter√≠stica.
   - $(X_{\text{max}})$ es el valor m√°ximo en la caracter√≠stica.

2. **Z-Score Normalization (Standardization):**

   $X_{\text{norm}} = \frac{X - \mu}{\sigma}$

   **Donde:**

   - $(X)$ es el valor original.
   - $(\mu)$ es la media de la caracter√≠stica.
   - $(\sigma)$ es la desviaci√≥n est√°ndar de la caracter√≠stica.

**Ejemplo con Gr√°fico en Python:**

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos de ejemplo
data = np.array([1, 5, 10, 20, 50, 100], dtype=float)

# Min-Max Scaling
min_max_scaled = (data - np.min(data)) / (np.max(data) - np.min(data))

# Z-Score Normalization
z_score_normalized = (data - np.mean(data)) / np.std(data)

# Personalizar la apariencia de la gr√°fica
plt.figure(figsize=(12, 4))

# Gr√°fica de datos originales
plt.subplot(1, 3, 1)
plt.plot(data, marker='o', linestyle='-', color='b')
plt.title('Datos Originales')
plt.xlabel('√çndice')
plt.ylabel('Valor')
plt.grid(True, linestyle='--', alpha=0.7)

# Gr√°fica de Min-Max Scaling
plt.subplot(1, 3, 2)
plt.plot(min_max_scaled, marker='o', linestyle='-', color='r')
plt.title('Min-Max Scaling')
plt.xlabel('√çndice')
plt.ylabel('Valor Normalizado [0, 1]')
plt.grid(True, linestyle='--', alpha=0.7)

# Gr√°fica de Z-Score Normalization
plt.subplot(1, 3, 3)
plt.plot(z_score_normalized, marker='o', linestyle='-', color='g')
plt.title('Z-Score Normalization')
plt.xlabel('√çndice')
plt.ylabel('Valor Normalizado (Z-Score)')
plt.grid(True, linestyle='--', alpha=0.7)

# Ajustar el dise√±o de las subgr√°ficas
plt.tight_layout()

# Mostrar la gr√°fica
plt.show()
```

En este ejemplo, se utiliza Min-Max Scaling y Z-Score Normalization para normalizar un conjunto de datos. El gr√°fico muestra la distribuci√≥n de los datos originales y c√≥mo cambian despu√©s de aplicar cada m√©todo de normalizaci√≥n.
