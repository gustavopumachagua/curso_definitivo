# **27. Uniform Distribution**

La distribuci贸n uniforme es un tipo de distribuci贸n de probabilidad en la que todas las variables tienen la misma probabilidad de ocurrir. En otras palabras, cada valor posible de la variable aleatoria tiene la misma probabilidad de ser observado.

**Terminolog铆a:**

1. **$(a)$ y $(b)$:** Son los par谩metros que definen el intervalo en el que la variable aleatoria puede tomar valores. \(a\) es el valor m铆nimo y \(b\) es el valor m谩ximo.

**Funci贸n de Densidad de Probabilidad (PDF):**

La funci贸n de densidad de probabilidad (PDF) de una distribuci贸n uniforme en el intervalo $[a, b]$ es constante dentro del intervalo y cero fuera de 茅l. La f贸rmula es:

$f(x|a, b) = \frac{1}{b - a}$

donde:
- $(a)$ es el valor m铆nimo del intervalo.
- $(b)$ es el valor m谩ximo del intervalo.

**Ejemplo con Gr谩fico en Python:**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform

# Par谩metros de la distribuci贸n uniforme
a = 0
b = 1

# Generar datos para la distribuci贸n uniforme
x = np.linspace(-0.1, 1.1, 1000)
pdf = uniform.pdf(x, a, b - a)

# Visualizar la distribuci贸n uniforme
plt.plot(x, pdf, label=f'U({a}, {b})')
plt.title('Distribuci贸n Uniforme')
plt.xlabel('X')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.show()
```

En este ejemplo, se genera y visualiza una distribuci贸n uniforme en el intervalo $[0, 1]$. Puedes ajustar los valores de $(a)$ y $(b)$ para explorar diferentes distribuciones uniformes. La gr谩fica muestra que la densidad de probabilidad es constante dentro del intervalo y cero fuera de 茅l.
