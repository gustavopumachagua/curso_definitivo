# **27. Uniform Distribution**

La distribución uniforme es un tipo de distribución de probabilidad en la que todas las variables tienen la misma probabilidad de ocurrir. En otras palabras, cada valor posible de la variable aleatoria tiene la misma probabilidad de ser observado.

**Terminología:**

1. **$(a)$ y $(b)$:** Son los parámetros que definen el intervalo en el que la variable aleatoria puede tomar valores. \(a\) es el valor mínimo y \(b\) es el valor máximo.

**Función de Densidad de Probabilidad (PDF):**

La función de densidad de probabilidad (PDF) de una distribución uniforme en el intervalo $[a, b]$ es constante dentro del intervalo y cero fuera de él. La fórmula es:

$f(x|a, b) = \frac{1}{b - a}$

donde:
- $(a)$ es el valor mínimo del intervalo.
- $(b)$ es el valor máximo del intervalo.

**Ejemplo con Gráfico en Python:**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform

# Parámetros de la distribución uniforme
a = 0
b = 1

# Generar datos para la distribución uniforme
x = np.linspace(-0.1, 1.1, 1000)
pdf = uniform.pdf(x, a, b - a)

# Visualizar la distribución uniforme
plt.plot(x, pdf, label=f'U({a}, {b})')
plt.title('Distribución Uniforme')
plt.xlabel('X')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.show()
```

En este ejemplo, se genera y visualiza una distribución uniforme en el intervalo $[0, 1]$. Puedes ajustar los valores de $(a)$ y $(b)$ para explorar diferentes distribuciones uniformes. La gráfica muestra que la densidad de probabilidad es constante dentro del intervalo y cero fuera de él.
