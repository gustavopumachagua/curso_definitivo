# **26. Normal Distribution**

La distribución normal, también conocida como distribución gaussiana, es una de las distribuciones de probabilidad más importantes en estadísticas y teoría de probabilidad. Es completamente especificada por dos parámetros: la media $(\mu)$ y la desviación estándar $(\sigma)$. La forma de la distribución es simétrica en torno a la media y tiene forma de campana.

**Terminología:**

1. **Media $(\mu)$:** Representa el valor central de la distribución.

2. **Desviación Estándar $(\sigma)$:** Indica cuánto se dispersan los valores con respecto a la media. Una desviación estándar mayor implica una distribución más amplia.

3. **Varianza $(\sigma^2)$:** Es el cuadrado de la desviación estándar. Proporciona una medida de la dispersión de los valores.

**Función de Densidad de Probabilidad (PDF):**

La fórmula de la función de densidad de probabilidad (PDF) de una distribución normal es:

$f(x|\mu, \sigma) = \frac{1}{\sqrt{2\pi\sigma^2}} \cdot e^{-\frac{(x-\mu)^2}{2\sigma^2}}$

Donde:

- $(x)$ es la variable aleatoria.
- $(\mu)$ es la media.
- $(\sigma)$ es la desviación estándar.
- $(e)$ es la base del logaritmo natural.

**Ejemplo con Gráfico en Python:**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parámetros de la distribución normal
mu = 0
sigma = 1

# Generar datos para la distribución normal
x = np.linspace(-5, 5, 100)
pdf = norm.pdf(x, mu, sigma)

# Visualizar la distribución normal con personalización
plt.figure(figsize=(8, 6))
plt.plot(x, pdf, label=f'N({mu}, {sigma})', color='blue', linewidth=2)

# Línea vertical en la media
plt.axvline(mu, color='red', linestyle='--', label='Media')

# Área sombreada bajo la curva
plt.fill_between(x, pdf, color='blue', alpha=0.1, label='Área sombreada')

# Personalizaciones adicionales
plt.title('Distribución Normal (Gaussiana)')
plt.xlabel('X')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

En este ejemplo, se genera y visualiza una distribución normal estándar (con media $(\mu = 0)$ y desviación estándar $(\sigma = 1)$. Puedes ajustar los valores de $(\mu)$ y $(\sigma)$ para explorar diferentes distribuciones normales. La gráfica muestra la forma de campana característica de la distribución normal.
