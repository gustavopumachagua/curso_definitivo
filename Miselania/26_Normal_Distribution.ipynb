# **26. Normal Distribution**

La distribuci贸n normal, tambi茅n conocida como distribuci贸n gaussiana, es una de las distribuciones de probabilidad m谩s importantes en estad铆sticas y teor铆a de probabilidad. Es completamente especificada por dos par谩metros: la media $(\mu)$ y la desviaci贸n est谩ndar $(\sigma)$. La forma de la distribuci贸n es sim茅trica en torno a la media y tiene forma de campana.

**Terminolog铆a:**

1. **Media $(\mu)$:** Representa el valor central de la distribuci贸n.

2. **Desviaci贸n Est谩ndar $(\sigma)$:** Indica cu谩nto se dispersan los valores con respecto a la media. Una desviaci贸n est谩ndar mayor implica una distribuci贸n m谩s amplia.

3. **Varianza $(\sigma^2)$:** Es el cuadrado de la desviaci贸n est谩ndar. Proporciona una medida de la dispersi贸n de los valores.

**Funci贸n de Densidad de Probabilidad (PDF):**

La f贸rmula de la funci贸n de densidad de probabilidad (PDF) de una distribuci贸n normal es:

$f(x|\mu, \sigma) = \frac{1}{\sqrt{2\pi\sigma^2}} \cdot e^{-\frac{(x-\mu)^2}{2\sigma^2}}$

Donde:

- $(x)$ es la variable aleatoria.
- $(\mu)$ es la media.
- $(\sigma)$ es la desviaci贸n est谩ndar.
- $(e)$ es la base del logaritmo natural.

**Ejemplo con Gr谩fico en Python:**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Par谩metros de la distribuci贸n normal
mu = 0
sigma = 1

# Generar datos para la distribuci贸n normal
x = np.linspace(-5, 5, 100)
pdf = norm.pdf(x, mu, sigma)

# Visualizar la distribuci贸n normal con personalizaci贸n
plt.figure(figsize=(8, 6))
plt.plot(x, pdf, label=f'N({mu}, {sigma})', color='blue', linewidth=2)

# L铆nea vertical en la media
plt.axvline(mu, color='red', linestyle='--', label='Media')

# rea sombreada bajo la curva
plt.fill_between(x, pdf, color='blue', alpha=0.1, label='rea sombreada')

# Personalizaciones adicionales
plt.title('Distribuci贸n Normal (Gaussiana)')
plt.xlabel('X')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

En este ejemplo, se genera y visualiza una distribuci贸n normal est谩ndar (con media $(\mu = 0)$ y desviaci贸n est谩ndar $(\sigma = 1)$. Puedes ajustar los valores de $(\mu)$ y $(\sigma)$ para explorar diferentes distribuciones normales. La gr谩fica muestra la forma de campana caracter铆stica de la distribuci贸n normal.
