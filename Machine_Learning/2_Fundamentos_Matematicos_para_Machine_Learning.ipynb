a. **Vectores y Matrices:**

Un vector es una secuencia de n煤meros, y una matriz es una colecci贸n bidimensional de n煤meros.

**Ejemplo:**

2. **C谩lculo:**

El c谩lculo es esencial para entender c贸mo los modelos de machine learning aprenden y optimizan. Aqu铆 se mencionan dos conceptos importantes:

a. **Derivadas:**

La derivada de una funci贸n describe la tasa de cambio instant谩neo de la funci贸n en un punto dado.

**Ejemplo:**

```python
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np

# Definir la funci贸n
x = sp.symbols('x')
f = x**2

# Calcular la derivada
f_prime = sp.diff(f, x)

# Graficar la funci贸n y su derivada
x_vals = np.linspace(-5, 5, 100)
f_vals = np.array([f.subs(x, val) for val in x_vals])
f_prime_vals = np.array([f_prime.subs(x, val) for val in x_vals])

plt.plot(x_vals, f_vals, label='$f(x) = x^2$')
plt.plot(x_vals, f_prime_vals, label="f'(x) = 2x", linestyle='dashed')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Funci贸n y su Derivada')
plt.show()
```

b. **Integrales:**

La integral de una funci贸n representa el 谩rea bajo la curva de la funci贸n.

**Ejemplo:**

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# Definir la funci贸n
x = sp.symbols('x')
f = x**2

# Calcular la derivada
f_prime = sp.diff(f, x)

# Convertir la funci贸n y la derivada a funciones de numpy
f_np = sp.lambdify(x, f, 'numpy')
f_prime_np = sp.lambdify(x, f_prime, 'numpy')

# Generar datos
x_vals = np.linspace(-5, 5, 100)
f_vals = f_np(x_vals)
f_prime_vals = f_prime_np(x_vals)

plt.plot(x_vals, f_vals, label='$f(x) = x^2$')
plt.plot(x_vals, f_prime_vals, label="f'(x) = 2x", linestyle='dashed')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Funci贸n y su Derivada')
plt.show()
```

**Aplicaci贸n en Machine Learning:**

- **Gradientes y Descenso de Gradiente:**
  - En optimizaci贸n de modelos, el gradiente (derivada) se utiliza para encontrar los m铆nimos locales de una funci贸n de p茅rdida.
  - El descenso de gradiente ajusta iterativamente los par谩metros del modelo para minimizar la funci贸n de p茅rdida.

- **Integraci贸n en Features Polin贸micas:**
  - En modelos polin贸micos, la integraci贸n se puede utilizar para calcular el 谩rea bajo la curva de una caracter铆stica, lo que puede ser 煤til en la ingenier铆a de caracter铆sticas.

El entendimiento de estos conceptos proporciona una base s贸lida para comprender el funcionamiento interno de algoritmos y modelos de machine learning.

## **Probabilidades y Estad铆sticas**

### **Fundamentos estad铆sticos para entender la incertidumbre en los modelos**

1. **Probabilidad:**

La probabilidad es una medida num茅rica de la certeza o posibilidad de que ocurra un evento. En machine learning, se utiliza para cuantificar la incertidumbre asociada con las predicciones de un modelo.

a. **Probabilidad de un Evento:**

$P(A) = \frac{\text{N煤mero de casos favorables para } A}{\text{N煤mero total de casos posibles}}$

b. **Regla de la Suma:**

$P(A \cup B) = P(A) + P(B) - P(A \cap B)$

**Ejemplo:**

```python
import matplotlib.pyplot as plt
import numpy as np

# Generar datos aleatorios con distribuci贸n binomial
data_binomial = np.random.binomial(n=10, p=0.5, size=1000)

# Calcular la probabilidad emp铆rica
prob_success = np.sum(data_binomial == 5) / len(data_binomial)

# Graficar el histograma con personalizaci贸n
plt.hist(data_binomial, bins=np.arange(0, 12) - 0.5, density=True, alpha=0.75, color='skyblue', edgecolor='black', linewidth=1.2)
plt.axvline(x=5, color='red', linestyle='--', label='Valor deseado')

# Personalizar etiquetas y t铆tulos
plt.title('Distribuci贸n Binomial Personalizada', fontsize=16)
plt.xlabel('N煤mero de xitos', fontsize=14)
plt.ylabel('Frecuencia Relativa', fontsize=14)
plt.legend()

# A帽adir cuadr铆cula
plt.grid(True, linestyle='--', alpha=0.5)

# A帽adir texto
plt.text(5, 0.08, f'Probabilidad Emp铆rica: {prob_success:.2f}', fontsize=12, ha='center', va='center', color='purple')

# Mostrar la gr谩fica
plt.show()
```

2. **Distribuciones de Probabilidad:**

Las distribuciones de probabilidad describen la probabilidad de ocurrencia de cada posible resultado en un conjunto de datos.

a. **Distribuci贸n Normal (Gaussiana):**

$f(x|\mu, \sigma) = \frac{1}{\sigma\sqrt{2\pi}}e^{-\frac{1}{2}\left(\frac{x-\mu}{\sigma}\right)^2}$

**Ejemplo:**

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

# Generar datos aleatorios con distribuci贸n normal
data_normal = np.random.normal(loc=0, scale=1, size=1000)

# Graficar el histograma
plt.hist(data_normal, bins=30, density=True, alpha=0.75, color='skyblue', edgecolor='black', linewidth=1.2)

# Graficar la funci贸n de densidad de probabilidad
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, 0, 1)
plt.plot(x, p, 'k', linestyle='dashed', linewidth=2, label='Densidad de Probabilidad')

# Personalizar etiquetas y t铆tulos
plt.title('Distribuci贸n Normal Personalizada', fontsize=16)
plt.xlabel('Valor', fontsize=14)
plt.ylabel('Densidad de Probabilidad', fontsize=14)

# A帽adir leyenda
plt.legend()

# A帽adir cuadr铆cula
plt.grid(True, linestyle='--', alpha=0.5)

# Mostrar la gr谩fica
plt.show()
```

3. **Estad铆sticas Descriptivas:**

Las estad铆sticas descriptivas resumen y describen las caracter铆sticas principales de un conjunto de datos.

a. **Media y Desviaci贸n Est谩ndar:**

$\text{Media} = \frac{\sum_{i=1}^{n} x_i}{n}$

$\text{Desviaci贸n Est谩ndar} = \sqrt{\frac{\sum_{i=1}^{n} (x_i - \text{Media})^2}{n}}$

**Ejemplo:**

```python
import numpy as np

# Generar datos aleatorios
data = np.random.normal(loc=5, scale=2, size=1000)

# Calcular media y desviaci贸n est谩ndar
mean_val = np.mean(data)
std_dev = np.std(data)

print(f'Media: {mean_val:.2f}')
print(f'Desviaci贸n Est谩ndar: {std_dev:.2f}')
```

4. **Intervalos de Confianza:**

Los intervalos de confianza proporcionan un rango plausible de valores para un par谩metro desconocido.

$\text{Intervalo de Confianza} = \text{Estimador} \pm \text{Margen de Error}$

**Ejemplo:**

```python
from scipy import stats

# Generar datos aleatorios
data = np.random.normal(loc=5, scale=2, size=1000)

# Calcular intervalo de confianza para la media
confidence_interval = stats.norm.interval(0.95, loc=np.mean(data), scale=stats.sem(data))

print(f'Intervalo de Confianza para la Media: {confidence_interval}')
```

**Aplicaci贸n en Machine Learning:**

- **Incertidumbre en Predicciones:**

  - Los intervalos de confianza pueden utilizarse para cuantificar la incertidumbre alrededor de las predicciones de un modelo, proporcionando l铆mites plausibles para los valores predichos.

- **Evaluaci贸n de Modelos:**

  - Las estad铆sticas descriptivas y las distribuciones de probabilidad ayudan a evaluar el rendimiento de los modelos y comprender la variabilidad en los datos.

- **Toma de Decisiones:**

  - La comprensi贸n de la probabilidad y la incertidumbre es esencial para tomar decisiones informadas basadas en los resultados de los modelos.

Estos conceptos estad铆sticos son fundamentales para entender y gestionar la incertidumbre en los modelos de machine learning. Incorporarlos en el an谩lisis de datos y la evaluaci贸n de modelos fortalece la capacidad de interpretar y utilizar los resultados de manera efectiva.

## **Matriz de Confusi贸n:**

![Matrix Confusion](../img/matrix_confusion.jpg "Matrix Confusion")

La matriz de confusi贸n es una tabla que se utiliza en problemas de clasificaci贸n para evaluar el rendimiento de un modelo. Para un problema de clasificaci贸n binaria, la matriz de confusi贸n tiene cuatro entradas principales:



**F贸rmulas asociadas:**

- **Precisi贸n (Precision):**

 $(\frac{TP}{TP + FP})$

- **Sensibilidad o tasa de verdaderos positivos (Recall):**

 $(\frac{TP}{TP + FN})$

- **Exactitud (Accuracy):**

$(\frac{TP + TN}{TP + TN + FP + FN})$

- **Especificidad:**

$(\frac{TN}{TN + FP})$

- **Valor predictivo negativo (NPV):**

 $(\frac{TN}{TN + FN})$

- **Tasa de falsos positivos (FPR):**

 $(\frac{FP}{FP + TN})$

- **F1-Score:**

 $(\frac{2 \cdot \text{Precision} \cdot \text{Recall}}{\text{Precision} + \text{Recall}})$

**Ejemplo con DataFrame y gr谩ficos en Python:**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

# Ejemplo de un DataFrame
data = {'Real': [1, 0, 1, 2, 1, 0, 1, 2, 2, 0],
        'Predicho': [1, 0, 1, 2, 0, 0, 1, 2, 1, 1]}

df = pd.DataFrame(data)

# Calcular la matriz de confusi贸n
cm = confusion_matrix(df['Real'], df['Predicho'])

# Configurar el gr谩fico
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", cbar=False,
            xticklabels=["Clase 0", "Clase 1", "Clase 2"],
            yticklabels=["Clase 0", "Clase 1", "Clase 2"])
plt.xlabel("Clase predicha")
plt.ylabel("Clase real")
plt.title("Matriz de Confusi贸n")

# Mostrar el gr谩fico
plt.show()
```

Este c贸digo utiliza un DataFrame de pandas para almacenar las etiquetas reales y predichas. Luego, se calcula la matriz de confusi贸n con `confusion_matrix` de scikit-learn y se visualiza con un mapa de calor usando `heatmap` de seaborn.
