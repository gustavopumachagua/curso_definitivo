# **3. Recopilación y Análisis de Requisitos**

## **Identificación y Documentación de Requisitos**

### **Técnicas para recopilar y priorizar requisitos**

1. **Identificación de Requisitos:**

La identificación de requisitos es un proceso crucial en el análisis de negocios. Implica la recopilación y documentación de las necesidades y expectativas de los stakeholders. Aquí se detallan los pasos y ejemplos:

**Pasos:**

1. **Entender el Contexto:**

   - Realizar entrevistas con stakeholders.
   - Revisar documentos existentes.

2. **Identificar Stakeholders:**

   - Listar todas las partes interesadas.
   - Determinar sus expectativas y requisitos.

3. **Elicitación de Requisitos:**

   - Utilizar diversas técnicas (entrevistas, encuestas, talleres).
   - Ejemplo: **Entrevista con el Cliente:**
     - **Pregunta:** "¿Cuáles son los objetivos clave de su negocio?"

4. **Análisis de Documentos:**

   - Revisar informes, contratos, políticas, etc.
   - Identificar requisitos implícitos y explícitos.

2. **Documentación de Requisitos:**

La documentación precisa es esencial para evitar malentendidos. Se pueden utilizar formatos como listas, diagramas y casos de uso. Ejemplo:

**Formato de Documento:**

**Requisito:**
- **ID:** REQ001
- **Descripción:** El sistema debe permitir a los usuarios registrarse.
- **Prioridad:** Alta
- **Stakeholder:** Usuario final
- **Comentarios:** Necesario para acceder a funciones personalizadas.


**Técnicas para Recopilar y Priorizar Requisitos:**

1. **Técnicas de Recopilación:**

**Entrevistas:**

- **Descripción:** Conversaciones directas con stakeholders.
- **Ejemplo:** Entrevista con el equipo de marketing para comprender los requisitos del nuevo sistema de CRM.

**Observación:**

- **Descripción:** Observar directamente las operaciones.
- **Ejemplo:** Observar cómo los empleados ingresan datos en el sistema actual.

**Cuestionarios:**

- **Descripción:** Formular preguntas estructuradas.
- **Ejemplo:** Enviar cuestionarios a los usuarios finales para recopilar sus necesidades y expectativas.

2. **Técnicas de Priorización:**

**Matriz de Priorización:**

- **Descripción:** Asignar valores numéricos a cada requisito.
- **Ejemplo:**

  | Requisito | Importancia (1-5) | Complejidad (1-5) | Puntuación |
  |-----------|---------------------|-------------------|------------|
  | REQ001    | 4                   | 3                 | 12         |
  | REQ002    | 5                   | 2                 | 10         |

**Método de MoSCoW:**

- **Descripción:** Clasificar requisitos en Must-haves, Should-haves, Could-haves y Won't-haves.
- **Ejemplo:** Marcar un requisito como "Must-have" si es crítico para el éxito del proyecto.

**Conclusiones:**

La identificación y documentación adecuadas de requisitos son esenciales para el éxito del proyecto. Utilizar técnicas efectivas de recopilación y priorización garantiza que los requisitos se capturen de manera completa y se aborden según su importancia.

### **Creación de especificaciones de requisitos**

La creación de especificaciones de requisitos es un paso crucial en el análisis de negocios. Estas especificaciones detalladas proporcionan una guía clara para el diseño y desarrollo del sistema. Aquí se presentan los pasos y ejemplos para la creación de especificaciones de requisitos:

1. **Detalles Generales:**

   - **Nombre del Proyecto:**
   - **Fecha:**
   - **Versión:**
   - **Equipo Responsable:**

2. **Descripción del Sistema:**

   - **Descripción General del Sistema:**
     - Descripción narrativa del propósito y la funcionalidad del sistema.

3. **Requisitos Funcionales:**

   - **RF001: Registro de Usuario:**

     - *Descripción:* El sistema debe permitir a los usuarios registrarse proporcionando su nombre, correo electrónico y contraseña.
     - *Precondiciones:* El usuario debe tener acceso a internet.
     - *Postcondiciones:* El sistema crea una cuenta de usuario única.

   - **RF002: Inicio de Sesión:**

     - *Descripción:* El sistema debe permitir a los usuarios iniciar sesión utilizando su correo electrónico y contraseña.
     - *Precondiciones:* El usuario debe estar registrado en el sistema.
     - *Postcondiciones:* El sistema redirige al usuario a su panel de control.

4. **Requisitos No Funcionales:**

   - **RNF001: Rendimiento del Sistema:**
     - *Descripción:* El sistema debe cargar las páginas en menos de 3 segundos.
     - *Criterio de Aceptación:* Todas las páginas deben cargar dentro del límite de tiempo especificado.

   - **RNF002: Seguridad:**
     - *Descripción:* Los datos de los usuarios deben almacenarse de forma segura utilizando cifrado.
     - *Criterio de Aceptación:* La auditoría de seguridad debe aprobar el sistema después de la implementación.

5. **Requisitos de Interfaz:**

   - **RI001: Interfaz de Usuario:**
     - *Descripción:* El sistema debe tener una interfaz de usuario intuitiva con menús de navegación claramente etiquetados.
     - *Criterio de Aceptación:* Los usuarios deben poder completar tareas comunes sin capacitación adicional.

   - **RI002: Integración con API Externa:**
     - *Descripción:* El sistema debe integrarse con una API externa para la autenticación de usuarios.
     - *Criterio de Aceptación:* Las pruebas de integración deben validar la comunicación correcta con la API.

6. **Requisitos de Rendimiento:**

   - **RR001: Tiempo de Respuesta:**
     - *Descripción:* El sistema debe responder a las solicitudes del usuario en menos de 1 segundo.
     - *Criterio de Aceptación:* Las pruebas de carga deben verificar el rendimiento bajo condiciones de estrés.

7. **Requisitos de Seguridad:**

   - **RS001: Acceso Autorizado:**
     - *Descripción:* Solo los usuarios autorizados deben tener acceso a funciones administrativas.
     - *Criterio de Aceptación:* La autenticación de dos factores debe implementarse para las cuentas de administrador.

8. **Anexos:**

   - **Diagramas de Flujo:**
     - Adjuntar diagramas que ilustren el flujo de trabajo del sistema.

**Conclusiones:**

La creación de especificaciones de requisitos es esencial para el desarrollo exitoso de un sistema. Estas especificaciones proporcionan una base sólida para el diseño, la implementación y la prueba, asegurando que el producto final cumpla con las expectativas y necesidades de los stakeholders.

## **Validación y Verificación de Requisitos**

### **Aseguramiento de la calidad en los requisitos**

El aseguramiento de la calidad en requisitos es un proceso fundamental para garantizar que los requisitos del sistema sean claros, completos, consistentes y verificables. Aquí hay un enfoque paso a paso con ejemplos:

1. **Revisión por Pares:**

   - **Definición:** Dos o más personas revisan los requisitos de manera conjunta.
   - **Ejemplo:** Dos analistas de negocios revisan los requisitos funcionales para el módulo de autenticación.

2. **Validación con Stakeholders:**

   - **Definición:** Verificar que los requisitos reflejen las expectativas y necesidades de los stakeholders.
   - **Ejemplo:** Presentar los requisitos del sistema a los usuarios finales para su revisión y comentarios.

3. **Trazabilidad de Requisitos:**

   - **Definición:** Establecer una relación clara entre requisitos y fuentes, como documentos comerciales o solicitudes de cambio.
   - **Ejemplo:** Asociar cada requisito con el caso de uso correspondiente.

4. **Consistencia y Claridad:**

   - **Definición:** Asegurarse de que no haya ambigüedades y que los requisitos no entren en conflicto entre sí.
   - **Ejemplo:** Revisar que los términos utilizados en los requisitos sean coherentes en todo el documento.

5. **Verificabilidad:**

   - **Definición:** Confirmar que se pueden establecer criterios objetivos para verificar si un requisito se cumple.
   - **Ejemplo:** Especificar métricas claras para evaluar el rendimiento del sistema.

6. **Estandarización y Plantillas:**

   - **Definición:** Utilizar plantillas y estándares predefinidos para la documentación de requisitos.
   - **Ejemplo:** Adoptar una plantilla específica que incluya secciones como descripción, prioridad y criterios de aceptación.

7. **Pruebas de Aceptación:**

   - **Definición:** Desarrollar casos de prueba basados en requisitos para garantizar que el sistema cumple con las expectativas.
   - **Ejemplo:** Crear casos de prueba para validar que todas las funciones especificadas en los requisitos funcionales están implementadas correctamente.

8. **Auditorías y Revisiones Externas:**

   - **Definición:** Contratar a terceros para realizar auditorías independientes de la calidad de los requisitos.
   - **Ejemplo:** Contratar a una empresa de consultoría para realizar una revisión externa de los requisitos.

9. **Gestión de Cambios:**

   - **Definición:** Implementar un proceso para gestionar cambios en los requisitos de manera controlada.
   - **Ejemplo:** Establecer un comité de cambio que evalúe y apruebe o rechace solicitudes de cambio de requisitos.

10. **Herramientas de Validación Automática:**

   - **Definición:** Utilizar herramientas de software para realizar verificaciones automáticas de calidad.
   - **Ejemplo:** Emplear herramientas que verifiquen la consistencia gramatical y la trazabilidad entre requisitos.

**Conclusiones:**

El aseguramiento de la calidad en requisitos es un proceso continuo que implica la colaboración de diversos stakeholders y la aplicación de buenas prácticas de gestión de la calidad. La implementación efectiva de estos pasos garantiza que los requisitos sirvan como una base sólida para el desarrollo y la implementación exitosa del sistema.

### **Técnicas de revisión y validación**

Las técnicas de revisión y validación son procesos esenciales en el desarrollo de software y en la gestión de proyectos. Estas técnicas buscan asegurar la calidad y la precisión de los productos entregables a lo largo del ciclo de vida del desarrollo. A continuación, se describen algunas de las técnicas más comunes:

1. **Revisión por Pares:**

   - **Definición:** En esta técnica, dos o más personas revisan de manera conjunta un documento, código fuente u otro artefacto del proyecto.
   - **Propósito:** Identificar errores, inconsistencias y mejorar la calidad del trabajo a través del intercambio de ideas.
   - **Ejemplo:** Un programador revisa el código de otro colega para encontrar posibles errores o mejorar la legibilidad.

2. **Inspecciones Formales:**

   - **Definición:** Similar a la revisión por pares, pero con un enfoque más estructurado y formal. Implica el uso de roles definidos, agendas y métricas específicas.
   - **Propósito:** Detectar defectos de manera más exhaustiva y sistemática que las revisiones informales.
   - **Ejemplo:** Realizar inspecciones formales de documentos de requisitos para asegurar la consistencia y la cobertura completa.

3. **Caminatas a Través del Código (Code Walkthroughs):**

   - **Definición:** Una revisión detallada del código fuente, donde los desarrolladores explican y discuten cada parte del código.
   - **Propósito:** Mejorar la comprensión del código, detectar posibles errores y fomentar el conocimiento compartido del sistema.
   - **Ejemplo:** Un equipo realiza una caminata a través del código recién implementado para asegurarse de que todos comprendan su funcionalidad.

4. **Pruebas de Software:**

   - **Definición:** Ejecución de pruebas específicas para evaluar el comportamiento de un sistema o componente.
   - **Propósito:** Verificar que el software cumple con los requisitos y detectar posibles defectos.
   - **Ejemplo:** Pruebas de unidad, pruebas de integración, pruebas de sistema y pruebas de aceptación del usuario.

5. **Validación de Requisitos:**

   - **Definición:** Confirmación de que los requisitos del sistema cumplen con las expectativas y necesidades del cliente o stakeholders.
   - **Propósito:** Evitar malentendidos y garantizar que el sistema entregado satisfaga los objetivos comerciales.
   - **Ejemplo:** Revisión y aprobación formal de los requisitos por parte de los stakeholders.

6. **Modelado y Prototipado:**

   - **Definición:** Creación de modelos visuales o prototipos para representar aspectos clave del sistema antes de la implementación completa.
   - **Propósito:** Validar y refinar los conceptos del sistema antes de invertir recursos significativos en el desarrollo.
   - **Ejemplo:** Creación de maquetas de interfaz de usuario para obtener retroalimentación temprana de los usuarios.

7. **Validación Cruzada (Cross-Validation):**

   - **Definición:** En el contexto de modelos predictivos, se refiere a la evaluación de un modelo mediante la partición del conjunto de datos en subconjuntos de entrenamiento y prueba.
   - **Propósito:** Evaluar la capacidad de generalización del modelo y detectar problemas como sobreajuste.
   - **Ejemplo:** Validación cruzada k-fold en modelos de aprendizaje automático.

8. **Revisión Formal de Diseño:**

   - **Definición:** Un examen sistemático del diseño de un sistema, con roles y procedimientos definidos.
   - **Propósito:** Identificar posibles problemas de diseño antes de la implementación.
   - **Ejemplo:** Revisión formal de los diagramas de diseño de un sistema.

9. **Revisión de Documentación:**

   - **Definición:** Evaluación de documentos de proyecto para asegurar que estén completos, precisos y cumplan con los estándares definidos.
   - **Propósito:** Garantizar que la documentación sea una fuente precisa y útil de información sobre el proyecto.
   - **Ejemplo:** Revisión de manuales de usuario para verificar la claridad y la exhaustividad.

10. **Validación de Datos:**

    - **Definición:** Confirmación de que los datos utilizados en un proyecto son precisos, completos y relevantes.
    - **Propósito:** Evitar problemas derivados de datos incorrectos o inadecuados.
    - **Ejemplo:** Validación de datos de entrada para un modelo de aprendizaje automático.

**Importancia:**

La aplicación de estas técnicas es crucial para minimizar riesgos, mejorar la calidad del software y garantizar que el producto final cumpla con los requisitos y expectativas de los stakeholders. Además, promueven la colaboración efectiva, la detección temprana de problemas y la mejora continua del proceso de desarrollo.
