# **4. An√°lisis de Datos**

## **Conceptos B√°sicos de An√°lisis de Datos**

### **Uso de estad√≠sticas descriptivas**

Las estad√≠sticas descriptivas son una parte esencial del an√°lisis de datos, ya que proporcionan un resumen compacto y comprensible de las caracter√≠sticas principales de un conjunto de datos. Vamos a explorar algunas de las estad√≠sticas descriptivas m√°s comunes y c√≥mo se pueden utilizar con ejemplos y gr√°ficos en c√≥digo Python.

**Estad√≠sticas Descriptivas Comunes:**

1. **Media (Promedio):** La suma de todos los valores dividida por el n√∫mero de elementos.

2. **Mediana:** El valor medio de un conjunto de datos cuando est√° ordenado.

3. **Desviaci√≥n Est√°ndar:** Una medida de la dispersi√≥n de los valores en un conjunto de datos.

4. **Percentiles:** El valor por debajo del cual cae un porcentaje dado de observaciones en un conjunto de datos.

5. **Boxplot (Diagrama de Caja):** Proporciona una representaci√≥n visual de la mediana, los cuartiles y los valores at√≠picos.

**Ejemplo en Python:**

Vamos a generar un conjunto de datos ficticio y aplicar estas estad√≠sticas descriptivas:

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos aleatorios
np.random.seed(42)
data = np.random.normal(loc=0, scale=1, size=1000)

# Calcular estad√≠sticas descriptivas
mean_value = np.mean(data)
median_value = np.median(data)
std_dev = np.std(data)
percentile_25 = np.percentile(data, 25)
percentile_75 = np.percentile(data, 75)

# Graficar un boxplot
plt.boxplot(data, vert=False)
plt.title('Boxplot del Conjunto de Datos')
plt.show()

# Imprimir las estad√≠sticas
print(f'Media: {mean_value:.2f}')
print(f'Mediana: {median_value:.2f}')
print(f'Desviaci√≥n Est√°ndar: {std_dev:.2f}')
print(f'Percentil 25: {percentile_25:.2f}')
print(f'Percentil 75: {percentile_75:.2f}')
```

Este c√≥digo genera un conjunto de datos normalmente distribuido, calcula estad√≠sticas descriptivas y muestra un boxplot. Puedes adaptar este c√≥digo seg√∫n tus necesidades y aplicarlo a conjuntos de datos reales.

Las estad√≠sticas descriptivas te proporcionan una visi√≥n r√°pida de la tendencia central, la dispersi√≥n y la presencia de valores at√≠picos en tus datos. El boxplot es particularmente √∫til para visualizar estas caracter√≠sticas de manera concisa.

### **An√°lisis exploratorio de datos**

El An√°lisis Exploratorio de Datos (EDA, por sus siglas en ingl√©s) es una fase crucial en el proceso de an√°lisis de datos. Su objetivo es comprender la estructura del conjunto de datos, identificar patrones, detectar valores at√≠picos y formular hip√≥tesis para investigaciones adicionales. Vamos a explorar algunas t√©cnicas comunes de EDA con ejemplos en Python.

**T√©cnicas Comunes de EDA:**

1. **Histogramas y Gr√°ficos de Densidad:**
   - Proporcionan una visi√≥n de la distribuci√≥n de los datos.

2. **Diagramas de dispersi√≥n:**
   - Muestran la relaci√≥n entre dos variables.

3. **Matriz de correlaci√≥n:**
   - Describe las relaciones lineales entre pares de variables.

4. **Diagrama de caja por categor√≠as:**
   - Permite comparar la distribuci√≥n de una variable entre diferentes categor√≠as.

5. **Gr√°ficos de conteo:**
   - Muestra la frecuencia de categor√≠as en variables categ√≥ricas.

6. **Gr√°ficos de viol√≠n:**
   - Combinan informaci√≥n de boxplot y densidad.

**Ejemplo en Python:**

Vamos a utilizar un conjunto de datos de ejemplo y aplicar estas t√©cnicas:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar un conjunto de datos de ejemplo
tips = sns.load_dataset("tips")

# Histograma y Gr√°fico de Densidad
sns.histplot(tips['total_bill'], kde=True)
plt.title('Histograma y Gr√°fico de Densidad de Total Bill')
plt.show()

# Diagrama de Dispersi√≥n
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title('Diagrama de Dispersi√≥n entre Total Bill y Tip')
plt.show()

# Diagrama de Caja por Categor√≠as
sns.boxplot(x='day', y='total_bill', data=tips)
plt.title('Diagrama de Caja de Total Bill por D√≠a')
plt.show()

# Gr√°fico de Viol√≠n
sns.violinplot(x='day', y='total_bill', data=tips)
plt.title('Gr√°fico de Viol√≠n de Total Bill por D√≠a')
plt.show()

# Gr√°fico de Conteo
sns.countplot(x='day', data=tips)
plt.title('Gr√°fico de Conteo de D√≠as')
plt.show()
```

Estas visualizaciones te proporcionan informaci√≥n valiosa sobre la distribuci√≥n de las variables, relaciones entre ellas y patrones espec√≠ficos en tus datos. Puedes adaptar estas t√©cnicas seg√∫n la naturaleza de tu conjunto de datos y tus objetivos de an√°lisis.

### **Uso de algoritmos de clustering y clasificaci√≥n**

La identificaci√≥n de patrones y tendencias es una parte fundamental de la miner√≠a de datos. Puedes utilizar diversas t√©cnicas y algoritmos para descubrir patrones en tus datos. **Identificaci√≥n de Patrones con An√°lisis de Series Temporales**

Supongamos que tienes datos de series temporales y deseas identificar patrones y tendencias a lo largo del tiempo. Utilizaremos un conjunto de datos simple de temperatura para este ejemplo.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Crear un conjunto de datos de ejemplo de series temporales
fecha_inicio = '2022-01-01'
fecha_fin = '2022-12-31'
fechas = pd.date_range(start=fecha_inicio, end=fecha_fin)
num_dias = len(fechas)

# Generar temperaturas aleatorias para cada d√≠a
temperaturas = np.random.randint(20, 71, size=num_dias)

data = {'Fecha': fechas, 'Temperatura': temperaturas}
df = pd.DataFrame(data)

# Visualizar la serie temporal
plt.figure(figsize=(12, 6))
plt.plot(df['Fecha'], df['Temperatura'], marker='o')
plt.title('Temperatura a lo largo del tiempo')
plt.xlabel('Fecha')
plt.ylabel('Temperatura')
plt.grid(True)
plt.show()
```

Esta gr√°fica te dar√° una idea visual de c√≥mo cambia la temperatura con el tiempo. Puedes identificar patrones estacionales, tendencias ascendentes o descendentes, etc.

**Identificaci√≥n de Patrones con Clustering**

Supongamos que tienes un conjunto de datos multidimensional y deseas identificar grupos de patrones similares. Utilizaremos un conjunto de datos de iris y aplicaremos el algoritmo de K-Means para identificar patrones.

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

# Cargar el conjunto de datos de iris
iris = load_iris()
X = iris.data

# Normalizar los datos
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Aplicar K-Means con 3 clusters (puedes ajustar esto seg√∫n tus necesidades)
kmeans = KMeans(n_clusters=3, n_init=10, random_state=42)
y_kmeans = kmeans.fit_predict(X_scaled)

# Personalizar la gr√°fica
plt.figure(figsize=(8, 6))  # Ajustar tama√±o de la figura

# Scatter plot para los puntos de datos y los centros de los clusters
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=y_kmeans, cmap='viridis', label='Datos')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=200, c='red', marker='X', label='Centros de Cluster')

# Etiquetas y leyenda
plt.title('K-Means Clustering')
plt.xlabel('Caracter√≠stica 1 (Normalizada)')
plt.ylabel('Caracter√≠stica 2 (Normalizada)')
plt.legend()

# A√±adir cuadr√≠cula
plt.grid(True, linestyle='--', alpha=0.5)

# Mostrar la gr√°fica
plt.show()
```

Esta gr√°fica muestra c√≥mo K-Means agrupa los datos en clusters. Puedes ajustar el n√∫mero de clusters y explorar patrones en diferentes dimensiones.

Estos son solo ejemplos b√°sicos, y la elecci√≥n de t√©cnicas espec√≠ficas depender√° de tu conjunto de datos y tus objetivos. La miner√≠a de datos implica una exploraci√≥n profunda y la aplicaci√≥n de m√∫ltiples t√©cnicas para descubrir patrones valiosos en tus datos.
