1. **Medidas de Tendencia Central:**

**Media:**

La media es el promedio aritm茅tico de un conjunto de datos.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generar datos de ejemplo
datos = np.random.normal(loc=50, scale=10, size=1000)

# Calcular la media
media = np.mean(datos)
print(f"Media: {media}")

# Tama帽o de la figura y estilo
plt.figure(figsize=(10, 6))
sns.set(style='darkgrid')  # Utiliza el estilo de seaborn

# Histograma con bordes negros y transparencia
plt.hist(datos, bins=30, alpha=0.7, color='blue', edgecolor='black', label='Datos')

# L铆nea de media con ajustes
plt.axvline(media, color='red', linestyle='dashed', linewidth=2, label='Media', alpha=0.8)

# Etiquetas y leyenda
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.legend()

# T铆tulo y cuadr铆cula
plt.title('Histograma con Media')
plt.grid(True, linestyle='--', alpha=0.5)

# Informaci贸n estad铆stica adicional
std_dev = np.std(datos)
plt.text(10, 80, f'Media: {media:.2f}\nDesviaci贸n Est谩ndar: {std_dev:.2f}', fontsize=10)

# Mostrar la gr谩fica
plt.show()
```

### **An谩lisis de la distribuci贸n de datos**

El an谩lisis de la distribuci贸n de datos implica examinar la forma y caracter铆sticas fundamentales de c贸mo est谩n distribuidos los valores en un conjunto de datos. Aqu铆 hay una explicaci贸n detallada con ejemplos y gr谩ficos personalizados en c贸digo Python.

4. **QQ Plot:**

El QQ plot compara los cuantiles observados con los cuantiles esperados de una distribuci贸n te贸rica (como la normal). Es 煤til para evaluar la normalidad de los datos.

```python
import scipy.stats as stats
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Generar datos de ejemplo
datos = np.random.normal(loc=50, scale=10, size=1000)

# Crear un QQ plot
sm.qqplot(datos, line='45', fmt='b')  # Elimina el argumento color
plt.title('QQ Plot')
plt.show()
```

5. **Distribuciones Espec铆ficas:**

Ajustar datos a una distribuci贸n espec铆fica y comparar visualmente.

```python
# Ajustar a una distribuci贸n normal
sns.histplot(datos, kde=True, color='blue', stat='density', label='Datos')
plt.title('Ajuste a Distribuci贸n Normal')
plt.xlabel('Valor')
plt.ylabel('Densidad')
plt.legend()

# Superponer la densidad de una distribuci贸n normal
x = np.linspace(np.min(datos), np.max(datos), 100)
y = stats.norm.pdf(x, np.mean(datos), np.std(datos))
plt.plot(x, y, color='red', label='Distribuci贸n Normal')
plt.legend()
plt.show()
```

### **Visualizaci贸n de relaciones y patrones en los datos**

La visualizaci贸n de relaciones y patrones en los datos es esencial para comprender la estructura y las interacciones dentro de un conjunto de datos. Vamos a explorar diversas t茅cnicas utilizando las bibliotecas de Python, como Matplotlib y Seaborn.

1. **Diagrama de Dispersi贸n con Ajuste de L铆nea:**

Un diagrama de dispersi贸n con una l铆nea de ajuste puede revelar patrones en la relaci贸n entre dos variables.

```python
import numpy as np
import matplotlib.pyplot as plt

# Crear datos de ejemplo
x = np.linspace(0, 10, 100)
y = 2 * x + np.random.normal(0, 1, 100)

# Crear un diagrama de dispersi贸n con ajuste de l铆nea
plt.scatter(x, y, color='blue', label='Datos')
plt.plot(x, 2 * x, color='red', linestyle='--', label='Ajuste de L铆nea')
plt.title('Diagrama de Dispersi贸n con Ajuste de L铆nea')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.legend()
plt.show()
```

2. **Mapa de Calor (Heatmap):**

Un mapa de calor es 煤til para visualizar la relaci贸n entre dos variables categ贸ricas.

```python
import seaborn as sns

# Crear datos de ejemplo
data = np.random.rand(10, 12)

# Crear un mapa de calor
sns.heatmap(data, cmap='Blues', annot=True, fmt=".2f", linewidths=.5)
plt.title('Mapa de Calor')
plt.show()
```

3. **Pairplot:**

Un pairplot muestra relaciones entre m煤ltiples variables en un conjunto de datos.

```python
import seaborn as sns
import pandas as pd

# Crear datos de ejemplo
data = sns.load_dataset('iris')

# Crear un pairplot
sns.pairplot(data, hue='species', markers=['o', 's', 'D'])
plt.suptitle('Pairplot de Datos Iris', y=1.02)
plt.show()
```

Un gr谩fico de viol铆n puede mostrar la distribuci贸n de una variable en diferentes categor铆as.


6. **Diagrama de Correlaci贸n:**

Un diagrama de correlaci贸n es 煤til para visualizar la fuerza y direcci贸n de las relaciones entre variables.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el conjunto de datos de iris
data = sns.load_dataset('iris')

# Codificar la variable categ贸rica 'species'
data_encoded = pd.get_dummies(data, columns=['species'])

# Calcular la matriz de correlaci贸n
correlation_matrix = data_encoded.corr()

# Crear un mapa de calor
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Diagrama de Correlaci贸n')
plt.show()
```
