# **3. Estadísticas Descriptivas y Visualización de Datos**

## **Estadísticas Descriptivas**

Las estadísticas descriptivas son un conjunto de técnicas que resumen y describen las características principales de un conjunto de datos. Estas medidas proporcionan información sobre la distribución, dispersión y forma de los datos. A continuación, se presenta una explicación detallada con ejemplos y gráficos personalizados en código Python.

1. **Medidas de Tendencia Central:**

**Media:**

La media es el promedio aritmético de un conjunto de datos.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generar datos de ejemplo
datos = np.random.normal(loc=50, scale=10, size=1000)

# Calcular la media
media = np.mean(datos)
print(f"Media: {media}")

# Tamaño de la figura y estilo
plt.figure(figsize=(10, 6))
sns.set(style='darkgrid')  # Utiliza el estilo de seaborn

# Histograma con bordes negros y transparencia
plt.hist(datos, bins=30, alpha=0.7, color='blue', edgecolor='black', label='Datos')

# Línea de media con ajustes
plt.axvline(media, color='red', linestyle='dashed', linewidth=2, label='Media', alpha=0.8)

# Etiquetas y leyenda
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.legend()

# Título y cuadrícula
plt.title('Histograma con Media')
plt.grid(True, linestyle='--', alpha=0.5)

# Información estadística adicional
std_dev = np.std(datos)
plt.text(10, 80, f'Media: {media:.2f}\nDesviación Estándar: {std_dev:.2f}', fontsize=10)

# Mostrar la gráfica
plt.show()
```

**Mediana:**

La mediana es el valor que divide el conjunto de datos en dos partes iguales.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generar datos de ejemplo
datos = np.random.normal(loc=50, scale=10, size=1000)

# Calcular la mediana
mediana = np.median(datos)
print(f"Mediana: {mediana}")

# Tamaño de la figura y estilo
plt.figure(figsize=(10, 6))
sns.set(style='darkgrid')  # Utiliza el estilo de seaborn

# Histograma con bordes negros y transparencia
plt.hist(datos, bins=30, alpha=0.7, color='green', edgecolor='black', label='Datos')

# Línea de mediana con ajustes
plt.axvline(mediana, color='purple', linestyle='dashed', linewidth=2, label='Mediana')

# Etiquetas y leyenda
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.legend()

# Título y cuadrícula
plt.title('Histograma con Mediana')
plt.grid(True, linestyle='--', alpha=0.5)

# Información estadística adicional
std_dev = np.std(datos)
plt.text(10, 80, f'Mediana: {mediana:.2f}\nDesviación Estándar: {std_dev:.2f}', fontsize=10)

# Mostrar la gráfica
plt.show()
```

2. **Medidas de Dispersión:**

**Desviación Estándar:**

La desviación estándar mide la dispersión de los datos con respecto a la media.

```python
# Calcular la desviación estándar
desviacion_estandar = np.std(datos)
print(f"Desviación Estándar: {desviacion_estandar}")
```

**Rango:**

El rango es la diferencia entre el valor máximo y el valor mínimo en un conjunto de datos.

```python
# Calcular el rango
rango = np.max(datos) - np.min(datos)
print(f"Rango: {rango}")
```

3. **Visualización con Seaborn:**

Seaborn es una biblioteca de visualización de datos que complementa a Matplotlib.

```python
import seaborn as sns

# Visualizar la distribución y la densidad
sns.histplot(datos, kde=True, color='skyblue')
plt.title('Distribución de Datos con Densidad')
plt.show()

# Visualizar un diagrama de caja (boxplot)
sns.boxplot(x=datos, color='lightcoral')
plt.title('Diagrama de Caja')
plt.show()
```

Estas son solo algunas de las medidas de estadísticas descriptivas y visualizaciones disponibles en Python. Utilizando bibliotecas como NumPy, Matplotlib y Seaborn, puedes realizar un análisis más profundo y visualizar las características esenciales de tus datos.

### **Medidas de tendencia central y dispersión**

Las medidas de tendencia central y dispersión son herramientas estadísticas fundamentales que proporcionan información clave sobre la distribución de los datos. A continuación, se detallan estas medidas con ejemplos y gráficos personalizados en código Python.

1. **Medidas de Tendencia Central:**

**Media:**

La media es el promedio aritmético de un conjunto de datos. Se calcula sumando todos los valores y dividiendo por el número total de observaciones.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos de ejemplo
datos = np.random.normal(loc=50, scale=10, size=1000)

# Calcular la media
media = np.mean(datos)
print(f"Media: {media}")

# Personalizar la gráfica
plt.figure(figsize=(10, 6))  # Ajustar tamaño de la figura
plt.hist(datos, bins=30, alpha=0.7, color='blue', edgecolor='black', label='Datos')  # Añadir bordes negros

# Línea de la media con ajustes
plt.axvline(media, color='red', linestyle='dashed', linewidth=2, label='Media')

# Etiquetas y leyenda
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.title('Histograma con Media')
plt.legend()

# Cuadrícula
plt.grid(True, linestyle='--', alpha=0.5)

# Añadir información estadística adicional
std_dev = np.std(datos)
plt.text(10, 80, f'Media: {media:.2f}\nDesviación Estándar: {std_dev:.2f}', fontsize=10)

# Mostrar la gráfica
plt.show()
```

**Mediana:**

La mediana es el valor que divide el conjunto de datos en dos partes iguales. Se encuentra ordenando los datos y seleccionando el valor en el centro.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos de ejemplo
datos = np.random.normal(loc=50, scale=10, size=1000)

# Calcular la mediana
mediana = np.median(datos)
print(f"Mediana: {mediana}")

# Personalizar la gráfica de la mediana
plt.figure(figsize=(10, 6))  # Ajustar tamaño de la figura
plt.hist(datos, bins=30, alpha=0.7, color='green', edgecolor='black', label='Datos')  # Añadir bordes negros

# Línea de la mediana con ajustes
plt.axvline(mediana, color='purple', linestyle='dashed', linewidth=2, label='Mediana')

# Etiquetas y leyenda
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.title('Histograma con Mediana')
plt.legend()

# Cuadrícula
plt.grid(True, linestyle='--', alpha=0.5)

# Mostrar la gráfica
plt.show()
```

### **Análisis de la distribución de datos**

El análisis de la distribución de datos implica examinar la forma y características fundamentales de cómo están distribuidos los valores en un conjunto de datos. Aquí hay una explicación detallada con ejemplos y gráficos personalizados en código Python.

1. **Histograma:**

El histograma es una representación visual de la distribución de frecuencias de un conjunto de datos. Muestra la frecuencia o probabilidad de que un rango particular de valores ocurra.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos de ejemplo
datos = np.random.normal(loc=50, scale=10, size=1000)

# Crear un histograma
plt.hist(datos, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Histograma de Datos')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.show()
```

2. **Densidad de Probabilidad:**

La densidad de probabilidad es una estimación suave de la distribución de probabilidad subyacente de los datos. Puede visualizarse mediante un gráfico de densidad.

```python
import seaborn as sns

# Visualizar la distribución y la densidad
sns.histplot(datos, kde=True, color='green')
plt.title('Histograma con Densidad')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.show()
```

3. **Diagrama de Caja (Boxplot):**

El diagrama de caja proporciona información sobre la dispersión y simetría de los datos, así como sobre la presencia de valores atípicos.

```python
# Visualizar un diagrama de caja
sns.boxplot(x=datos, color='lightcoral')
plt.title('Diagrama de Caja')
plt.xlabel('Datos')
plt.show()
```

4. **QQ Plot:**

El QQ plot compara los cuantiles observados con los cuantiles esperados de una distribución teórica (como la normal). Es útil para evaluar la normalidad de los datos.

```python
import scipy.stats as stats
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Generar datos de ejemplo
datos = np.random.normal(loc=50, scale=10, size=1000)

# Crear un QQ plot
sm.qqplot(datos, line='45', fmt='b')  # Elimina el argumento color
plt.title('QQ Plot')
plt.show()
```

5. **Distribuciones Específicas:**

Ajustar datos a una distribución específica y comparar visualmente.

```python
# Ajustar a una distribución normal
sns.histplot(datos, kde=True, color='blue', stat='density', label='Datos')
plt.title('Ajuste a Distribución Normal')
plt.xlabel('Valor')
plt.ylabel('Densidad')
plt.legend()

# Superponer la densidad de una distribución normal
x = np.linspace(np.min(datos), np.max(datos), 100)
y = stats.norm.pdf(x, np.mean(datos), np.std(datos))
plt.plot(x, y, color='red', label='Distribución Normal')
plt.legend()
plt.show()
```

6. **Medidas Resumen:**

Calcular medidas resumen como la media, mediana, desviación estándar, etc.

```python
# Calcular medidas resumen
media = np.mean(datos)
mediana = np.median(datos)
desviacion_estandar = np.std(datos)

print(f"Media: {media}")
print(f"Mediana: {mediana}")
print(f"Desviación Estándar: {desviacion_estandar}")
```

Este análisis proporciona información crucial sobre la forma y naturaleza de los datos, lo que ayuda en la toma de decisiones y en la elección de técnicas de análisis estadístico adecuadas. Es importante adaptar estas técnicas según el tipo de datos y los objetivos específicos del análisis.

### **Visualización de Datos con Matplotlib y Seaborn**

### **Creación de gráficos básicos**

Crear gráficos es una parte esencial del análisis de datos y visualización. Python ofrece varias bibliotecas poderosas para la creación de gráficos, como Matplotlib y Seaborn. Aquí tienes una explicación detallada con ejemplos y gráficos personalizados en código Python.

1. **Gráfico de Línea:**

Un gráfico de línea es útil para mostrar la relación entre dos variables a lo largo de un eje.

```python
import numpy as np
import matplotlib.pyplot as plt

# Crear datos de ejemplo
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Crear un gráfico de línea
plt.plot(x, y, label='Función Seno', color='blue', linestyle='--', marker='o')
plt.title('Gráfico de Línea')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.legend()
plt.show()
```

2. **Gráfico de Barras:**

Un gráfico de barras es útil para comparar diferentes categorías.

```python
# Crear datos de ejemplo
categorias = ['A', 'B', 'C', 'D']
valores = [15, 24, 30, 22]

# Crear un gráfico de barras
plt.bar(categorias, valores, color='green')
plt.title('Gráfico de Barras')
plt.xlabel('Categorías')
plt.ylabel('Valores')
plt.show()
```

3. **Gráfico de Dispersión:**

Un gráfico de dispersión muestra la relación entre dos variables y es útil para identificar patrones.

```python
# Crear datos de ejemplo
x = np.random.rand(50)
y = 2 * x + np.random.randn(50)

# Crear un gráfico de dispersión
plt.scatter(x, y, color='red', marker='o')
plt.title('Gráfico de Dispersión')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.show()
```

4. **Histograma:**

Un histograma muestra la distribución de un conjunto de datos.

```python
# Crear datos de ejemplo
datos = np.random.randn(1000)

# Crear un histograma
plt.hist(datos, bins=30, color='purple', edgecolor='black')
plt.title('Histograma')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.show()
```

5. **Diagrama de Caja (Boxplot):**

El diagrama de caja muestra la distribución y la presencia de valores atípicos.

```python
# Crear datos de ejemplo
datos = np.random.normal(0, 1, 100)

# Crear un diagrama de caja
plt.boxplot(datos, vert=False)
plt.title('Diagrama de Caja')
plt.xlabel('Valores')
plt.show()
```

6. **Gráfico de Pastel:**

Un gráfico de pastel representa proporciones de un conjunto.

```python
# Crear datos de ejemplo
tamaños = [20, 30, 40, 10]
etiquetas = ['A', 'B', 'C', 'D']

# Crear un gráfico de pastel
plt.pie(tamaños, labels=etiquetas, autopct='%1.1f%%', colors=['gold', 'lightcoral', 'lightskyblue', 'lightgreen'])
plt.title('Gráfico de Pastel')
plt.show()
```

Estos son solo algunos ejemplos básicos. Puedes personalizar aún más los gráficos según tus necesidades utilizando las numerosas opciones de configuración disponibles en las bibliotecas de visualización de Python.

### **Visualización de relaciones y patrones en los datos**

La visualización de relaciones y patrones en los datos es esencial para comprender la estructura y las interacciones dentro de un conjunto de datos. Vamos a explorar diversas técnicas utilizando las bibliotecas de Python, como Matplotlib y Seaborn.

1. **Diagrama de Dispersión con Ajuste de Línea:**

Un diagrama de dispersión con una línea de ajuste puede revelar patrones en la relación entre dos variables.

```python
import numpy as np
import matplotlib.pyplot as plt

# Crear datos de ejemplo
x = np.linspace(0, 10, 100)
y = 2 * x + np.random.normal(0, 1, 100)

# Crear un diagrama de dispersión con ajuste de línea
plt.scatter(x, y, color='blue', label='Datos')
plt.plot(x, 2 * x, color='red', linestyle='--', label='Ajuste de Línea')
plt.title('Diagrama de Dispersión con Ajuste de Línea')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.legend()
plt.show()
```

2. **Mapa de Calor (Heatmap):**

Un mapa de calor es útil para visualizar la relación entre dos variables categóricas.

```python
import seaborn as sns

# Crear datos de ejemplo
data = np.random.rand(10, 12)

# Crear un mapa de calor
sns.heatmap(data, cmap='Blues', annot=True, fmt=".2f", linewidths=.5)
plt.title('Mapa de Calor')
plt.show()
```

3. **Pairplot:**

Un pairplot muestra relaciones entre múltiples variables en un conjunto de datos.

```python
import seaborn as sns
import pandas as pd

# Crear datos de ejemplo
data = sns.load_dataset('iris')

# Crear un pairplot
sns.pairplot(data, hue='species', markers=['o', 's', 'D'])
plt.suptitle('Pairplot de Datos Iris', y=1.02)
plt.show()
```

4. **Gráfico de Violín:**

Un gráfico de violín puede mostrar la distribución de una variable en diferentes categorías.

```python
# Crear un gráfico de violín
sns.violinplot(x='species', y='sepal_length', data=data, palette='Set2')
plt.title('Gráfico de Violín')
plt.xlabel('Especies')
plt.ylabel('Longitud del Sépalo')
plt.show()
```

5. **Diagrama de Caja y Bigotes:**

Un diagrama de caja y bigotes proporciona información sobre la distribución y la presencia de valores atípicos.


```python
# Crear un diagrama de caja y bigotes
sns.boxplot(x='species', y='sepal_length', data=data, palette='Pastel1')
plt.title('Diagrama de Caja y Bigotes')
plt.xlabel('Especies')
plt.ylabel('Longitud del Sépalo')
plt.show()
```


6. **Diagrama de Correlación:**

Un diagrama de correlación es útil para visualizar la fuerza y dirección de las relaciones entre variables.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el conjunto de datos de iris
data = sns.load_dataset('iris')

# Codificar la variable categórica 'species'
data_encoded = pd.get_dummies(data, columns=['species'])

# Calcular la matriz de correlación
correlation_matrix = data_encoded.corr()

# Crear un mapa de calor
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Diagrama de Correlación')
plt.show()
```

Estos son solo algunos ejemplos, y la elección de la técnica dependerá de la naturaleza de tus datos y de lo que estés tratando de explorar. La visualización efectiva de datos es esencial para obtener información valiosa y tomar decisiones informadas.
