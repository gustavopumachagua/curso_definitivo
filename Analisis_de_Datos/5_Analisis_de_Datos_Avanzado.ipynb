## **An谩lisis Estad铆stico Inferencial**

El an谩lisis estad铆stico inferencial es una rama de la estad铆stica que se centra en realizar inferencias o conclusiones sobre una poblaci贸n en base a una muestra de datos recopilados de esa poblaci贸n. En lugar de describir simplemente los datos, como se hace en el an谩lisis estad铆stico descriptivo, el an谩lisis inferencial busca generalizar las conclusiones a toda la poblaci贸n.

Las t茅cnicas de an谩lisis estad铆stico inferencial se basan en la teor铆a de la probabilidad y permiten realizar afirmaciones o inferencias sobre las caracter铆sticas de una poblaci贸n, como medias, proporciones, diferencias entre grupos, entre otros.

A continuaci贸n, proporcionar茅 una explicaci贸n general y un ejemplo de c贸mo realizar an谩lisis estad铆stico inferencial con Python utilizando una prueba de hip贸tesis t de dos muestras. Imaginemos que queremos comparar las alturas promedio de hombres y mujeres en una poblaci贸n.

**Prueba de Hip贸tesis (T de dos muestras)**

1. **Hip贸tesis nula (H0):** No hay diferencia significativa entre las alturas promedio de hombres y mujeres.
2. **Hip贸tesis alternativa (H1):** Hay una diferencia significativa entre las alturas promedio de hombres y mujeres.

````python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Generar datos de altura para hombres y mujeres
np.random.seed(42)
altura_hombres = np.random.normal(loc=175, scale=7, size=100)
altura_mujeres = np.random.normal(loc=162, scale=6, size=100)

# Realizar una prueba t de dos muestras
t_stat, p_value = stats.ttest_ind(altura_hombres, altura_mujeres)

# Imprimir resultados
print(f"Estad铆stico t: {t_stat}")
print(f"Valor p: {p_value}")

# Visualizar los datos
plt.hist(altura_hombres, alpha=0.5, label='Hombres', color='blue')
plt.hist(altura_mujeres, alpha=0.5, label='Mujeres', color='pink')
plt.title('Distribuci贸n de Alturas por G茅nero')
plt.xlabel('Altura')
plt.ylabel('Frecuencia')
plt.legend()
plt.show()
````

En este ejemplo:

- `altura_hombres` y `altura_mujeres` son conjuntos de datos generados aleatoriamente que representan alturas de hombres y mujeres.
- `stats.ttest_ind` realiza una prueba t de dos muestras para evaluar si hay una diferencia significativa entre las alturas promedio de ambos grupos.
- El estad铆stico t y el valor p se utilizan para tomar decisiones sobre si rechazar o no la hip贸tesis nula.

Es importante destacar que este es solo un ejemplo simple. En situaciones reales, el an谩lisis estad铆stico inferencial puede ser m谩s complejo, y las elecciones de prueba pueden depender de la naturaleza de los datos y las preguntas de investigaci贸n.

Recuerda que la interpretaci贸n correcta y apropiada de los resultados de las pruebas estad铆sticas es esencial para realizar inferencias v谩lidas sobre la poblaci贸n. Adem谩s, se deben considerar otras consideraciones, como el tama帽o de la muestra y la distribuci贸n de los datos.

### **Pruebas de hip贸tesis**

Las pruebas de hip贸tesis son procedimientos estad铆sticos que nos permiten tomar decisiones sobre afirmaciones o suposiciones (hip贸tesis) que hacemos acerca de una poblaci贸n, bas谩ndonos en la informaci贸n recopilada de una muestra. Estas pruebas ayudan a evaluar si hay evidencia suficiente para rechazar o no una afirmaci贸n sobre la poblaci贸n.

Aqu铆, proporcionar茅 una explicaci贸n general y un ejemplo utilizando una prueba t de una muestra en Python.

**Pasos Comunes en una Prueba de Hip贸tesis:**

1. **Formulaci贸n de Hip贸tesis:**

   - **Hip贸tesis nula (H0):** Afirmaci贸n inicial que asumimos como verdadera.
   - **Hip贸tesis alternativa (H1):** Afirmaci贸n que buscamos respaldar con evidencia.

2. **Selecci贸n de Nivel de Significancia (伪):**

   - Es la probabilidad de cometer un error de tipo I (rechazar H0 cuando es verdadera). Com煤nmente se elige 伪 = 0.05.

3. **Recopilaci贸n de Datos y C谩lculos:**

   - Recopilar datos de una muestra.
   - Calcular estad铆sticos seg煤n el tipo de prueba.

4. **Toma de Decisiones:**

   - Comparar el valor p (probabilidad de obtener resultados tan extremos como los observados, asumiendo que H0 es verdadera) con 伪.
   - Si el valor p < 伪, rechazamos H0 a favor de H1; de lo contrario, no hay suficiente evidencia para rechazar H0.

**Ejemplo: Prueba t de una Muestra**

Supongamos que queremos probar si la altura promedio de una muestra de personas es significativamente diferente de 170 cm. Aqu铆 hay un ejemplo en Python:

````python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Generar datos de altura de una muestra
np.random.seed(42)
altura_muestra = np.random.normal(loc=172, scale=5, size=30)

# Realizar una prueba t de una muestra
t_stat, p_value = stats.ttest_1samp(altura_muestra, popmean=170)

# Imprimir resultados
print(f"Estad铆stico t: {t_stat}")
print(f"Valor p: {p_value}")

# Visualizar los datos
plt.hist(altura_muestra, bins=15, alpha=0.5, color='blue')
plt.axvline(x=170, color='red', linestyle='--', label='Altura te贸rica')
plt.title('Distribuci贸n de Alturas en la Muestra')
plt.xlabel('Altura')
plt.ylabel('Frecuencia')
plt.legend()
plt.show()
````

En este ejemplo:

- `altura_muestra` es un conjunto de datos de altura generado aleatoriamente.
- `stats.ttest_1samp` realiza una prueba t de una muestra para evaluar si la altura promedio de la muestra es significativamente diferente de 170 cm.
- El estad铆stico t y el valor p se utilizan para tomar decisiones sobre si rechazar o no la hip贸tesis nula.

Es importante interpretar cuidadosamente los resultados y considerar el contexto del problema al realizar pruebas de hip贸tesis. Adem谩s, ten en cuenta que este ejemplo es simplificado y que existen muchas otras pruebas de hip贸tesis para diferentes situaciones.

### **Regresi贸n lineal y correlaci贸n**

La regresi贸n lineal y la correlaci贸n son t茅cnicas estad铆sticas utilizadas para estudiar la relaci贸n entre dos variables. La regresi贸n lineal se centra en predecir o modelar el valor de una variable dependiente bas谩ndose en los valores de una o m谩s variables independientes, mientras que la correlaci贸n mide la fuerza y direcci贸n de la relaci贸n lineal entre dos variables.

**Regresi贸n Lineal:**

La regresi贸n lineal es un modelo estad铆stico que asume una relaci贸n lineal entre la variable dependiente  $(Y)$ y las variables independientes $(X_1, X_2, ..., X_n)$. La ecuaci贸n general de una regresi贸n lineal simple es:

$Y = \beta_0 + \beta_1 X + \epsilon$

- $(Y)$ es la variable dependiente.
- $(X)$ es la variable independiente.
- $(\beta_0)$ es la ordenada al origen (intercepto).
- $(\beta_1)$ es la pendiente de la l铆nea.
- $(\epsilon)$ es el t茅rmino de error.

**Correlaci贸n:**

La correlaci贸n cuantifica la fuerza y direcci贸n de una relaci贸n lineal entre dos variables. El coeficiente de correlaci贸n de Pearson $(r)$ es una medida com煤n de correlaci贸n, con valores en el rango $[-1, 1]$.

- $(r = 1)$ implica una correlaci贸n positiva perfecta.
- $(r = -1)$ implica una correlaci贸n negativa perfecta.
- $(r = 0)$ implica falta de correlaci贸n lineal.

**Ejemplo en Python:**

Vamos a generar datos ficticios y realizar una regresi贸n lineal y calcular el coeficiente de correlaci贸n usando Python:

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

# Generar datos ficticios
np.random.seed(42)
X = np.random.rand(100) * 10
Y = 2 * X + 1 + np.random.randn(100) * 2  # Ecuaci贸n con error

# Realizar regresi贸n lineal
slope, intercept, r_value, p_value, std_err = linregress(X, Y)

# Calcular valores predichos
Y_pred = slope * X + intercept

# Calcular coeficiente de correlaci贸n
correlation_coefficient = np.corrcoef(X, Y)[0, 1]

# Graficar los datos y la regresi贸n lineal
plt.scatter(X, Y, label='Datos')
plt.plot(X, Y_pred, color='red', label=f'Regresi贸n Lineal (r = {correlation_coefficient:.2f})')
plt.title('Regresi贸n Lineal y Correlaci贸n')
plt.xlabel('Variable Independiente (X)')
plt.ylabel('Variable Dependiente (Y)')
plt.legend()
plt.show()
````

En este ejemplo:

- Generamos datos ficticios con una relaci贸n lineal y un poco de ruido.
- Calculamos la regresi贸n lineal usando `linregress` de `scipy.stats`.
- Graficamos los datos junto con la l铆nea de regresi贸n.

Este c贸digo proporcionar谩 una visualizaci贸n de una regresi贸n lineal y mostrar谩 el coeficiente de correlaci贸n $(r)$ en el gr谩fico. La interpretaci贸n del coeficiente de correlaci贸n es clave para comprender la fuerza y direcci贸n de la relaci贸n lineal.

## **An谩lisis de Series Temporales**

### **Conceptos b谩sicos de series temporales**

Las series temporales son conjuntos de datos ordenados temporalmente, donde las observaciones se recopilan, registran o registran en intervalos regulares. Estos datos son fundamentales en 谩reas como finanzas, meteorolog铆a, econom铆a y muchas otras disciplinas. A continuaci贸n, se presentan algunos conceptos clave:

1. **Tendencia:**

   - *Definici贸n:* La direcci贸n general en la que se mueve una serie temporal a largo plazo.
   - *Ejemplo:* Aumento constante de las ventas de una empresa a lo largo de varios a帽os.

2. **Estacionalidad:**

   - *Definici贸n:* Patrones recurrentes o ciclos que se repiten en intervalos regulares.
   - *Ejemplo:* Incremento en las ventas durante las temporadas de vacaciones cada a帽o.

3. **Ciclo:**

   - *Definici贸n:* Variaciones recurrentes que no est谩n necesariamente relacionadas con la duraci贸n de un a帽o.
   - *Ejemplo:* Ciclos econ贸micos, como recesiones y expansiones.

4. **Ruido (o Residuo):**

   - *Definici贸n:* Componente irregular o aleatorio que no sigue ning煤n patr贸n discernible.
   - *Ejemplo:* Fluctuaciones impredecibles en las ventas diarias debido a factores inesperados.

5. **Autocorrelaci贸n:**

   - *Definici贸n:* Grado de similitud entre una observaci贸n y observaciones previas en la serie temporal.
   - *Ejemplo:* La temperatura de hoy puede estar correlacionada con la temperatura de ayer.

**Ejemplo en Python:**

Vamos a generar una serie temporal simulada que incluya tendencia, estacionalidad y ruido, y luego visualizaremos estos conceptos.

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Generar datos de una serie temporal simulada
np.random.seed(42)
timesteps = 100
t = np.arange(timesteps)
trend = 0.5 * t  # Tendencia lineal
seasonal = 10 * np.sin(2 * np.pi * t / 12)  # Estacionalidad mensual
noise = np.random.normal(0, 5, timesteps)  # Ruido

# Combinar componentes para formar la serie temporal
time_series = trend + seasonal + noise

# Crear DataFrame con la serie temporal
df = pd.DataFrame({'Fecha': pd.date_range('2022-01-01', periods=timesteps), 'Valor': time_series})

# Visualizar la serie temporal
plt.figure(figsize=(10, 6))
plt.plot(df['Fecha'], df['Valor'], label='Serie Temporal Simulada')
plt.title('Serie Temporal con Tendencia, Estacionalidad y Ruido')
plt.xlabel('Fecha')
plt.ylabel('Valor')
plt.legend()
plt.show()
````

En este ejemplo:

- `trend` representa una tendencia lineal creciente con el tiempo.
- `seasonal` simula una componente estacional mensual.
- `noise` introduce un componente de ruido aleatorio.
- La serie temporal resultante (`time_series`) es la combinaci贸n de estos componentes.

El gr谩fico resultante visualiza c贸mo se ven estos conceptos en una serie temporal simulada.

### **An谩lisis y pron贸stico**

El an谩lisis y pron贸stico en series temporales implican comprender patrones pasados, identificar tendencias y estacionalidades, y predecir valores futuros. Vamos a explorar estos conceptos utilizando Python con la biblioteca `statsmodels` para el an谩lisis y `prophet` para el pron贸stico.

**An谩lisis de la Serie Temporal:**

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm

# Generar datos de una serie temporal simulada
np.random.seed(42)
timesteps = 100
t = np.arange(timesteps)
trend = 0.5 * t  # Tendencia lineal
seasonal = 10 * np.sin(2 * np.pi * t / 12)  # Estacionalidad mensual
noise = np.random.normal(0, 5, timesteps)  # Ruido

# Combinar componentes para formar la serie temporal
time_series = trend + seasonal + noise

# Crear DataFrame con la serie temporal
df = pd.DataFrame({'Fecha': pd.date_range('2022-01-01', periods=timesteps), 'Valor': time_series})

# Descomposici贸n de la serie temporal
result = sm.tsa.seasonal_decompose(time_series, model='additive', period=12)

# Visualizar componentes
plt.figure(figsize=(12, 8))
plt.subplot(4, 1, 1)
plt.plot(df['Fecha'], df['Valor'], label='Serie Temporal Original')
plt.legend()

plt.subplot(4, 1, 2)
plt.plot(df['Fecha'], result.trend, label='Tendencia')
plt.legend()

plt.subplot(4, 1, 3)
plt.plot(df['Fecha'], result.seasonal, label='Estacionalidad')
plt.legend()

plt.subplot(4, 1, 4)
plt.plot(df['Fecha'], result.resid, label='Residuo (Ruido)')
plt.legend()

plt.tight_layout()
plt.show()
````

En este ejemplo:

- Generamos una serie temporal simulada con tendencia, estacionalidad y ruido.
- Utilizamos la descomposici贸n estacional aditiva de `statsmodels` para separar la serie en tendencia, estacionalidad y residuo.

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm

# Generar datos de una serie temporal simulada
np.random.seed(42)
timesteps = 100
t = np.arange(timesteps)
trend = 0.5 * t  # Tendencia lineal
seasonal = 10 * np.sin(2 * np.pi * t / 12)  # Estacionalidad mensual
noise = np.random.normal(0, 5, timesteps)  # Ruido

# Combinar componentes para formar la serie temporal
time_series = trend + seasonal + noise

# Crear DataFrame con la serie temporal
df = pd.DataFrame({'Fecha': pd.date_range('2022-01-01', periods=timesteps), 'Valor': time_series})

# Ajustar un modelo SARIMA (Seasonal AutoRegressive Integrated Moving Average)
order = (1, 1, 1)  # (p, d, q)
seasonal_order = (0, 0, 0, 0)  # (P, D, Q, S)
model = sm.tsa.SARIMAX(df['Valor'], order=order, seasonal_order=seasonal_order, enforce_stationarity=False, enforce_invertibility=False)
results = model.fit()

# Generar pron贸sticos
forecast_steps = 12
forecast = results.get_forecast(steps=forecast_steps)

# Visualizar el pron贸stico
plt.figure(figsize=(12, 6))
plt.plot(df['Fecha'], df['Valor'], label='Observado')
forecast_dates = pd.date_range('2022-01-01', periods=timesteps + forecast_steps)
plt.plot(forecast_dates[-forecast_steps:], forecast.predicted_mean, label='Pron贸stico')
plt.title('Pron贸stico con SARIMA')
plt.legend()
plt.show()
````

En este ejemplo:

- Utilizamos la biblioteca `prophet` para pronosticar los valores futuros de la serie temporal.
- Creamos un modelo Prophet, ajustamos el modelo con datos hist贸ricos y generamos un DataFrame con fechas futuras.
- Realizamos el pron贸stico y visualizamos los resultados.

Este c贸digo proporciona una introducci贸n b谩sica al an谩lisis y pron贸stico de series temporales. Es importante tener en cuenta que estos son ejemplos simplificados, y el an谩lisis y pron贸stico de series temporales en entornos del mundo real pueden ser m谩s complejos y requerir t茅cnicas m谩s avanzadas.
