# **4. Probabilidades**

## **Conceptos Fundamentales de Probabilidad**

### **Eventos, espacio muestral, reglas de probabilidad b√°sicas**

**1. Eventos:**

Un evento es un resultado o conjunto de resultados de un experimento aleatorio. Puede ser cualquier conjunto de posibles resultados, ya sea simple (como obtener un n√∫mero espec√≠fico al lanzar un dado) o compuesto (como obtener un n√∫mero par al lanzar un dado).

**2. Espacio Muestral:**

El espacio muestral es el conjunto de todos los posibles resultados de un experimento aleatorio. Se denota com√∫nmente por $( S )$.

**3. Reglas de Probabilidad B√°sicas:**

**Regla de la Suma:**

$P(A \cup B) = P(A) + P(B) - P(A \cap B)$

**Regla del Producto:**

$P(A \cap B) = P(A) \cdot P(B|A)$

**Complemento:**

$P(A') = 1 - P(A)$

**Regla de la Probabilidad Total:**

$P(B) = P(A_1) \cdot P(B|A_1) + P(A_2) \cdot P(B|A_2) + \ldots + P(A_n) \cdot P(B|A_n)$

**Ejemplo y C√≥digo Python:**

Supongamos un experimento de lanzar dos dados:

```python
import itertools

# Espacio Muestral
dados = [1, 2, 3, 4, 5, 6]
espacio_muestral = list(itertools.product(dados, repeat=2))

# Evento A: Suma mayor o igual a 8
evento_A = [(d1, d2) for d1, d2 in espacio_muestral if d1 + d2 >= 8]

# Evento B: Dado 1 es par
evento_B = [(d1, d2) for d1, d2 in espacio_muestral if d1 % 2 == 0]

# Probabilidades
P_A = len(evento_A) / len(espacio_muestral)
P_B_given_A = len([(d1, d2) for (d1, d2) in evento_A if d1 % 2 == 0]) / len(evento_A)
P_B = len(evento_B) / len(espacio_muestral)

# Regla de la Suma
P_A_union_B = P_A + P_B - P_A * P_B_given_A

# Regla del Producto
P_A_intersect_B = P_A * P_B_given_A

# Complemento
P_A_complemento = 1 - P_A

# Regla de la Probabilidad Total
P_B_total = P_A * P_B_given_A + (1 - P_A) * P_B_given_A

print("Espacio Muestral:", espacio_muestral)
print("Evento A:", evento_A)
print("Evento B:", evento_B)

print("\nProbabilidades:")
print("P(A):", P_A)
print("P(B|A):", P_B_given_A)
print("P(B):", P_B)

print("\nRegla de la Suma:")
print("P(A U B):", P_A_union_B)

print("\nRegla del Producto:")
print("P(A ‚à© B):", P_A_intersect_B)

print("\nComplemento:")
print("P(A'): ", P_A_complemento)

print("\nRegla de la Probabilidad Total:")
print("P(B):", P_B_total)
```

Este c√≥digo en Python simula un experimento de lanzar dos dados, calcula eventos, espacio muestral y aplica las reglas de probabilidad b√°sicas. Los resultados se obtienen utilizando f√≥rmulas matem√°ticas y se imprimen para su visualizaci√≥n.

### **Probabilidad condicional**

La probabilidad condicional se refiere a la probabilidad de que ocurra un evento $( B )$ dado que otro evento $( A )$ ha ocurrido. Se denota por  $P(B|A)$ y se calcula como:

$P(B|A) = \frac{P(A \cap B)}{P(A)}$

donde $P(A)$ es la probabilidad de que ocurra $( A )$ y $P(A \cap B)$ es la probabilidad de que ocurran ambos eventos $( A )$ y $( B )$.

**Ejemplo y C√≥digo Python:**

Supongamos que lanzamos dos dados y estamos interesados en la probabilidad de que la suma sea mayor o igual a 8 dado que el primer dado es un n√∫mero par.

```python
import itertools
import matplotlib.pyplot as plt

# Espacio Muestral
dados = [1, 2, 3, 4, 5, 6]
espacio_muestral = list(itertools.product(dados, repeat=2))

# Evento A: Primer dado es par
evento_A = [(d1, d2) for d1, d2 in espacio_muestral if d1 % 2 == 0]

# Evento B: Suma mayor o igual a 8
evento_B = [(d1, d2) for d1, d2 in espacio_muestral if d1 + d2 >= 8]

# Probabilidad de A
P_A = len(evento_A) / len(espacio_muestral)

# Probabilidad conjunta de A y B
P_A_intersect_B = len([(d1, d2) for (d1, d2) in evento_A if (d1, d2) in evento_B]) / len(espacio_muestral)

# Probabilidad condicional de B dado A
P_B_given_A = P_A_intersect_B / P_A

# Visualizaci√≥n en gr√°fico
fig, ax = plt.subplots()
ax.scatter(*zip(*espacio_muestral), color='blue', alpha=0.1, label='Espacio Muestral')
ax.scatter(*zip(*evento_A), color='green', label='Evento A')
ax.scatter(*zip(*evento_B), color='red', label='Evento B')
ax.set_xlabel('Dado 1')
ax.set_ylabel('Dado 2')
ax.legend()
plt.title('Espacio Muestral, Evento A y Evento B')
plt.show()

print("Probabilidad condicional P(B|A):", P_B_given_A)
```

Este c√≥digo en Python simula el lanzamiento de dos dados y calcula la probabilidad condicional de que la suma sea mayor o igual a 8 dado que el primer dado es un n√∫mero par. Adem√°s, muestra un gr√°fico para visualizar el espacio muestral, el evento $( A )$ y el evento $( B )$.

## **Distribuciones de Probabilidad**

### **Distribuciones discretas (binomial, poisson)**

1. **Distribuci√≥n Binomial:**

La distribuci√≥n binomial describe el n√∫mero de √©xitos en una secuencia fija de ensayos independientes, cada uno con la misma probabilidad de √©xito.

**F√≥rmula Matem√°tica:**

$P(X = k) = \binom{n}{k} \cdot p^k \cdot (1 - p)^{n-k}$

donde:

- $( n )$ es el n√∫mero total de ensayos,
- $( k )$ es el n√∫mero de √©xitos,
- $( p )$ es la probabilidad de √©xito en un solo ensayo,
- $(1 - p)$ es la probabilidad de fracaso en un solo ensayo,
- $\binom{n}{k}$ es el coeficiente binomial, tambi√©n conocido como "n choose k".

**Ejemplo y C√≥digo Python:**

Supongamos que lanzamos un dado justo (6 caras) 5 veces y queremos encontrar la probabilidad de obtener exactamente 2 veces el n√∫mero 4.

```python
import scipy.stats as stats
import matplotlib.pyplot as plt
import numpy as np

# Par√°metros de la distribuci√≥n binomial
n = 5  # n√∫mero total de ensayos (lanzamientos del dado)
p = 1/6  # probabilidad de √©xito en un solo ensayo (obtener el n√∫mero 4)
k = 2  # n√∫mero de √©xitos deseados

# Probabilidad de obtener exactamente k √©xitos
probabilidad = stats.binom.pmf(k, n, p)

# Generar datos para graficar la distribuci√≥n
x = np.arange(0, n + 1)
y = stats.binom.pmf(x, n, p)

# Visualizaci√≥n en gr√°fico
plt.bar(x, y, align='center', alpha=0.7)
plt.title('Distribuci√≥n Binomial')
plt.xlabel('N√∫mero de √âxitos')
plt.ylabel('Probabilidad')
plt.show()

print("Probabilidad de obtener exactamente 2 veces el n√∫mero 4:", probabilidad)
```

Este c√≥digo en Python utiliza la biblioteca `scipy.stats` para calcular la probabilidad de obtener exactamente 2 √©xitos al lanzar un dado justo 5 veces, siguiendo una distribuci√≥n binomial.

2. **Distribuci√≥n de Poisson:**

La distribuci√≥n de Poisson modela la probabilidad de un n√∫mero dado de eventos que ocurren en un intervalo fijo de tiempo o espacio.

**F√≥rmula Matem√°tica:**

$P(X = k) = \frac{e^{-\lambda} \cdot \lambda^k}{k!}$

donde:

- $( k )$ es el n√∫mero de eventos,
- $( \lambda )$ es la tasa promedio de eventos por intervalo,
- $( e )$ es la base del logaritmo natural.

**Ejemplo y C√≥digo Python:**

Supongamos que, en promedio, se producen 3 accidentes de tr√°fico por d√≠a. Queremos encontrar la probabilidad de que ocurran exactamente 2 accidentes en un d√≠a dado.

```python
# Par√°metros de la distribuci√≥n de Poisson
lmbda = 3  # tasa promedio de eventos por d√≠a
k = 2  # n√∫mero de eventos deseados

# Probabilidad de obtener exactamente k eventos
probabilidad_poisson = stats.poisson.pmf(k, lmbda)

# Generar datos para graficar la distribuci√≥n
x_poisson = np.arange(0, 10)
y_poisson = stats.poisson.pmf(x_poisson, lmbda)

# Visualizaci√≥n en gr√°fico
plt.bar(x_poisson, y_poisson, align='center', alpha=0.7)
plt.title('Distribuci√≥n de Poisson')
plt.xlabel('N√∫mero de Eventos')
plt.ylabel('Probabilidad')
plt.show()

print("Probabilidad de que ocurran exactamente 2 accidentes en un d√≠a:", probabilidad_poisson)
```

Este c√≥digo en Python utiliza la biblioteca `scipy.stats` para calcular la probabilidad de que ocurran exactamente 2 accidentes en un d√≠a, siguiendo una distribuci√≥n de Poisson con una tasa promedio de 3 accidentes por d√≠a.

### **Distribuciones continuas (normal, exponencial)**

1. **Distribuci√≥n Normal:**

La distribuci√≥n normal (o gaussiana) es una de las distribuciones m√°s comunes en estad√≠sticas. Se caracteriza por su forma de campana y est√° completamente determinada por su media $( \mu )$ y su desviaci√≥n est√°ndar $( \sigma )$.

**F√≥rmula Matem√°tica:**

La funci√≥n de densidad de probabilidad (PDF) de una distribuci√≥n normal es:

$f(x|\mu, \sigma) = \frac{1}{\sqrt{2\pi\sigma^2}} \cdot e^{-\frac{(x - \mu)^2}{2\sigma^2}}$

donde:

- $( \mu )$ es la media,
- $( \sigma )$ es la desviaci√≥n est√°ndar,
- $( e )$ es la base del logaritmo natural.

**Ejemplo y C√≥digo Python:**

Supongamos que estamos modelando la altura de estudiantes universitarios, y sabemos que la altura sigue una distribuci√≥n normal con media $( \mu = 170 )$ cm y desviaci√≥n est√°ndar $( \sigma = 5 )$ cm. Queremos encontrar la probabilidad de que un estudiante tenga una altura entre 165 y 175 cm.

```python
import scipy.stats as stats
import matplotlib.pyplot as plt
import numpy as np

# Par√°metros de la distribuci√≥n normal
media = 170
desviacion_estandar = 5

# Probabilidad de que un estudiante tenga altura entre 165 y 175 cm
probabilidad_normal = stats.norm.cdf(175, media, desviacion_estandar) - stats.norm.cdf(165, media, desviacion_estandar)

# Generar datos para graficar la distribuci√≥n
x_normal = np.linspace(150, 190, 1000)
y_normal = stats.norm.pdf(x_normal, media, desviacion_estandar)

# Visualizaci√≥n en gr√°fico
plt.plot(x_normal, y_normal, label='Distribuci√≥n Normal')
plt.fill_betweenx(y_normal, x_normal, where=[(x >= 165) and (x <= 175) for x in x_normal], alpha=0.3, label='√Årea de inter√©s')
plt.title('Distribuci√≥n Normal')
plt.xlabel('Altura (cm)')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.show()

print("Probabilidad de que un estudiante tenga altura entre 165 y 175 cm:", probabilidad_normal)
```

Este c√≥digo en Python utiliza la biblioteca `scipy.stats` para calcular la probabilidad de que un estudiante tenga una altura entre 165 y 175 cm, siguiendo una distribuci√≥n normal con media 170 cm y desviaci√≥n est√°ndar 5 cm.

2. **Distribuci√≥n Exponencial:**

La distribuci√≥n exponencial modela el tiempo entre eventos en un proceso de Poisson, donde los eventos ocurren de manera continua e independiente a una tasa constante.

**F√≥rmula Matem√°tica:**

La funci√≥n de densidad de probabilidad (PDF) de una distribuci√≥n exponencial es:

$f(x|\lambda) = \lambda \cdot e^{-\lambda x}$

donde:

- $( \lambda )$ es la tasa de ocurrencia de eventos,
- $( e )$ es la base del logaritmo natural.

**Ejemplo y C√≥digo Python:**

Supongamos que estamos modelando el tiempo entre llegadas de clientes a un servicio, y sabemos que el tiempo sigue una distribuci√≥n exponencial con una tasa de llegada de $( \lambda = 0.1 )$ clientes por minuto. Queremos encontrar la probabilidad de que el tiempo entre dos llegadas consecutivas sea menor a 8 minutos.

```python
# Par√°metro de la distribuci√≥n exponencial
tasa_llegada = 0.1

# Probabilidad de que el tiempo entre dos llegadas consecutivas sea menor a 8 minutos
probabilidad_exponencial = stats.expon.cdf(8, scale=1/tasa_llegada)

# Generar datos para graficar la distribuci√≥n
x_exponencial = np.linspace(0, 30, 1000)
y_exponencial = stats.expon.pdf(x_exponencial, scale=1/tasa_llegada)

# Visualizaci√≥n en gr√°fico
plt.plot(x_exponencial, y_exponencial, label='Distribuci√≥n Exponencial')
plt.fill_betweenx(y_exponencial, x_exponencial, where=(x_exponencial < 8), alpha=0.3, label='√Årea de inter√©s')
plt.title('Distribuci√≥n Exponencial')
plt.xlabel('Tiempo entre llegadas (minutos)')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.show()

print("Probabilidad de que el tiempo entre dos llegadas consecutivas sea menor a 8 minutos:", probabilidad_exponencial)
```

Este c√≥digo en Python utiliza la biblioteca `scipy.stats` para calcular la probabilidad de que el tiempo entre dos llegadas consecutivas sea menor a 8 minutos, siguiendo una distribuci√≥n exponencial con una tasa de llegada de 0.1 clientes por minuto.
