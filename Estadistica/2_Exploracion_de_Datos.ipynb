# **2. Exploraci贸n de Datos**

## **Medidas de Tendencia Central**

### **Media, mediana y moda**

La media, mediana y moda son medidas de tendencia central utilizadas en estad铆sticas descriptivas para resumir conjuntos de datos. Aqu铆 hay una explicaci贸n detallada con ejemplos y gr谩ficos en Python:

**1. Media:**

La media es el promedio de un conjunto de datos y se calcula sumando todos los valores y dividiendo por el n煤mero total de observaciones.

$\text{Media} = \frac{\sum_{i=1}^{n} X_i}{n}$

```python
import numpy as np
import matplotlib.pyplot as plt

# Crear datos para ejemplo
datos = np.random.normal(loc=50, scale=10, size=100)

# Calcular la media
media = np.mean(datos)

# Gr谩fico de histograma con personalizaciones
plt.figure(figsize=(10, 6))
plt.hist(datos, bins=20, color='skyblue', edgecolor='black')

# L铆nea vertical para la media
plt.axvline(media, color='red', linestyle='dashed', linewidth=2, label=f'Media = {media:.2f}')

# T铆tulo y etiquetas del eje
plt.title('Histograma con Media', fontsize=16)
plt.xlabel('Datos', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)

# Leyenda
plt.legend()

# Mostrar la gr谩fica
plt.show()
```

**2. Mediana:**

La mediana es el valor que separa la mitad superior e inferior de un conjunto de datos ordenados. Para encontrar la mediana, primero ordenamos los datos y luego seleccionamos el valor en el medio.

La mediana es el valor que se encuentra en el centro de un conjunto de datos ordenados, de manera que divide el conjunto en dos partes iguales. La f贸rmula para calcular la mediana depende de si la cantidad de datos $(n)$ en el conjunto es impar o par.

1. **Para una cantidad impar de datos $(n)$ impar:**

   La mediana $(M)$ es simplemente el valor que ocupa la posici贸n central en el conjunto ordenado. Si los datos est谩n ordenados de menor a mayor, la posici贸n de la mediana es $(\frac{n + 1}{2})$.

   $M = \text{valor en la posici贸n } \frac{n + 1}{2}$

2. **Para una cantidad par de datos $(n)$ par:**

   En este caso, la mediana es el promedio de los dos valores centrales. Si los datos est谩n ordenados de menor a mayor, las posiciones de los valores centrales son $(\frac{n}{2})$ y $(\frac{n}{2} + 1)$.

   $M = \frac{\text{valor en la posici贸n } \frac{n}{2} + \text{valor en la posici贸n } \frac{n}{2} + 1}{2}$

En resumen, la mediana es una medida de tendencia central que no se ve afectada por valores extremos (a diferencia del promedio). Si la cantidad de datos es impar, la mediana es el valor central; si la cantidad es par, es el promedio de los dos valores centrales.

```python
import numpy as np
import matplotlib.pyplot as plt

# Crear datos para ejemplo
datos = np.random.normal(loc=50, scale=10, size=100)

# Calcular la mediana
mediana = np.median(datos)

# Gr谩fico de histograma con personalizaciones
plt.figure(figsize=(10, 6))
plt.hist(datos, bins=20, color='skyblue', edgecolor='black')

# L铆nea vertical para la mediana
plt.axvline(mediana, color='green', linestyle='dashed', linewidth=2, label=f'Mediana = {mediana:.2f}')

# T铆tulo y etiquetas del eje
plt.title('Histograma con Mediana', fontsize=16)
plt.xlabel('Datos', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)

# Leyenda
plt.legend()

# A帽adir una nota sobre la mediana
plt.text(mediana + 2, 10, f'Mediana = {mediana:.2f}', color='green', fontsize=12, bbox=dict(facecolor='white', alpha=0.8))

# Mostrar la gr谩fica
plt.show()
```

**3. Moda:**

La moda es el valor que aparece con mayor frecuencia en un conjunto de datos. Puede haber una moda (datos unimodales), dos modas (datos bimodales), o m谩s.

Estos ejemplos utilizan datos aleatorios generados con distribuci贸n normal, y los gr谩ficos muestran la posici贸n de la media, mediana y moda en el contexto de un histograma. Estas medidas proporcionan informaci贸n sobre la tendencia central de los datos.

## **Medidas de Dispersi贸n**

### **Rango, desviaci贸n est谩ndar y varianza**

**Rango:**

El rango en estad铆sticas es la diferencia entre el valor m谩ximo y el valor m铆nimo en un conjunto de datos. Proporciona una medida de la dispersi贸n o amplitud del conjunto. Matem谩ticamente, se calcula como $( R = \text{valor m谩ximo} - \text{valor m铆nimo} )$.

Voy a generar datos de ejemplo y mostrar el c谩lculo del rango con un gr谩fico:

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos de ejemplo
datos = np.random.randint(0, 100, size=50)

# Calcular el rango
rango = np.max(datos) - np.min(datos)

# Gr谩fico de los datos
plt.figure(figsize=(10, 6))
plt.scatter(range(len(datos)), datos, color='blue', label='Datos')
plt.title('Datos de Ejemplo con Rango')
plt.xlabel('ndice')
plt.ylabel('Valor')
plt.axhline(y=np.max(datos), color='red', linestyle='--', label='Valor M谩ximo')
plt.axhline(y=np.min(datos), color='green', linestyle='--', label='Valor M铆nimo')
plt.legend()
plt.show()

print(f"El rango de los datos es: {rango}")
```

**Desviaci贸n est谩ndar y Varianza:**

La varianza y la desviaci贸n est谩ndar son medidas de dispersi贸n que indican cu谩nto se alejan los valores individuales del centro del conjunto de datos.

La varianza $(S^2)$ es el promedio de las diferencias al cuadrado entre cada valor y la media del conjunto. La desviaci贸n est谩ndar $(S)$ es simplemente la ra铆z cuadrada de la varianza $(S = \sqrt{S^2})$.

Ahora, generemos datos de ejemplo, calculemos la varianza y la desviaci贸n est谩ndar, y creemos un gr谩fico para visualizar la dispersi贸n:

```python
# Generar datos de ejemplo
datos = np.random.normal(loc=50, scale=10, size=100)

# Calcular la varianza y desviaci贸n est谩ndar
varianza = np.var(datos)
desviacion_estandar = np.std(datos)

# Gr谩fico de histograma para visualizar la dispersi贸n
plt.figure(figsize=(10, 6))
plt.hist(datos, bins=20, color='skyblue', edgecolor='black')
plt.title('Histograma con Desviaci贸n Est谩ndar y Varianza')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.axvline(x=np.mean(datos), color='red', linestyle='dashed', linewidth=2, label='Media')
plt.axvline(x=np.mean(datos) + desviacion_estandar, color='orange', linestyle='dashed', linewidth=2, label='Desviaci贸n Est谩ndar')
plt.axvline(x=np.mean(datos) - desviacion_estandar, color='orange', linestyle='dashed', linewidth=2)
plt.legend()
plt.show()

print(f"La varianza de los datos es: {varianza}")
print(f"La desviaci贸n est谩ndar de los datos es: {desviacion_estandar}")
```

En este ejemplo, se generaron datos normalmente distribuidos con una media de 50 y una desviaci贸n est谩ndar de 10. El gr谩fico muestra la distribuci贸n de los datos junto con la media y la desviaci贸n est谩ndar.

### **Interpretaci贸n de la dispersi贸n de datos**

La dispersi贸n en estad铆sticas se refiere a la extensi贸n o propagaci贸n de los valores en un conjunto de datos. Las medidas de dispersi贸n proporcionan informaci贸n sobre cu谩n dispersos o concentrados est谩n los valores en torno a una medida central, como la media. Las principales medidas de dispersi贸n son el rango, la varianza y la desviaci贸n est谩ndar. Aqu铆, explicar茅 la interpretaci贸n de estas medidas con ejemplos y gr谩ficos personalizados en Python.

**1. Rango:**

El rango mide la amplitud total de los datos, es decir, la diferencia entre el valor m谩ximo y el valor m铆nimo. Un rango grande indica una mayor dispersi贸n.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos de ejemplo
datos = np.random.randint(0, 100, size=50)

# Calcular el rango
rango = np.max(datos) - np.min(datos)

# Gr谩fico de los datos
plt.figure(figsize=(10, 6))
plt.scatter(range(len(datos)), datos, color='blue', label='Datos')
plt.title('Datos de Ejemplo con Rango')
plt.xlabel('ndice')
plt.ylabel('Valor')
plt.axhline(y=np.max(datos), color='red', linestyle='--', label='Valor M谩ximo')
plt.axhline(y=np.min(datos), color='green', linestyle='--', label='Valor M铆nimo')
plt.legend()
plt.show()

print(f"El rango de los datos es: {rango}")
```

En este ejemplo, el gr谩fico muestra la dispersi贸n de los datos y c贸mo el rango se calcula a partir del valor m谩ximo y m铆nimo.

**2. Varianza y Desviaci贸n Est谩ndar:**

La varianza y la desviaci贸n est谩ndar miden cu谩n dispersos est谩n los datos en relaci贸n con la media. Una varianza grande indica mayor dispersi贸n.

```python
# Generar datos de ejemplo
datos = np.random.normal(loc=50, scale=10, size=100)

# Calcular la varianza y desviaci贸n est谩ndar
varianza = np.var(datos)
desviacion_estandar = np.std(datos)

# Gr谩fico de histograma para visualizar la dispersi贸n
plt.figure(figsize=(10, 6))
plt.hist(datos, bins=20, color='skyblue', edgecolor='black')
plt.title('Histograma con Desviaci贸n Est谩ndar y Varianza')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.axvline(x=np.mean(datos), color='red', linestyle='dashed', linewidth=2, label='Media')
plt.axvline(x=np.mean(datos) + desviacion_estandar, color='orange', linestyle='dashed', linewidth=2, label='Desviaci贸n Est谩ndar')
plt.axvline(x=np.mean(datos) - desviacion_estandar, color='orange', linestyle='dashed', linewidth=2)
plt.legend()
plt.show()

print(f"La varianza de los datos es: {varianza}")
print(f"La desviaci贸n est谩ndar de los datos es: {desviacion_estandar}")
```

Este gr谩fico muestra c贸mo la varianza y la desviaci贸n est谩ndar describen la dispersi贸n de los datos alrededor de la media. La l铆nea roja representa la media, y las l铆neas naranjas indican la desviaci贸n est谩ndar. Cuanto mayor sea la desviaci贸n est谩ndar, mayor ser谩 la dispersi贸n de los datos.
