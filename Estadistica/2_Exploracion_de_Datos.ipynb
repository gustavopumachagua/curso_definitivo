# **2. Exploraci√≥n de Datos**

## **Medidas de Tendencia Central**

### **Media, mediana y moda**

La media, mediana y moda son medidas de tendencia central utilizadas en estad√≠sticas descriptivas para resumir conjuntos de datos. Aqu√≠ hay una explicaci√≥n detallada con ejemplos y gr√°ficos en Python:

**1. Media:**

La media es el promedio de un conjunto de datos y se calcula sumando todos los valores y dividiendo por el n√∫mero total de observaciones.

$\text{Media} = \frac{\sum_{i=1}^{n} X_i}{n}$

```python
import numpy as np
import matplotlib.pyplot as plt

# Crear datos para ejemplo
datos = np.random.normal(loc=50, scale=10, size=100)

# Calcular la media
media = np.mean(datos)

# Gr√°fico de histograma con personalizaciones
plt.figure(figsize=(10, 6))
plt.hist(datos, bins=20, color='skyblue', edgecolor='black')

# L√≠nea vertical para la media
plt.axvline(media, color='red', linestyle='dashed', linewidth=2, label=f'Media = {media:.2f}')

# T√≠tulo y etiquetas del eje
plt.title('Histograma con Media', fontsize=16)
plt.xlabel('Datos', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)

# Leyenda
plt.legend()

# Mostrar la gr√°fica
plt.show()
```

**2. Mediana:**

La mediana es el valor que separa la mitad superior e inferior de un conjunto de datos ordenados. Para encontrar la mediana, primero ordenamos los datos y luego seleccionamos el valor en el medio.

La mediana es el valor que se encuentra en el centro de un conjunto de datos ordenados, de manera que divide el conjunto en dos partes iguales. La f√≥rmula para calcular la mediana depende de si la cantidad de datos $(n)$ en el conjunto es impar o par.

1. **Para una cantidad impar de datos $(n)$ impar:**

   La mediana $(M)$ es simplemente el valor que ocupa la posici√≥n central en el conjunto ordenado. Si los datos est√°n ordenados de menor a mayor, la posici√≥n de la mediana es $(\frac{n + 1}{2})$.

   $M = \text{valor en la posici√≥n } \frac{n + 1}{2}$

2. **Para una cantidad par de datos $(n)$ par:**

   En este caso, la mediana es el promedio de los dos valores centrales. Si los datos est√°n ordenados de menor a mayor, las posiciones de los valores centrales son $(\frac{n}{2})$ y $(\frac{n}{2} + 1)$.

   $M = \frac{\text{valor en la posici√≥n } \frac{n}{2} + \text{valor en la posici√≥n } \frac{n}{2} + 1}{2}$

En resumen, la mediana es una medida de tendencia central que no se ve afectada por valores extremos (a diferencia del promedio). Si la cantidad de datos es impar, la mediana es el valor central; si la cantidad es par, es el promedio de los dos valores centrales.

```python
import numpy as np
import matplotlib.pyplot as plt

# Crear datos para ejemplo
datos = np.random.normal(loc=50, scale=10, size=100)

# Calcular la mediana
mediana = np.median(datos)

# Gr√°fico de histograma con personalizaciones
plt.figure(figsize=(10, 6))
plt.hist(datos, bins=20, color='skyblue', edgecolor='black')

# L√≠nea vertical para la mediana
plt.axvline(mediana, color='green', linestyle='dashed', linewidth=2, label=f'Mediana = {mediana:.2f}')

# T√≠tulo y etiquetas del eje
plt.title('Histograma con Mediana', fontsize=16)
plt.xlabel('Datos', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)

# Leyenda
plt.legend()

# A√±adir una nota sobre la mediana
plt.text(mediana + 2, 10, f'Mediana = {mediana:.2f}', color='green', fontsize=12, bbox=dict(facecolor='white', alpha=0.8))

# Mostrar la gr√°fica
plt.show()
```

**3. Moda:**

La moda es el valor que aparece con mayor frecuencia en un conjunto de datos. Puede haber una moda (datos unimodales), dos modas (datos bimodales), o m√°s.

Estos ejemplos utilizan datos aleatorios generados con distribuci√≥n normal, y los gr√°ficos muestran la posici√≥n de la media, mediana y moda en el contexto de un histograma. Estas medidas proporcionan informaci√≥n sobre la tendencia central de los datos.

### **Interpretaci√≥n y uso en diferentes contextos**

**1. Media (Promedio):**

La media es el promedio aritm√©tico de un conjunto de datos y se calcula sumando todos los valores y dividiendo por el n√∫mero de observaciones.

$\text{Media} = \frac{\sum_{i=1}^{n} x_i}{n}$

**Interpretaci√≥n:**

La media representa el "centro de masa" del conjunto de datos y es sensible a los valores at√≠picos (outliers).

**C√≥digo Python con Gr√°fico Personalizado:**

```python
import numpy as np
import matplotlib.pyplot as plt

# Crear datos para ejemplo
datos = np.random.normal(loc=50, scale=10, size=100)

# Calcular la media
media = np.mean(datos)

# Gr√°fico de histograma con la media
plt.figure(figsize=(10, 6))
plt.hist(datos, bins=20, color='skyblue', edgecolor='black')
plt.axvline(media, color='red', linestyle='dashed', linewidth=2, label=f'Media = {media:.2f}')
plt.title('Histograma con Media')
plt.xlabel('Datos')
plt.ylabel('Frecuencia')
plt.legend()
plt.show()
```

Este gr√°fico muestra un histograma con una l√≠nea punteada que representa la media.

**2. Mediana:**

La mediana es el valor que divide el conjunto de datos ordenado en dos partes iguales.

**Interpretaci√≥n:**

La mediana es menos sensible a los valores at√≠picos y proporciona informaci√≥n sobre la ubicaci√≥n central real del conjunto de datos.

**C√≥digo Python con Gr√°fico Personalizado:**

```python
# Calcular la mediana
mediana = np.median(datos)

# Gr√°fico de histograma con la mediana
plt.figure(figsize=(10, 6))
plt.hist(datos, bins=20, color='skyblue', edgecolor='black')
plt.axvline(mediana, color='green', linestyle='dashed', linewidth=2, label=f'Mediana = {mediana:.2f}')
plt.title('Histograma con Mediana')
plt.xlabel('Datos')
plt.ylabel('Frecuencia')
plt.legend()
plt.show()
```

Este gr√°fico muestra un histograma con una l√≠nea punteada que representa la mediana.

**3. Moda:**

La moda es el valor que ocurre con mayor frecuencia en el conjunto de datos.

**Interpretaci√≥n:**

La moda es √∫til para identificar los valores m√°s comunes en un conjunto de datos.

## **Medidas de Dispersi√≥n**

### **Rango, desviaci√≥n est√°ndar y varianza**

**Rango:**

El rango en estad√≠sticas es la diferencia entre el valor m√°ximo y el valor m√≠nimo en un conjunto de datos. Proporciona una medida de la dispersi√≥n o amplitud del conjunto. Matem√°ticamente, se calcula como $( R = \text{valor m√°ximo} - \text{valor m√≠nimo} )$.

Voy a generar datos de ejemplo y mostrar el c√°lculo del rango con un gr√°fico:

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos de ejemplo
datos = np.random.randint(0, 100, size=50)

# Calcular el rango
rango = np.max(datos) - np.min(datos)

# Gr√°fico de los datos
plt.figure(figsize=(10, 6))
plt.scatter(range(len(datos)), datos, color='blue', label='Datos')
plt.title('Datos de Ejemplo con Rango')
plt.xlabel('√çndice')
plt.ylabel('Valor')
plt.axhline(y=np.max(datos), color='red', linestyle='--', label='Valor M√°ximo')
plt.axhline(y=np.min(datos), color='green', linestyle='--', label='Valor M√≠nimo')
plt.legend()
plt.show()

print(f"El rango de los datos es: {rango}")
```

**Desviaci√≥n est√°ndar y Varianza:**

La varianza y la desviaci√≥n est√°ndar son medidas de dispersi√≥n que indican cu√°nto se alejan los valores individuales del centro del conjunto de datos.

La varianza $(S^2)$ es el promedio de las diferencias al cuadrado entre cada valor y la media del conjunto. La desviaci√≥n est√°ndar $(S)$ es simplemente la ra√≠z cuadrada de la varianza $(S = \sqrt{S^2})$.

Ahora, generemos datos de ejemplo, calculemos la varianza y la desviaci√≥n est√°ndar, y creemos un gr√°fico para visualizar la dispersi√≥n:

```python
# Generar datos de ejemplo
datos = np.random.normal(loc=50, scale=10, size=100)

# Calcular la varianza y desviaci√≥n est√°ndar
varianza = np.var(datos)
desviacion_estandar = np.std(datos)

# Gr√°fico de histograma para visualizar la dispersi√≥n
plt.figure(figsize=(10, 6))
plt.hist(datos, bins=20, color='skyblue', edgecolor='black')
plt.title('Histograma con Desviaci√≥n Est√°ndar y Varianza')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.axvline(x=np.mean(datos), color='red', linestyle='dashed', linewidth=2, label='Media')
plt.axvline(x=np.mean(datos) + desviacion_estandar, color='orange', linestyle='dashed', linewidth=2, label='Desviaci√≥n Est√°ndar')
plt.axvline(x=np.mean(datos) - desviacion_estandar, color='orange', linestyle='dashed', linewidth=2)
plt.legend()
plt.show()

print(f"La varianza de los datos es: {varianza}")
print(f"La desviaci√≥n est√°ndar de los datos es: {desviacion_estandar}")
```

En este ejemplo, se generaron datos normalmente distribuidos con una media de 50 y una desviaci√≥n est√°ndar de 10. El gr√°fico muestra la distribuci√≥n de los datos junto con la media y la desviaci√≥n est√°ndar.

### **Interpretaci√≥n de la dispersi√≥n de datos**

La dispersi√≥n en estad√≠sticas se refiere a la extensi√≥n o propagaci√≥n de los valores en un conjunto de datos. Las medidas de dispersi√≥n proporcionan informaci√≥n sobre cu√°n dispersos o concentrados est√°n los valores en torno a una medida central, como la media. Las principales medidas de dispersi√≥n son el rango, la varianza y la desviaci√≥n est√°ndar. Aqu√≠, explicar√© la interpretaci√≥n de estas medidas con ejemplos y gr√°ficos personalizados en Python.

**1. Rango:**

El rango mide la amplitud total de los datos, es decir, la diferencia entre el valor m√°ximo y el valor m√≠nimo. Un rango grande indica una mayor dispersi√≥n.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos de ejemplo
datos = np.random.randint(0, 100, size=50)

# Calcular el rango
rango = np.max(datos) - np.min(datos)

# Gr√°fico de los datos
plt.figure(figsize=(10, 6))
plt.scatter(range(len(datos)), datos, color='blue', label='Datos')
plt.title('Datos de Ejemplo con Rango')
plt.xlabel('√çndice')
plt.ylabel('Valor')
plt.axhline(y=np.max(datos), color='red', linestyle='--', label='Valor M√°ximo')
plt.axhline(y=np.min(datos), color='green', linestyle='--', label='Valor M√≠nimo')
plt.legend()
plt.show()

print(f"El rango de los datos es: {rango}")
```

En este ejemplo, el gr√°fico muestra la dispersi√≥n de los datos y c√≥mo el rango se calcula a partir del valor m√°ximo y m√≠nimo.

**2. Varianza y Desviaci√≥n Est√°ndar:**

La varianza y la desviaci√≥n est√°ndar miden cu√°n dispersos est√°n los datos en relaci√≥n con la media. Una varianza grande indica mayor dispersi√≥n.

```python
# Generar datos de ejemplo
datos = np.random.normal(loc=50, scale=10, size=100)

# Calcular la varianza y desviaci√≥n est√°ndar
varianza = np.var(datos)
desviacion_estandar = np.std(datos)

# Gr√°fico de histograma para visualizar la dispersi√≥n
plt.figure(figsize=(10, 6))
plt.hist(datos, bins=20, color='skyblue', edgecolor='black')
plt.title('Histograma con Desviaci√≥n Est√°ndar y Varianza')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.axvline(x=np.mean(datos), color='red', linestyle='dashed', linewidth=2, label='Media')
plt.axvline(x=np.mean(datos) + desviacion_estandar, color='orange', linestyle='dashed', linewidth=2, label='Desviaci√≥n Est√°ndar')
plt.axvline(x=np.mean(datos) - desviacion_estandar, color='orange', linestyle='dashed', linewidth=2)
plt.legend()
plt.show()

print(f"La varianza de los datos es: {varianza}")
print(f"La desviaci√≥n est√°ndar de los datos es: {desviacion_estandar}")
```

Este gr√°fico muestra c√≥mo la varianza y la desviaci√≥n est√°ndar describen la dispersi√≥n de los datos alrededor de la media. La l√≠nea roja representa la media, y las l√≠neas naranjas indican la desviaci√≥n est√°ndar. Cuanto mayor sea la desviaci√≥n est√°ndar, mayor ser√° la dispersi√≥n de los datos.
