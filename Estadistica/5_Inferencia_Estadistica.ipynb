# **5. Inferencia Estad√≠stica**

## **Muestreo y Estimaci√≥n Puntual**

### **M√©todos de muestreo**

Los m√©todos de muestreo son t√©cnicas utilizadas para seleccionar una muestra representativa de una poblaci√≥n con el objetivo de realizar inferencias sobre la poblaci√≥n completa. Aqu√≠ se presentar√°n dos m√©todos comunes: muestreo aleatorio simple y muestreo estratificado.

1. **Muestreo Aleatorio Simple:**

En este m√©todo, cada individuo de la poblaci√≥n tiene la misma probabilidad de ser seleccionado. Se selecciona una muestra sin reemplazo, lo que significa que una vez que un individuo es seleccionado, no puede ser seleccionado nuevamente.

**Ejemplo y C√≥digo Python:**

Supongamos que queremos seleccionar una muestra aleatoria simple de tama√±o $( n )$ de una poblaci√≥n de alturas de estudiantes.

```python
import numpy as np
import matplotlib.pyplot as plt

# Poblaci√≥n de alturas (simulada)
np.random.seed(42)
poblacion_alturas = np.random.normal(loc=170, scale=5, size=1000)

# Muestreo Aleatorio Simple
tamano_muestra = 100
muestra_aleatoria = np.random.choice(poblacion_alturas, size=tamano_muestra, replace=False)

# Personalizaci√≥n de la gr√°fica
plt.figure(figsize=(10, 6))  # Tama√±o de la figura

# Histograma de la poblaci√≥n
plt.hist(poblacion_alturas, bins=30, alpha=0.7, color='blue', label='Poblaci√≥n', edgecolor='black')

# Histograma de la muestra aleatoria
plt.hist(muestra_aleatoria, bins=30, alpha=0.7, color='orange', label='Muestra Aleatoria', edgecolor='black')

# L√≠neas de referencia
plt.axvline(np.mean(poblacion_alturas), color='blue', linestyle='dashed', linewidth=2, label='Media Poblacional')
plt.axvline(np.mean(muestra_aleatoria), color='orange', linestyle='dashed', linewidth=2, label='Media Muestra')

# Etiquetas y t√≠tulos
plt.xlabel('Altura', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)
plt.title('Muestreo Aleatorio Simple', fontsize=16)

# Cuadr√≠cula
plt.grid(True, linestyle='--', alpha=0.5)

# Leyenda
plt.legend()

# Mostrar la gr√°fica
plt.show()
```

Este c√≥digo en Python simula una poblaci√≥n de alturas de estudiantes y realiza un muestreo aleatorio simple para seleccionar una muestra de tama√±o 100. El gr√°fico muestra la distribuci√≥n de alturas en la poblaci√≥n y la muestra seleccionada.

2. **Muestreo Estratificado:**

En el muestreo estratificado, la poblaci√≥n se divide en subgrupos homog√©neos llamados estratos, y luego se selecciona una muestra de cada estrato. Este m√©todo es √∫til cuando la poblaci√≥n tiene heterogeneidad interna.

**Ejemplo y C√≥digo Python:**

Supongamos que queremos realizar un muestreo estratificado de una poblaci√≥n de estudiantes, dividi√©ndolos en dos estratos seg√∫n su g√©nero.

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulaci√≥n de poblaci√≥n estratificada por g√©nero
np.random.seed(42)
alturas_hombres = np.random.normal(loc=175, scale=5, size=500)
alturas_mujeres = np.random.normal(loc=165, scale=5, size=500)

# Muestreo Estratificado
tamano_muestra_por_estrato = 50
muestra_estratificada = np.concatenate([
    np.random.choice(alturas_hombres, size=tamano_muestra_por_estrato, replace=False),
    np.random.choice(alturas_mujeres, size=tamano_muestra_por_estrato, replace=False)
])

# Personalizaci√≥n de la gr√°fica
plt.figure(figsize=(10, 6))  # Tama√±o de la figura

# Histograma de alturas de hombres
plt.hist(alturas_hombres, bins=30, alpha=0.7, color='blue', label='Hombres', edgecolor='black')

# Histograma de alturas de mujeres
plt.hist(alturas_mujeres, bins=30, alpha=0.7, color='pink', label='Mujeres', edgecolor='black')

# Histograma de la muestra estratificada
plt.hist(muestra_estratificada, bins=30, alpha=0.7, color='green', label='Muestra Estratificada', edgecolor='black')

# Etiquetas y t√≠tulos
plt.xlabel('Altura', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)
plt.title('Muestreo Estratificado por G√©nero', fontsize=16)

# Cuadr√≠cula
plt.grid(True, linestyle='--', alpha=0.5)

# Leyenda
plt.legend()

# Mostrar la gr√°fica
plt.show()
```

Este c√≥digo en Python simula una poblaci√≥n de estudiantes divididos por g√©nero y realiza un muestreo estratificado, seleccionando una muestra de tama√±o 50 de cada estrato. El gr√°fico muestra la distribuci√≥n de alturas en la poblaci√≥n y la muestra estratificada.

### **Estimadores y estimaci√≥n puntual**

**Estimador:**

Un estimador es una regla o f√≥rmula que se utiliza para calcular una estimaci√≥n de un par√°metro poblacional. En otras palabras, es una funci√≥n de los datos muestrales que se utiliza para hacer inferencias sobre un par√°metro desconocido de la poblaci√≥n.

**Estimaci√≥n Puntual:**

La estimaci√≥n puntual es el valor num√©rico espec√≠fico que proporciona el estimador para el par√°metro poblacional. Es una √∫nica estimaci√≥n basada en los datos muestrales y se representa t√≠picamente por un solo n√∫mero.

**Ejemplo y C√≥digo Python:**

Supongamos que queremos estimar la media $( \mu )$ de una poblaci√≥n a partir de una muestra. En este caso, la media muestral $( \bar{X} )$ es un estimador puntual para $( \mu )$.

**F√≥rmula Matem√°tica:**

$\bar{X} = \frac{1}{n} \sum_{i=1}^{n} X_i$

donde:

- $( \bar{X} )$ es la media muestral,
- $( n )$ es el tama√±o de la muestra,
- $( X_i )$ son las observaciones individuales en la muestra.

**Ejemplo en Python:**

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar una muestra de una distribuci√≥n normal
np.random.seed(42)
muestra = np.random.normal(loc=10, scale=2, size=100)

# Calcular la media muestral (estimador puntual)
media_muestral = np.mean(muestra)

# Personalizaci√≥n de la gr√°fica
plt.figure(figsize=(10, 6))  # Tama√±o de la figura

# Histograma de la muestra
plt.hist(muestra, bins=20, alpha=0.7, color='skyblue', edgecolor='black', label='Muestra')

# L√≠nea vertical para la media muestral
plt.axvline(media_muestral, color='red', linestyle='dashed', linewidth=2, label='Media Muestral')

# Etiquetas y t√≠tulos
plt.xlabel('Valor', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)
plt.title('Estimaci√≥n Puntual de la Media', fontsize=16)

# Leyenda
plt.legend()

# A√±adir una cuadr√≠cula
plt.grid(True, linestyle='--', alpha=0.5)

# Mostrar la gr√°fica
plt.show()

print("Media Muestral (Estimador Puntual):", media_muestral)
```

En este c√≥digo en Python, generamos una muestra de una distribuci√≥n normal, calculamos la media muestral como estimador puntual y visualizamos la distribuci√≥n de la muestra con la media muestral marcada en rojo en un gr√°fico de histograma. La media muestral se utiliza como una estimaci√≥n puntual para la media poblacional.

## **Intervalos de Confianza**

### **Construcci√≥n e interpretaci√≥n**

**Intervalos de Confianza:**

Un intervalo de confianza (IC) es un rango de valores en el cual se espera que est√© el par√°metro poblacional con cierto nivel de confianza. Proporciona una medida de la incertidumbre asociada con la estimaci√≥n de un par√°metro y se expresa como (por ejemplo) "Estoy 95% seguro de que el valor real del par√°metro est√° dentro de este intervalo".

**Construcci√≥n de un Intervalo de Confianza para la Media (poblacional):**

Supongamos que queremos construir un intervalo de confianza para la media $( \mu )$ de una poblaci√≥n a partir de una muestra.

**F√≥rmula Matem√°tica (para una muestra grande):**

$\bar{X} \pm Z \cdot \frac{\sigma}{\sqrt{n}}$

donde:

- $( \bar{X} )$ es la media muestral,
- $( Z )$ es el valor cr√≠tico de la distribuci√≥n normal est√°ndar que corresponde al nivel de confianza deseado,
- $( \sigma )$ es la desviaci√≥n est√°ndar poblacional (si es conocida),
- $( n )$ es el tama√±o de la muestra.

**Ejemplo y C√≥digo Python:**

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Generar una muestra de una distribuci√≥n normal
np.random.seed(42)
muestra = np.random.normal(loc=10, scale=2, size=100)

# Calcular la media muestral y la desviaci√≥n est√°ndar muestral
media_muestral = np.mean(muestra)
desviacion_estandar_muestral = np.std(muestra, ddof=1)  # Usamos ddof=1 para obtener la desviaci√≥n est√°ndar muestral

# Nivel de confianza del 95%
nivel_confianza = 0.95

# Calcular el intervalo de confianza
intervalo_confianza = stats.norm.interval(nivel_confianza, loc=media_muestral, scale=desviacion_estandar_muestral/np.sqrt(len(muestra)))

# Visualizaci√≥n en gr√°fico
plt.hist(muestra, bins=20, alpha=0.7, color='skyblue', edgecolor='black', label='Muestra')
plt.axvline(media_muestral, color='red', linestyle='dashed', linewidth=2, label='Media Muestral')
plt.axvline(intervalo_confianza[0], color='green', linestyle='dashed', linewidth=2, label='L√≠mite Inferior IC')
plt.axvline(intervalo_confianza[1], color='green', linestyle='dashed', linewidth=2, label='L√≠mite Superior IC')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.title('Intervalo de Confianza para la Media')
plt.legend()
plt.show()

print(f"Media Muestral: {media_muestral}")
print(f"Intervalo de Confianza del {nivel_confianza * 100}%: ({intervalo_confianza[0]}, {intervalo_confianza[1]})")
```

En este c√≥digo, se genera una muestra de una distribuci√≥n normal, se calcula la media muestral y la desviaci√≥n est√°ndar muestral, y se construye un intervalo de confianza del 95% para la media poblacional. La gr√°fica muestra la distribuci√≥n de la muestra con la media muestral y los l√≠mites del intervalo de confianza.

### **Tama√±o de muestra y precisi√≥n**

**Tama√±o de Muestra:**

El tama√±o de muestra $(n)$ se refiere al n√∫mero de observaciones o elementos en una muestra tomada de una poblaci√≥n. En el contexto de los intervalos de confianza, el tama√±o de muestra juega un papel crucial en la estimaci√≥n de la precisi√≥n de la estimaci√≥n puntual. Un tama√±o de muestra m√°s grande generalmente resulta en intervalos de confianza m√°s estrechos y una mayor precisi√≥n en la estimaci√≥n del par√°metro poblacional.

**Precisi√≥n en Intervalos de Confianza:**

La precisi√≥n en un intervalo de confianza se relaciona con la amplitud del intervalo. Un intervalo m√°s estrecho indica una mayor precisi√≥n, ya que estamos siendo m√°s espec√≠ficos acerca de la ubicaci√≥n probable del par√°metro poblacional. Por el contrario, un intervalo m√°s amplio indica una menor precisi√≥n y m√°s incertidumbre acerca de la ubicaci√≥n del par√°metro.

**Ejemplo y C√≥digo Python:**

Vamos a generar intervalos de confianza para la media de una poblaci√≥n con diferentes tama√±os de muestra y analizar la precisi√≥n.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Configuraci√≥n de la simulaci√≥n
np.random.seed(42)
media_poblacional = 10
desviacion_poblacional = 2
tamanos_muestra = [30, 50, 100]
nivel_confianza = 0.95

# Simulaci√≥n de intervalos de confianza para diferentes tama√±os de muestra
intervalos_confianza = []

for n in tamanos_muestra:
    muestra = np.random.normal(loc=media_poblacional, scale=desviacion_poblacional, size=n)
    media_muestral = np.mean(muestra)
    error_estandar = stats.sem(muestra)
    intervalo_confianza = stats.t.interval(nivel_confianza, df=n-1, loc=media_muestral, scale=error_estandar)
    intervalos_confianza.append(intervalo_confianza)

# Visualizaci√≥n en gr√°fico
plt.figure(figsize=(12, 6))

for i, n in enumerate(tamanos_muestra):
    plt.plot([i+1, i+1], intervalos_confianza[i], marker='o', markersize=8, label=f'Tama√±o Muestra {n}')

plt.axhline(media_poblacional, color='red', linestyle='dashed', linewidth=2, label='Media Poblacional')
plt.xlabel('Tama√±o Muestra')
plt.ylabel('Intervalo de Confianza')
plt.title('Intervalos de Confianza para la Media con Diferentes Tama√±os de Muestra')
plt.legend()
plt.show()
```

En este c√≥digo, simulamos la construcci√≥n de intervalos de confianza para la media de una poblaci√≥n con diferentes tama√±os de muestra. Los intervalos se visualizan en un gr√°fico para comparar c√≥mo cambia la precisi√≥n con diferentes tama√±os de muestra. Ten en cuenta que se utiliza la distribuci√≥n t de Student para tener en cuenta la variabilidad adicional con tama√±os de muestra peque√±os.
