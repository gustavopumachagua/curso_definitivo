# **7. Regresión y Correlación**

## **Regresión Lineal Simple**

### **Modelo de regresión, coeficiente de determinación**

**Modelo de Regresión:**

Un modelo de regresión es una herramienta estadística que se utiliza para examinar la relación entre una variable dependiente (respuesta) y una o más variables independientes (predictores). El objetivo es desarrollar una ecuación matemática que describa la relación entre estas variables.

**Ecuación del Modelo de Regresión Simple:**

$Y = \beta_0 + \beta_1 \cdot X + \epsilon$

- $(Y)$ es la variable dependiente.
- $(X)$ es la variable independiente.
- $(\beta_0)$ es la ordenada al origen (intercepto).
- $(\beta_1)$ es la pendiente.
- $(\epsilon)$ es el error.

**Ecuación del Modelo de Regresión Múltiple:**

$Y = \beta_0 + \beta_1 \cdot X_1 + \beta_2 \cdot X_2 + \ldots + \beta_n \cdot X_n + \epsilon$

**Coeficiente de Determinación $(R^2)$:**

El coeficiente de determinación $(R^2)$ es una medida que indica la proporción de la variabilidad en la variable dependiente que puede explicarse por el modelo de regresión. $(R^2)$ varía entre 0 y 1, donde 0 indica que el modelo no explica ninguna variabilidad y 1 indica que el modelo explica toda la variabilidad.

$R^2 = 1 - \frac{\text{Suma de los cuadrados de los residuos}}{\text{Suma total de los cuadrados}}$

**Ejemplo y Código Python:**

Supongamos que queremos ajustar un modelo de regresión simple y calcular $(R^2)$ utilizando datos ficticios.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Crear datos ficticios
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Ajustar un modelo de regresión lineal
model = LinearRegression()
model.fit(X, y)

# Predicciones del modelo
y_pred = model.predict(X)

# Calcular R^2
r2 = r2_score(y, y_pred)

# Visualización en gráfico
plt.scatter(X, y, alpha=0.7, label='Datos Observados')
plt.plot(X, y_pred, color='red', linewidth=2, label='Modelo de Regresión')
plt.xlabel('Variable Independiente')
plt.ylabel('Variable Dependiente')
plt.title('Modelo de Regresión Lineal y R^2')
plt.legend()
plt.show()

# Imprimir resultados
print(f"Coeficientes del modelo: {model.coef_[0][0]} (pendiente), {model.intercept_[0]} (intercepto)")
print(f"Coeficiente de Determinación (R^2): {r2}")
```

En este código, hemos creado datos ficticios con una relación lineal y ajustado un modelo de regresión lineal. La visualización incluye una dispersión de puntos para los datos observados y la línea roja para el modelo de regresión. Además, imprimimos los coeficientes del modelo y el coeficiente de determinación $(R^2)$. Puedes ajustar este código para incluir múltiples variables independientes para un modelo de regresión múltiple.

### **Interpretación de resultados**

La interpretación de los resultados de un modelo de regresión y el coeficiente de determinación $(R^2)$ es fundamental para entender la relación entre las variables y la calidad del ajuste del modelo. A continuación, proporcionaré una explicación detallada con ejemplos y gráficos utilizando Python.

**Interpretación de Coeficientes del Modelo de Regresión:**

En un modelo de regresión lineal simple, la ecuación del modelo es:

$Y = \beta_0 + \beta_1 \cdot X + \epsilon$

Donde:

- $(\beta_0)$ es la ordenada al origen (intercepto).
- $(\beta_1)$ es la pendiente.
- $(X)$ es la variable independiente.
- $(Y)$ es la variable dependiente.

**Ejemplo en Python:**

```python
import numpy as np
from sklearn.linear_model import LinearRegression

# Crear datos ficticios
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Ajustar un modelo de regresión lineal
model = LinearRegression()
model.fit(X, y)

# Imprimir coeficientes del modelo
print(f"Intercepto (ordenada al origen): {model.intercept_[0]}")
print(f"Pendiente: {model.coef_[0][0]}")
```

En este ejemplo, el intercepto $(\beta_0)$ representa el valor esperado de $(Y)$ cuando $(X)$ es cero, y la pendiente $(\beta_1)$ indica cómo cambia $(Y)$ por unidad de cambio en $(X)$.

**Interpretación del Coeficiente de Determinación $(R^2)$:**

El coeficiente de determinación $(R^2)$ mide la proporción de la variabilidad en la variable dependiente que es explicada por el modelo. Su interpretación varía entre 0 (ninguna explicación) y 1 (explicación completa).

$R^2 = 1 - \frac{\text{Suma de los cuadrados de los residuos}}{\text{Suma total de los cuadrados}}$

**Ejemplo en Python:**

```python
from sklearn.metrics import r2_score

# Predicciones del modelo
y_pred = model.predict(X)

# Calcular R^2
r2 = r2_score(y, y_pred)

# Imprimir coeficiente de determinación (R^2)
print(f"Coeficiente de Determinación (R^2): {r2}")
```

**Visualización con Gráficos Personalizados:**

```python
import matplotlib.pyplot as plt

# Visualización en gráfico
plt.scatter(X, y, alpha=0.7, label='Datos Observados')
plt.plot(X, y_pred, color='red', linewidth=2, label='Modelo de Regresión')
plt.xlabel('Variable Independiente')
plt.ylabel('Variable Dependiente')
plt.title('Modelo de Regresión Lineal y R^2')
plt.legend()
plt.show()
```

En este gráfico, los puntos azules representan los datos observados, y la línea roja es la predicción del modelo de regresión. Este gráfico ayuda a visualizar cómo el modelo se ajusta a los datos observados.

**Interpretación Conjunta:**

- **Coeficientes del Modelo:**

  - El intercepto $(\beta_0)$ puede interpretarse como el valor esperado de la variable dependiente cuando la variable independiente es cero.
  - La pendiente $(\beta_1)$ indica el cambio promedio en la variable dependiente por unidad de cambio en la variable independiente.

- **Coeficiente de Determinación $(R^2)$:**

  - Un $(R^2)$ cercano a 1 indica que el modelo explica una gran proporción de la variabilidad en la variable dependiente.
  - Un $(R^2)$ cercano a 0 sugiere que el modelo no explica mucha variabilidad.

- **Gráfico de Regresión:**

  - Ayuda a visualizar cómo el modelo se ajusta a los datos observados.

La interpretación de estos resultados proporciona información valiosa sobre la relación entre las variables y la capacidad del modelo para explicar la variabilidad en los datos observados.

## **Regresión Múltiple**

### **Extensión del modelo de regresión a múltiples variables independientes**

Cuando trabajamos con modelos de regresión, es común extender el enfoque a múltiples variables independientes, creando así un modelo de regresión múltiple. La ecuación del modelo de regresión múltiple se ve así:

$Y = \beta_0 + \beta_1 \cdot X_1 + \beta_2 \cdot X_2 + \ldots + \beta_n \cdot X_n + \epsilon$

Donde:

- $( Y )$ es la variable dependiente.
- $( X_1, X_2, \ldots, X_n )$ son las variables independientes.
- $( \beta_0 )$ es el término de intersección (ordenada al origen).
- $( \beta_1, \beta_2, \ldots, \beta_n )$ son los coeficientes asociados a las variables independientes.
- $( \epsilon )$ es el término de error.

**Ejemplo con Código Python:**

Supongamos que tenemos un conjunto de datos con dos variables independientes $( X_1 )$ y $( X_2 )$, y queremos ajustar un modelo de regresión múltiple.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Crear datos ficticios con dos variables independientes
np.random.seed(42)
X1 = 2 * np.random.rand(100, 1)
X2 = 3 * np.random.rand(100, 1)
y = 4 + 3 * X1 + 5 * X2 + np.random.randn(100, 1)

# Ajustar un modelo de regresión múltiple
model = LinearRegression()
X_multi = np.column_stack((X1, X2))
model.fit(X_multi, y)

# Obtener coeficientes y realizar predicciones
intercepto = model.intercept_[0]
coeficientes = model.coef_[0]
y_pred = model.predict(X_multi)

# Calcular R^2
r2 = r2_score(y, y_pred)

# Imprimir coeficientes y R^2
print(f"Intercepto (ordenada al origen): {intercepto}")
print(f"Coeficientes: {coeficientes}")
print(f"Coeficiente de Determinación (R^2): {r2}")

# Visualización en gráfico 3D
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(X1, X2, y, color='blue', label='Datos Observados')
ax.scatter(X1, X2, y_pred, color='red', label='Predicciones del Modelo', marker='^')

ax.set_xlabel('Variable Independiente 1')
ax.set_ylabel('Variable Independiente 2')
ax.set_zlabel('Variable Dependiente')
ax.set_title('Modelo de Regresión Múltiple')

plt.legend()
plt.show()
```


En este ejemplo, hemos creado un conjunto de datos con dos variables independientes $(X_1)$ y $(X_2)$ y ajustado un modelo de regresión múltiple. La visualización incluye un gráfico 3D para mostrar cómo el modelo se ajusta a los datos observados y predice valores en función de dos variables independientes.

**Interpretación de Resultados:**

- **Coeficientes del Modelo:**

  - Cada coeficiente $(\beta_1, \beta_2)$ representa el cambio promedio en la variable dependiente por unidad de cambio en la variable independiente correspondiente, manteniendo las otras variables constantes.

- **Coeficiente de Determinación $(R^2)$:**

  - Indica la proporción de la variabilidad en la variable dependiente que es explicada por el modelo de regresión múltiple.

- **Visualización:**

  - El gráfico 3D proporciona una representación visual del modelo y cómo se ajusta a los datos.

La extensión a múltiples variables independientes permite modelar relaciones más complejas y capturar la variabilidad asociada a múltiples factores. La interpretación de los coeficientes sigue siendo similar, pero ahora consideramos el impacto de cada variable independiente controlando las demás.
