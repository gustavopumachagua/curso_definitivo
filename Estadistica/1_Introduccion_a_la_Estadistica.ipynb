# **1. Introducci√≥n a la Estad√≠stica**

## **Conceptos B√°sicos de Estad√≠stica**

### **Definiciones clave (poblaci√≥n, muestra, variable, etc.)**

Vamos a explorar las definiciones clave en estad√≠stica: poblaci√≥n, muestra, y variable, y c√≥mo puedes representarlas visualmente en Python utilizando gr√°ficos personalizados. Utilizaremos las bibliotecas `matplotlib` y `numpy` para crear visualizaciones simples.

Primero, aseg√∫rate de tener las bibliotecas instaladas:

```bash
pip install matplotlib numpy
```

Ahora, aqu√≠ tienes un ejemplo con explicaciones detalladas:

````python
import numpy as np
import matplotlib.pyplot as plt

# Definir una poblaci√≥n
np.random.seed(42)  # Para reproducibilidad
poblacion = np.random.normal(loc=50, scale=10, size=10000)  # Distribuci√≥n normal, media=50, desviaci√≥n est√°ndar=10

# Graficar la poblaci√≥n con personalizaciones
plt.figure(figsize=(10, 6))
plt.hist(poblacion, bins=30, color='skyblue', edgecolor='black', alpha=0.7)

# L√≠nea vertical para la media
media_poblacion = np.mean(poblacion)
plt.axvline(x=media_poblacion, color='red', linestyle='dashed', linewidth=2, label=f'Media = {media_poblacion:.2f}')

# L√≠nea vertical para la desviaci√≥n est√°ndar
std_dev_poblacion = np.std(poblacion)
plt.axvline(x=media_poblacion + std_dev_poblacion, color='orange', linestyle='dashed', linewidth=2, label=f'SD = {std_dev_poblacion:.2f}')
plt.axvline(x=media_poblacion - std_dev_poblacion, color='orange', linestyle='dashed', linewidth=2)

# T√≠tulo y etiquetas del eje
plt.title('Distribuci√≥n de la Poblaci√≥n - Media y Desviaci√≥n Est√°ndar', fontsize=16)
plt.xlabel('Valor', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)

# Leyenda
plt.legend()

# Mostrar la gr√°fica
plt.show()
````

En este ejemplo:

- **Poblaci√≥n:** Hemos creado una poblaci√≥n de 10,000 elementos que sigue una distribuci√≥n normal con una media de 50 y una desviaci√≥n est√°ndar de 10. La poblaci√≥n es el conjunto completo de datos que queremos estudiar.

- **Muestra:** Ahora, tomaremos una muestra de esta poblaci√≥n. Utilizaremos solo 500 elementos de la poblaci√≥n para formar nuestra muestra.

````python
# Tomar una muestra de la poblaci√≥n
muestra = np.random.choice(poblacion, size=500)

# Graficar la muestra junto con la poblaci√≥n con personalizaciones
plt.figure(figsize=(12, 6))
plt.hist(poblacion, bins=30, color='skyblue', alpha=0.5, label='Poblaci√≥n')
plt.hist(muestra, bins=30, color='orange', alpha=0.7, label='Muestra')

# T√≠tulo y etiquetas del eje
plt.title('Comparaci√≥n entre Poblaci√≥n y Muestra', fontsize=16)
plt.xlabel('Valor', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)

# Leyenda
plt.legend()

# Mostrar la gr√°fica
plt.show()
````

En esta visualizaci√≥n, puedes ver c√≥mo la muestra (en naranja) es un subconjunto m√°s peque√±o pero a√∫n representa la forma general de la poblaci√≥n.

- **Variable:** Ahora, introduzcamos una variable. Supongamos que estamos interesados en la altura de las personas en nuestra poblaci√≥n y muestra.

````python
# Generar datos de altura para la poblaci√≥n y la muestra
altura_poblacion = np.random.normal(loc=170, scale=10, size=10000)
altura_muestra = np.random.choice(altura_poblacion, size=500)

# Graficar las distribuciones de altura con personalizaciones
plt.figure(figsize=(12, 6))
plt.hist(altura_poblacion, bins=30, color='skyblue', alpha=0.5, label='Altura en Poblaci√≥n')
plt.hist(altura_muestra, bins=30, color='orange', alpha=0.7, label='Altura en Muestra')

# L√≠nea vertical para la media de la poblaci√≥n
media_altura_poblacion = np.mean(altura_poblacion)
plt.axvline(x=media_altura_poblacion, color='blue', linestyle='dashed', linewidth=2, label=f'Media Poblaci√≥n = {media_altura_poblacion:.2f}')

# L√≠nea vertical para la media de la muestra
media_altura_muestra = np.mean(altura_muestra)
plt.axvline(x=media_altura_muestra, color='orange', linestyle='dashed', linewidth=2, label=f'Media Muestra = {media_altura_muestra:.2f}')

# T√≠tulo y etiquetas del eje
plt.title('Distribuci√≥n de Altura en Poblaci√≥n y Muestra', fontsize=16)
plt.xlabel('Altura (cm)', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)

# Leyenda
plt.legend()

# Mostrar la gr√°fica
plt.show()
````

En este caso, hemos introducido la variable "altura", y graficamos c√≥mo se distribuye en la poblaci√≥n y la muestra.

Estos son ejemplos simples para comprender las definiciones clave en estad√≠stica. La visualizaci√≥n de datos es esencial para entender las caracter√≠sticas de la poblaci√≥n y la muestra, as√≠ como la distribuci√≥n de las variables de inter√©s.

### **Tipos de datos: cualitativos y cuantitativos**

En estad√≠stica, los datos se pueden clasificar en dos tipos principales: cualitativos y cuantitativos. Vamos a explorar cada uno de ellos con ejemplos y gr√°ficos personalizados utilizando Python.

**1. Datos Cualitativos:**

Los datos cualitativos representan atributos no num√©ricos y se dividen en dos subtipos: nominales y ordinales.

**a. Datos Nominales:**

Los datos nominales representan categor√≠as con nombres √∫nicos, pero no hay un orden espec√≠fico entre ellas.

**Ejemplo:** Colores de los autom√≥viles.

````python
import matplotlib.pyplot as plt

# Datos nominales
colores = ['red', 'blue', 'green', 'yellow', 'black']
conteo_colores = [25, 20, 15, 10, 30]

# Colores personalizados para las barras
colores_barras = ['lightcoral', 'royalblue', 'lightgreen', 'gold', 'gray']

# Graficar un gr√°fico de barras para datos nominales con personalizaciones
plt.figure(figsize=(10, 6))
plt.bar(colores, conteo_colores, color=colores_barras, edgecolor='black', linewidth=1.2)

# T√≠tulo y etiquetas del eje
plt.title('Distribuci√≥n de Colores de Autom√≥viles', fontsize=16)
plt.xlabel('Colores', fontsize=12)
plt.ylabel('Cantidad', fontsize=12)

# Mostrar la gr√°fica
plt.show()
````

**b. Datos Ordinales:**

Los datos ordinales tienen categor√≠as con un orden espec√≠fico, pero la distancia entre ellas no es cuantificable.

**Ejemplo:** Clasificaci√≥n de productos por calidad (Baja, Media, Alta).

````python
import matplotlib.pyplot as plt

# Datos ordinales
productos = ['Baja', 'Media', 'Alta']
conteo_calidad = [15, 30, 20]

# Colores personalizados para las barras
colores_barras = ['lightcoral', 'gold', 'lightgreen']

# Graficar un gr√°fico de barras para datos ordinales con personalizaciones
plt.figure(figsize=(10, 6))
plt.bar(productos, conteo_calidad, color=colores_barras, edgecolor='black', linewidth=1.2)

# T√≠tulo y etiquetas del eje
plt.title('Clasificaci√≥n de Productos por Calidad', fontsize=16)
plt.xlabel('Calidad', fontsize=12)
plt.ylabel('Cantidad', fontsize=12)

# Mostrar la gr√°fica
plt.show()
````

**2. Datos Cuantitativos:**

Los datos cuantitativos representan cantidades num√©ricas y se dividen en dos subtipos: discretos y continuos.

**a. Datos Discretos:**

Los datos discretos son valores num√©ricos contables y espec√≠ficos.

**Ejemplo:** N√∫mero de estudiantes en una clase.

````python
import matplotlib.pyplot as plt

# Datos discretos
num_estudiantes = [20, 25, 30, 35, 40]
grados = ['Primero', 'Segundo', 'Tercero', 'Cuarto', 'Quinto']

# Colores personalizados para las barras
colores_barras = ['lightblue', 'royalblue', 'cornflowerblue', 'steelblue', 'navy']

# Graficar un gr√°fico de barras para datos discretos con personalizaciones
plt.figure(figsize=(10, 6))
plt.bar(grados, num_estudiantes, color=colores_barras, edgecolor='black', linewidth=1.2)

# T√≠tulo y etiquetas del eje
plt.title('N√∫mero de Estudiantes por Grado', fontsize=16)
plt.xlabel('Grados', fontsize=12)
plt.ylabel('N√∫mero de Estudiantes', fontsize=12)

# Mostrar la gr√°fica
plt.show()
````

**b. Datos Continuos:**

Los datos continuos pueden tomar cualquier valor dentro de un rango y son medibles.

**Ejemplo:** Altura de los estudiantes.

````python
import numpy as np
import matplotlib.pyplot as plt

# Datos continuos
altura_estudiantes = np.random.normal(loc=170, scale=5, size=100)

# Graficar un histograma para datos continuos con personalizaciones
plt.figure(figsize=(10, 6))
plt.hist(altura_estudiantes, bins=20, color='mediumorchid', edgecolor='black', alpha=0.7)

# L√≠nea vertical para la media
media_altura_estudiantes = np.mean(altura_estudiantes)
plt.axvline(x=media_altura_estudiantes, color='darkorange', linestyle='dashed', linewidth=2, label=f'Media = {media_altura_estudiantes:.2f}')

# T√≠tulo y etiquetas del eje
plt.title('Distribuci√≥n de Altura de Estudiantes', fontsize=16)
plt.xlabel('Altura (cm)', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)

# Leyenda
plt.legend()

# Mostrar la gr√°fica
plt.show()
````

Estos ejemplos y gr√°ficos proporcionan una introducci√≥n a los tipos de datos cualitativos y cuantitativos, y c√≥mo visualizarlos en Python. La elecci√≥n entre gr√°ficos de barras y histogramas depende del tipo de datos que est√©s representando.

## **Descriptiva vs. Inferencial**

### **Diferencias entre estad√≠sticas descriptivas e inferenciales**

Las estad√≠sticas descriptivas e inferenciales son dos ramas principales de la estad√≠stica que se utilizan para analizar y comprender datos. A continuaci√≥n, explicar√© las diferencias entre ambas y proporcionar√© ejemplos con gr√°ficos en c√≥digo Python.

**Estad√≠sticas Descriptivas:**

Las estad√≠sticas descriptivas se centran en describir y resumir las caracter√≠sticas fundamentales de un conjunto de datos. Estas medidas proporcionan informaci√≥n sobre la tendencia central, la dispersi√≥n y la forma de la distribuci√≥n de los datos.

Ejemplo en Python:

````python
import numpy as np
import matplotlib.pyplot as plt

# Crear un conjunto de datos aleatorio
datos = np.random.normal(loc=50, scale=10, size=1000)

# Estad√≠sticas descriptivas
media = np.mean(datos)
mediana = np.median(datos)
desviacion_estandar = np.std(datos)

# Graficar un histograma con personalizaciones
plt.figure(figsize=(10, 6))
plt.hist(datos, bins=30, color='skyblue', edgecolor='black', alpha=0.7)

# L√≠nea vertical para la media
plt.axvline(x=media, color='orange', linestyle='dashed', linewidth=2, label=f'Media = {media:.2f}')

# L√≠nea vertical para la mediana
plt.axvline(x=mediana, color='green', linestyle='dashed', linewidth=2, label=f'Mediana = {mediana:.2f}')

# T√≠tulo y etiquetas del eje
plt.title('Histograma de Datos Aleatorios', fontsize=16)
plt.xlabel('Valor', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)

# Leyenda
plt.legend()

# Mostrar la gr√°fica
plt.show()

# Imprimir estad√≠sticas descriptivas
print(f'Media: {media}')
print(f'Mediana: {mediana}')
print(f'Desviaci√≥n Est√°ndar: {desviacion_estandar}')
````

En este ejemplo, se genera un conjunto de datos aleatorio y se calculan la media, la mediana y la desviaci√≥n est√°ndar. El histograma proporciona una visualizaci√≥n de la distribuci√≥n de los datos.

**Estad√≠sticas Inferenciales:**

Las estad√≠sticas inferenciales se utilizan para hacer inferencias o generalizaciones sobre una poblaci√≥n bas√°ndose en una muestra de datos. Esto implica utilizar m√©todos como pruebas de hip√≥tesis y estimaci√≥n de intervalos de confianza.

Ejemplo en Python:

````python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Crear dos muestras para comparar
muestra1 = np.random.normal(loc=50, scale=10, size=100)
muestra2 = np.random.normal(loc=55, scale=10, size=100)

# Prueba t de dos muestras
t_stat, p_valor = stats.ttest_ind(muestra1, muestra2)

# Boxplot para visualizar las muestras con personalizaciones
plt.figure(figsize=(10, 6))
plt.boxplot([muestra1, muestra2], labels=['Muestra 1', 'Muestra 2'], patch_artist=True, boxprops=dict(facecolor='lightblue', color='black'), medianprops=dict(color='orange'))

# L√≠nea horizontal para el valor medio de la muestra 1
plt.axhline(y=np.mean(muestra1), color='blue', linestyle='dashed', linewidth=2, label=f'Media Muestra 1 = {np.mean(muestra1):.2f}')

# L√≠nea horizontal para el valor medio de la muestra 2
plt.axhline(y=np.mean(muestra2), color='red', linestyle='dashed', linewidth=2, label=f'Media Muestra 2 = {np.mean(muestra2):.2f}')

# T√≠tulo y etiquetas del eje
plt.title('Comparaci√≥n de Muestras', fontsize=16)
plt.ylabel('Valor', fontsize=12)

# Leyenda
plt.legend()

# Mostrar la gr√°fica
plt.show()

# Imprimir resultados de la prueba t
print(f'Estad√≠stico t: {t_stat}')
print(f'Valor p: {p_valor}')
````

En este ejemplo, se comparan dos muestras utilizando una prueba t de dos muestras. El boxplot ayuda a visualizar las diferencias entre las muestras.

En resumen, las estad√≠sticas descriptivas se utilizan para resumir y describir datos, mientras que las estad√≠sticas inferenciales se utilizan para hacer inferencias sobre poblaciones bas√°ndose en muestras. Ambas son fundamentales para comprender y analizar conjuntos de datos.

### **Aplicaciones pr√°cticas**

Vamos a explorar aplicaciones pr√°cticas de estad√≠sticas descriptivas e inferenciales utilizando ejemplos con gr√°ficos en Python.

**1. Estad√≠sticas Descriptivas: An√°lisis de Ventas Mensuales**

Supongamos que tenemos datos mensuales de ventas de una tienda. Queremos utilizar estad√≠sticas descriptivas para entender la tendencia y la variabilidad de las ventas.

````python
import numpy as np
import matplotlib.pyplot as plt

# Crear datos de ventas mensuales
ventas_mensuales = np.random.normal(loc=50000, scale=10000, size=12)

# Estad√≠sticas descriptivas
media_ventas = np.mean(ventas_mensuales)
desviacion_ventas = np.std(ventas_mensuales)

# Definir los meses
meses = ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun', 'Jul', 'Ago', 'Sep', 'Oct', 'Nov', 'Dic']

# Gr√°fico de barras con personalizaciones
plt.figure(figsize=(10, 6))
plt.bar(meses, ventas_mensuales, color='skyblue', edgecolor='black')

# L√≠nea horizontal para la media de las ventas
plt.axhline(y=media_ventas, color='orange', linestyle='dashed', linewidth=2, label=f'Media de Ventas = {media_ventas:.2f}')

# T√≠tulo y etiquetas del eje
plt.title('Ventas Mensuales', fontsize=16)
plt.xlabel('Mes', fontsize=12)
plt.ylabel('Ventas', fontsize=12)

# Leyenda
plt.legend()

# Mostrar la gr√°fica
plt.show()

# Imprimir estad√≠sticas descriptivas
print(f'Media de Ventas: {media_ventas}')
print(f'Desviaci√≥n Est√°ndar de Ventas: {desviacion_ventas}')
````

En este ejemplo, utilizamos estad√≠sticas descriptivas para calcular la media y la desviaci√≥n est√°ndar de las ventas mensuales. El gr√°fico de barras muestra la variaci√≥n en las ventas a lo largo de los meses.

**2. Estad√≠sticas Inferenciales: Prueba A/B en una P√°gina Web**

Supongamos que una empresa realiza cambios en su p√°gina web y quiere evaluar si el nuevo dise√±o aumenta la tasa de clics en un bot√≥n espec√≠fico en comparaci√≥n con el dise√±o anterior.

````python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Datos de la tasa de clics para el dise√±o anterior y nuevo
dise√±o_anterior = np.random.normal(loc=0.1, scale=0.02, size=1000)
dise√±o_nuevo = np.random.normal(loc=0.12, scale=0.02, size=1000)

# Prueba t para comparar las tasas de clics
t_stat, p_valor = stats.ttest_ind(dise√±o_anterior, dise√±o_nuevo)

# Boxplot con personalizaciones
plt.figure(figsize=(10, 6))
plt.boxplot([dise√±o_anterior, dise√±o_nuevo], labels=['Dise√±o Anterior', 'Nuevo Dise√±o'], patch_artist=True, boxprops=dict(facecolor='lightblue', color='black'), medianprops=dict(color='orange'))

# L√≠nea horizontal para la media del dise√±o anterior
plt.axhline(y=np.mean(dise√±o_anterior), color='blue', linestyle='dashed', linewidth=2, label=f'Media Dise√±o Anterior = {np.mean(dise√±o_anterior):.2f}')

# L√≠nea horizontal para la media del nuevo dise√±o
plt.axhline(y=np.mean(dise√±o_nuevo), color='red', linestyle='dashed', linewidth=2, label=f'Media Nuevo Dise√±o = {np.mean(dise√±o_nuevo):.2f}')

# T√≠tulo y etiquetas del eje
plt.title('Comparaci√≥n de Tasas de Clics', fontsize=16)
plt.ylabel('Tasa de Clics', fontsize=12)

# Leyenda
plt.legend()

# Mostrar la gr√°fica
plt.show()

# Imprimir resultados de la prueba t
print(f'Estad√≠stico t: {t_stat}')
print(f'Valor p: {p_valor}')
````

En este ejemplo, utilizamos una prueba t para comparar las tasas de clics entre el dise√±o anterior y el nuevo. El boxplot muestra las distribuciones y las diferencias entre los dos dise√±os.

Estas aplicaciones representan solo dos ejemplos de c√≥mo se pueden utilizar las estad√≠sticas descriptivas e inferenciales en situaciones pr√°cticas para comprender datos y tomar decisiones informadas.
