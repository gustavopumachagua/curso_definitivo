### **Tokenizaci贸n, lematizaci贸n**

**Tokenizaci贸n:**

La tokenizaci贸n es el proceso de dividir un texto en unidades m谩s peque帽as llamadas "tokens". Un token puede ser una palabra, una frase, o incluso un car谩cter, dependiendo de la granularidad deseada. Los tokens son las unidades b谩sicas que se utilizan para el an谩lisis de texto.

**Ejemplo de Tokenizaci贸n con NLTK en Python:**

```python
import nltk
from nltk.tokenize import word_tokenize

nltk.download('punkt')  # Descargar datos necesarios para NLTK

# Texto de ejemplo
texto = "La tokenizaci贸n es una parte esencial del procesamiento de lenguaje natural."

# Tokenizaci贸n
tokens = word_tokenize(texto)
print(tokens)
```

En este ejemplo, el texto se tokeniza en palabras usando la funci贸n `word_tokenize` de NLTK. Los tokens resultantes son una lista de palabras en el texto.

**Lematizaci贸n:**

La lematizaci贸n es el proceso de reducir las palabras a su forma base, conocida como lema. Un lema es la forma can贸nica de una palabra, y la lematizaci贸n ayuda a reducir las palabras a su forma ra铆z.

**Ejemplo de Lematizaci贸n con NLTK en Python:**

En este ejemplo, el texto se lematiza utilizando el lematizador de NLTK. Cada palabra se reduce a su forma base (lema), lo que ayuda a simplificar el an谩lisis del texto.

### **Modelos de bolsa de palabras y modelos de lenguaje**

**Modelo de Bolsa de Palabras (Bag of Words - BoW):**

El Modelo de Bolsa de Palabras es una representaci贸n simplificada de un documento o conjunto de documentos donde se ignora la estructura gramatical y el orden de las palabras, centr谩ndose solo en la frecuencia de aparici贸n de las palabras. En este modelo, cada documento se representa como un vector, y cada palabra 煤nica en el conjunto de datos se convierte en una dimensi贸n en ese vector.

**Ejemplo de Bolsa de Palabras con CountVectorizer en Python:**

```python
from sklearn.feature_extraction.text import CountVectorizer
import matplotlib.pyplot as plt

# Datos de ejemplo
documentos = [
    "La inteligencia artificial es emocionante.",
    "La inteligencia artificial es el futuro.",
    "Los modelos de lenguaje son fundamentales para NLP."
]

# Crear un vectorizador de bolsa de palabras
vectorizador = CountVectorizer()

# Transformar los documentos en una matriz de t茅rminos-frecuencias
matriz_bolsa_palabras = vectorizador.fit_transform(documentos)

# Obtener las palabras del vocabulario
vocabulario = vectorizador.get_feature_names_out()

# Visualizar la matriz de bolsa de palabras
plt.imshow(matriz_bolsa_palabras.toarray(), cmap='viridis', aspect='auto', extent=(0, len(vocabulario), len(documentos), 0))
plt.colorbar()
plt.title('Matriz de Bolsa de Palabras')
plt.xlabel('Palabras (Vocabulario)')
plt.ylabel('Documentos')
plt.xticks(range(len(vocabulario)), vocabulario, rotation=45, ha='right')
plt.show()
```

En este ejemplo, utilizamos `CountVectorizer` de scikit-learn para crear la matriz de bolsa de palabras. La matriz resultante muestra la frecuencia de cada palabra en cada documento. Cada fila representa un documento, y cada columna representa una palabra 煤nica en el conjunto de datos.

**Modelo de Lenguaje:**

Un Modelo de Lenguaje es una representaci贸n probabil铆stica de la secuencia de palabras en un idioma. Estos modelos asignan probabilidades a las secuencias de palabras, lo que permite predecir la probabilidad de que una determinada secuencia sea v谩lida en el idioma dado. Pueden utilizarse para tareas como la generaci贸n de texto, correcci贸n ortogr谩fica, y traducci贸n autom谩tica.

**Ejemplo de Modelo de Lenguaje con Markov en Python:**

```python
import random

# Datos de ejemplo
corpus = "La inteligencia artificial es el futuro. La inteligencia artificial es emocionante."

# Crear un modelo de lenguaje de primer orden (Markov)
def entrenar_modelo(corpus):
    palabras = corpus.split()
    modelo = {}
    for i in range(len(palabras) - 1):
        palabra_actual = palabras[i]
        palabra_siguiente = palabras[i + 1]
        if palabra_actual not in modelo:
            modelo[palabra_actual] = []
        modelo[palabra_actual].append(palabra_siguiente)
    return modelo

# Generar texto usando el modelo entrenado
def generar_texto(modelo, semilla, longitud):
    texto_generado = semilla
    palabra_actual = semilla
    for _ in range(longitud - 1):
        if palabra_actual in modelo:
            palabra_siguiente = random.choice(modelo[palabra_actual])
            texto_generado += " " + palabra_siguiente
            palabra_actual = palabra_siguiente
        else:
            break
    return texto_generado

# Entrenar el modelo de lenguaje
modelo_lenguaje = entrenar_modelo(corpus)

# Generar texto utilizando el modelo
texto_generado = generar_texto(modelo_lenguaje, semilla="La", longitud=10)
print(texto_generado)
```

En este ejemplo, utilizamos un modelo de lenguaje de primer orden basado en cadenas de Markov. El modelo se entrena en un corpus y luego se utiliza para generar texto de manera probabil铆stica. Este modelo es simple y solo considera la probabilidad de la palabra siguiente dada la palabra actual. Modelos m谩s avanzados pueden considerar contextos m谩s largos y tener en cuenta la sintaxis y sem谩ntica de las palabras.

1. **An谩lisis de Sentimientos:**

El an谩lisis de sentimientos se utiliza para determinar la polaridad emocional de un texto, clasific谩ndolo como positivo, negativo o neutro. Esto es 煤til en redes sociales, comentarios de productos, revisiones, etc.

2. **Clasificaci贸n de Texto:**

La clasificaci贸n de texto se aplica para asignar categor铆as o etiquetas a un texto bas谩ndose en su contenido.

**Ejemplo con scikit-learn:**

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, confusion_matrix

# Datos de ejemplo
documentos = ["Esto es un texto positivo.", "Esto es un texto negativo.", "Este texto es neutro."]

# Etiquetas correspondientes
etiquetas = ["Positivo", "Negativo", "Neutro"]

# Crear un vectorizador de bolsa de palabras
vectorizador = CountVectorizer()
matriz_bolsa_palabras = vectorizador.fit_transform(documentos)

# Dividir los datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(matriz_bolsa_palabras, etiquetas, test_size=0.2, random_state=42)

# Crear un clasificador Naive Bayes
clasificador = MultinomialNB()
clasificador.fit(X_train, y_train)

# Hacer predicciones en el conjunto de prueba
y_pred = clasificador.predict(X_test)

# Calcular la precisi贸n del modelo
precision = accuracy_score(y_test, y_pred)
print(f"Precisi贸n del modelo: {precision:.2f}")

# Crear una matriz de confusi贸n
cm = confusion_matrix(y_test, y_pred)
print("Matriz de Confusi贸n:")
print(cm)
```

Este c贸digo utiliza un clasificador Naive Bayes para clasificar texto en categor铆as predefinidas.
