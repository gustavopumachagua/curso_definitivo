**Procesamiento de Texto:**

El procesamiento de texto implica la manipulación de texto para realizar tareas como tokenización, eliminación de stop words, lematización, entre otras.

**Ejemplo de Tokenización con NLTK:**

El análisis de texto involucra la extracción de información significativa del texto. Puede incluir tareas como análisis de sentimientos, extracción de entidades, y clasificación de texto.

**Ejemplo de Análisis de Sentimientos con NLTK:**

### **Tokenización, lematización**

**Tokenización:**

La tokenización es el proceso de dividir un texto en unidades más pequeñas llamadas "tokens". Un token puede ser una palabra, una frase, o incluso un carácter, dependiendo de la granularidad deseada. Los tokens son las unidades básicas que se utilizan para el análisis de texto.

**Ejemplo de Tokenización con NLTK en Python:**

```python
import nltk
from nltk.tokenize import word_tokenize

nltk.download('punkt')  # Descargar datos necesarios para NLTK

# Texto de ejemplo
texto = "La tokenización es una parte esencial del procesamiento de lenguaje natural."

# Tokenización
tokens = word_tokenize(texto)
print(tokens)
```

En este ejemplo, el texto se tokeniza en palabras usando la función `word_tokenize` de NLTK. Los tokens resultantes son una lista de palabras en el texto.

---

**Lematización:**

La lematización es el proceso de reducir las palabras a su forma base, conocida como lema. Un lema es la forma canónica de una palabra, y la lematización ayuda a reducir las palabras a su forma raíz.

**Ejemplo de Lematización con NLTK en Python:**

```python
from nltk.stem import WordNetLemmatizer

# Texto de ejemplo
texto = "Los científicos están investigando nuevos métodos."

# Lematización
lematizador = WordNetLemmatizer()
lemas = [lematizador.lemmatize(palabra) for palabra in word_tokenize(texto)]
print(lemas)
```

En este ejemplo, el texto se lematiza utilizando el lematizador de NLTK. Cada palabra se reduce a su forma base (lema), lo que ayuda a simplificar el análisis del texto.

---

**Visualización de Tokens y Lemas:**

Vamos a visualizar los tokens y lemas utilizando gráficos de barras para observar la diferencia.

```python
import matplotlib.pyplot as plt
from collections import Counter

# Crear un gráfico de barras para tokens
contador_tokens = Counter(tokens)
plt.bar(contador_tokens.keys(), contador_tokens.values())
plt.title('Frecuencia de Tokens')
plt.xlabel('Tokens')
plt.ylabel('Frecuencia')
plt.show()

# Crear un gráfico de barras para lemas
contador_lemas = Counter(lemas)
plt.bar(contador_lemas.keys(), contador_lemas.values())
plt.title('Frecuencia de Lemas')
plt.xlabel('Lemas')
plt.ylabel('Frecuencia')
plt.show()
```

Estos gráficos de barras visualizan la frecuencia de tokens y lemas en el texto. Puedes observar cómo la lematización reduce la variedad de palabras, lo que puede ser útil para análisis más precisos.

### **Modelos de bolsa de palabras y modelos de lenguaje**

**Modelo de Bolsa de Palabras (Bag of Words - BoW):**

El Modelo de Bolsa de Palabras es una representación simplificada de un documento o conjunto de documentos donde se ignora la estructura gramatical y el orden de las palabras, centrándose solo en la frecuencia de aparición de las palabras. En este modelo, cada documento se representa como un vector, y cada palabra única en el conjunto de datos se convierte en una dimensión en ese vector.

**Ejemplo de Bolsa de Palabras con CountVectorizer en Python:**

```python
from sklearn.feature_extraction.text import CountVectorizer
import matplotlib.pyplot as plt

# Datos de ejemplo
documentos = [
    "La inteligencia artificial es emocionante.",
    "La inteligencia artificial es el futuro.",
    "Los modelos de lenguaje son fundamentales para NLP."
]

# Crear un vectorizador de bolsa de palabras
vectorizador = CountVectorizer()

# Transformar los documentos en una matriz de términos-frecuencias
matriz_bolsa_palabras = vectorizador.fit_transform(documentos)

# Obtener las palabras del vocabulario
vocabulario = vectorizador.get_feature_names_out()

# Visualizar la matriz de bolsa de palabras
plt.imshow(matriz_bolsa_palabras.toarray(), cmap='viridis', aspect='auto', extent=(0, len(vocabulario), len(documentos), 0))
plt.colorbar()
plt.title('Matriz de Bolsa de Palabras')
plt.xlabel('Palabras (Vocabulario)')
plt.ylabel('Documentos')
plt.xticks(range(len(vocabulario)), vocabulario, rotation=45, ha='right')
plt.show()
```

En este ejemplo, utilizamos `CountVectorizer` de scikit-learn para crear la matriz de bolsa de palabras. La matriz resultante muestra la frecuencia de cada palabra en cada documento. Cada fila representa un documento, y cada columna representa una palabra única en el conjunto de datos.

---

**Modelo de Lenguaje:**

Un Modelo de Lenguaje es una representación probabilística de la secuencia de palabras en un idioma. Estos modelos asignan probabilidades a las secuencias de palabras, lo que permite predecir la probabilidad de que una determinada secuencia sea válida en el idioma dado. Pueden utilizarse para tareas como la generación de texto, corrección ortográfica, y traducción automática.

**Ejemplo de Modelo de Lenguaje con Markov en Python:**

```python
import random

# Datos de ejemplo
corpus = "La inteligencia artificial es el futuro. La inteligencia artificial es emocionante."

# Crear un modelo de lenguaje de primer orden (Markov)
def entrenar_modelo(corpus):
    palabras = corpus.split()
    modelo = {}
    for i in range(len(palabras) - 1):
        palabra_actual = palabras[i]
        palabra_siguiente = palabras[i + 1]
        if palabra_actual not in modelo:
            modelo[palabra_actual] = []
        modelo[palabra_actual].append(palabra_siguiente)
    return modelo

# Generar texto usando el modelo entrenado
def generar_texto(modelo, semilla, longitud):
    texto_generado = semilla
    palabra_actual = semilla
    for _ in range(longitud - 1):
        if palabra_actual in modelo:
            palabra_siguiente = random.choice(modelo[palabra_actual])
            texto_generado += " " + palabra_siguiente
            palabra_actual = palabra_siguiente
        else:
            break
    return texto_generado

# Entrenar el modelo de lenguaje
modelo_lenguaje = entrenar_modelo(corpus)

# Generar texto utilizando el modelo
texto_generado = generar_texto(modelo_lenguaje, semilla="La", longitud=10)
print(texto_generado)
```

En este ejemplo, utilizamos un modelo de lenguaje de primer orden basado en cadenas de Markov. El modelo se entrena en un corpus y luego se utiliza para generar texto de manera probabilística. Este modelo es simple y solo considera la probabilidad de la palabra siguiente dada la palabra actual. Modelos más avanzados pueden considerar contextos más largos y tener en cuenta la sintaxis y semántica de las palabras.

## **Análisis de Sentimientos**

### **Aplicaciones prácticas de NLP**

El Procesamiento de Lenguaje Natural (NLP) tiene diversas aplicaciones prácticas en el mundo real. A continuación, exploraremos algunas de ellas con ejemplos y gráficos en código Python.

1. **Análisis de Sentimientos:**

El análisis de sentimientos se utiliza para determinar la polaridad emocional de un texto, clasificándolo como positivo, negativo o neutro. Esto es útil en redes sociales, comentarios de productos, revisiones, etc.

**Ejemplo con Visualización:**

```python
from nltk.sentiment import SentimentIntensityAnalyzer
import matplotlib.pyplot as plt

# Texto de ejemplo
texto_sentimiento = "Me encanta trabajar en proyectos de inteligencia artificial."

# Crear un analizador de sentimientos
sia = SentimentIntensityAnalyzer()

# Obtener el puntaje de sentimiento
sentimiento = sia.polarity_scores(texto_sentimiento)

# Visualizar el resultado
plt.bar(sentimiento.keys(), sentimiento.values(), color=['green', 'red', 'gray', 'blue'])
plt.title('Análisis de Sentimientos')
plt.xlabel('Polaridad')
plt.ylabel('Puntaje')
plt.show()
```

Este código utiliza el analizador de sentimientos de NLTK y muestra un gráfico de barras que representa la polaridad del texto.

2. **Clasificación de Texto:**

La clasificación de texto se aplica para asignar categorías o etiquetas a un texto basándose en su contenido.

**Ejemplo con scikit-learn:**

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, confusion_matrix

# Datos de ejemplo
documentos = ["Esto es un texto positivo.", "Esto es un texto negativo.", "Este texto es neutro."]

# Etiquetas correspondientes
etiquetas = ["Positivo", "Negativo", "Neutro"]

# Crear un vectorizador de bolsa de palabras
vectorizador = CountVectorizer()
matriz_bolsa_palabras = vectorizador.fit_transform(documentos)

# Dividir los datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(matriz_bolsa_palabras, etiquetas, test_size=0.2, random_state=42)

# Crear un clasificador Naive Bayes
clasificador = MultinomialNB()
clasificador.fit(X_train, y_train)

# Hacer predicciones en el conjunto de prueba
y_pred = clasificador.predict(X_test)

# Calcular la precisión del modelo
precision = accuracy_score(y_test, y_pred)
print(f"Precisión del modelo: {precision:.2f}")

# Crear una matriz de confusión
cm = confusion_matrix(y_test, y_pred)
print("Matriz de Confusión:")
print(cm)
```

Este código utiliza un clasificador Naive Bayes para clasificar texto en categorías predefinidas.
