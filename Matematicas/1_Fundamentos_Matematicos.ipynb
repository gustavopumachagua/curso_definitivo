# **1. Fundamentos Matemáticos**

## **Álgebra Lineal**

### **Vectores y matrices**

**Vectores:**

Un vector es una entidad matemática que tiene magnitud y dirección. En el contexto matemático, los vectores se representan comúnmente como listas ordenadas de números. Un vector en $(n)$-dimensiones se denota como $\mathbf{v} = [v_1, v_2, \ldots, v_n]$.

**Operaciones con Vectores:**

1. **Suma de Vectores:**

   $\mathbf{v} + \mathbf{u} = [v_1 + u_1, v_2 + u_2, \ldots, v_n + u_n]$

2. **Multiplicación Escalar:**

   $c \cdot \mathbf{v} = [c \cdot v_1, c \cdot v_2, \ldots, c \cdot v_n]$

**Ejemplo en Python:**

````python
import numpy as np

# Definir dos vectores
v = np.array([1, 2, 3])
u = np.array([4, 5, 6])

# Suma de vectores
suma = v + u

# Multiplicación escalar
escalar = 2 * v

print("Vector v:", v)
print("Vector u:", u)
print("Suma de vectores:", suma)
print("Multiplicación escalar:", escalar)
````

**Matrices:**

Una matriz es una colección rectangular de números, símbolos o expresiones, dispuestos en filas y columnas. Se representa comúnmente como $A_{m \times n}$, donde $(m)$ es el número de filas y $(n)$ es el número de columnas.

**Operaciones con Matrices:**

1. **Suma de Matrices:**

   $C = A + B \quad \text{donde } c_{ij} = a_{ij} + b_{ij}$

2. **Multiplicación Escalar:**

   $c \cdot A \quad \text{donde } c_{ij} = c \cdot a_{ij}$

3. **Producto Matricial:**

   $C = A \cdot B \quad \text{donde } c_{ij} = \sum_{k=1}^{n} a_{ik} \cdot b_{kj}$

**Ejemplo en Python:**

````python
import numpy as np

# Definir dos matrices
A = np.array([[1, 2, 3], [4, 5, 6]])
B = np.array([[7, 8], [9, 10], [11, 12]])

# Transponer la matriz B para que tenga dimensiones (2, 3)
B_transpuesta = B.T

# Suma de matrices
suma_matriz = A + B_transpuesta

# Multiplicación escalar
escalar_matriz = 2 * A

# Producto matricial
producto_matricial = np.dot(A, B)

print("Matriz A:\n", A)
print("Matriz B (transpuesta):\n", B_transpuesta)
print("Suma de matrices:\n", suma_matriz)
print("Multiplicación escalar:\n", escalar_matriz)
print("Producto matricial:\n", producto_matricial)
````

### **Operaciones básicas**

**Operaciones Básicas con Vectores:**

**1. Suma de Vectores:**

La suma de dos vectores $(\mathbf{v})$ y $(\mathbf{u})$ se realiza componente por componente.

$\mathbf{v} + \mathbf{u} = [v_1 + u_1, v_2 + u_2, \ldots, v_n + u_n]$

**Ejemplo con Gráfico:**

````python
import numpy as np
import matplotlib.pyplot as plt

# Definir dos vectores
v = np.array([1, 2])
u = np.array([2, 1])

# Suma de vectores
suma = v + u

# Graficar los vectores
plt.figure()
plt.plot([0, v[0]], [0, v[1]], label='Vector v', color='r', marker='o')
plt.plot([0, u[0]], [0, u[1]], label='Vector u', color='b', marker='o')
plt.plot([0, suma[0]], [0, suma[1]], label='Suma v+u', color='g', marker='o')
plt.legend()
plt.grid(True)
plt.title('Suma de Vectores')
plt.show()
````

**2. Multiplicación Escalar:**

La multiplicación de un vector $\mathbf{v}$ por un escalar $(c)$ se realiza multiplicando cada componente por $(c)$.

$c \cdot \mathbf{v} = [c \cdot v_1, c \cdot v_2, \ldots, c \cdot v_n]$

**Ejemplo con Gráfico:**

````python
import numpy as np
import matplotlib.pyplot as plt

# Definir un vector
v = np.array([1, 2])

# Multiplicación escalar
escalar = 2 * v

# Graficar el vector original y el resultado
plt.figure()
plt.plot([0, v[0]], [0, v[1]], label='Vector v', color='r', marker='o')
plt.plot([0, escalar[0]], [0, escalar[1]], label='2 * v', color='b', marker='o')
plt.legend()
plt.grid(True)
plt.title('Multiplicación Escalar de un Vector')
plt.show()
````

**Operaciones Básicas con Matrices:**

**1. Suma de Matrices:**

La suma de dos matrices $(A)$ y $(B)$ se realiza sumando las correspondientes entradas.

$C = A + B \quad \text{donde } c_{ij} = a_{ij} + b_{ij}$

**Ejemplo con Gráfico:**

````python
import numpy as np
import matplotlib.pyplot as plt

# Definir dos matrices
A = np.array([[1, 2], [3, 4]])
B = np.array([[2, 1], [0, 1]])

# Suma de matrices
suma_matriz = A + B

# Graficar las matrices y su suma
fig, axs = plt.subplots(1, 3, figsize=(10, 4))

axs[0].imshow(A, cmap='Blues', interpolation='none')
axs[0].set_title('Matriz A')

axs[1].imshow(B, cmap='Blues', interpolation='none')
axs[1].set_title('Matriz B')

axs[2].imshow(suma_matriz, cmap='Blues', interpolation='none')
axs[2].set_title('Suma A + B')

plt.show()
````

**2. Multiplicación Escalar y Producto Matricial:**

La multiplicación de una matriz $(A)$ por un escalar $(c)$ se realiza multiplicando cada entrada por $(c)$.

$c \cdot A \quad \text{donde } c_{ij} = c \cdot a_{ij}$

El producto matricial de dos matrices $(A)$ y $(B)$ se obtiene sumando los productos de las entradas correspondientes.

$C = A \cdot B \quad \text{donde } c_{ij} = \sum_{k=1}^{n} a_{ik} \cdot b_{kj}$

**Ejemplo con Gráfico:**

````python
import numpy as np
import matplotlib.pyplot as plt

# Definir dos matrices
A = np.array([[1, 2], [3, 4]])
B = np.array([[2, 0], [1, 1]])

# Multiplicación escalar
escalar_matriz = 2 * A

# Producto matricial
producto_matricial = np.dot(A, B)

# Graficar las matrices y sus operaciones
fig, axs = plt.subplots(1, 3, figsize=(12, 4))

axs[0].imshow(A, cmap='Blues', interpolation='none')
axs[0].set_title('Matriz A')

axs[1].imshow(escalar_matriz, cmap='Blues', interpolation='none')
axs[1].set_title('2 * A')

axs[2].imshow(producto_matricial, cmap='Blues', interpolation='none')
axs[2].set_title('Producto A * B')

plt.show()
````

Estos ejemplos deberían ayudarte a entender las operaciones básicas con vectores y matrices, y cómo representarlas gráficamente en Python. Si tienes más preguntas o necesitas aclaraciones adicionales, no dudes en preguntar.

### **Espacios vectoriales**

Un espacio vectorial es un conjunto de elementos llamados vectores, junto con dos operaciones definidas: la suma de vectores y la multiplicación por escalares. Estas operaciones deben cumplir ciertas propiedades para que el conjunto sea considerado un espacio vectorial. Las propiedades incluyen la cerradura bajo la suma y la multiplicación por escalares, la existencia de un vector cero, la existencia de inversos aditivos, y la distributividad de la suma y la multiplicación por escalares.

**Propiedades de un Espacio Vectorial:**

1. **Cerradura bajo la Suma:**

   $\forall \mathbf{u}, \mathbf{v} \in V, \quad \mathbf{u} + \mathbf{v} \in V$

2. **Cerradura bajo la Multiplicación por Escalar:**

   $\forall \mathbf{u} \in V, \quad c \cdot \mathbf{u} \in V$

3. **Asociatividad de la Suma:**

   $\forall \mathbf{u}, \mathbf{v}, \mathbf{w} \in V, \quad (\mathbf{u} + \mathbf{v}) + \mathbf{w} = \mathbf{u} + (\mathbf{v} + \mathbf{w})$

4. **Existencia de un Elemento Neutro para la Suma:**

   $\exists \mathbf{0} \in V \ \text{tal que} \ \forall \mathbf{u} \in V, \quad \mathbf{u} + \mathbf{0} = \mathbf{u}$

5. **Existencia de Inverso Aditivo:**

   $\forall \mathbf{u} \in V, \ \exists (-\mathbf{u}) \in V \ \text{tal que} \ \mathbf{u} + (-\mathbf{u}) = \mathbf{0}$

6. **Compatibilidad con la Multiplicación Escalar:**

   $\forall \mathbf{u} \in V, \quad 1 \cdot \mathbf{u} = \mathbf{u}$

7. **Distributividad de la Suma respecto a la Multiplicación Escalar:**

   $\forall \mathbf{u} \in V, \forall c, d \in \mathbb{R}, \quad (c + d) \cdot \mathbf{u} = c \cdot \mathbf{u} + d \cdot \mathbf{u}$

8. **Distributividad de la Multiplicación Escalar respecto a la Suma de Vectores:**

   $\forall \mathbf{u} \in V, \forall c, d \in \mathbb{R}, \quad c \cdot (\mathbf{u} + \mathbf{v}) = c \cdot \mathbf{u} + d \cdot \mathbf{v}$

**Ejemplo con Gráfico en Python:**

````python
import numpy as np
import matplotlib.pyplot as plt

# Definir un espacio vectorial en R^2
V = np.array([[0, 0], [2, 3]])

# Graficar el espacio vectorial
plt.quiver(*V.T, angles='xy', scale_units='xy', scale=1, color=['r', 'b'])
plt.xlim(-1, 4)
plt.ylim(-1, 4)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.title('Espacio Vectorial en $R^2$')
plt.show()
````

En este ejemplo, el conjunto $(V)$ está representado por dos vectores en $(\mathbb{R}^2)$. Puedes observar que estos vectores cumplen con las propiedades de un espacio vectorial, como la cerradura bajo la suma y la multiplicación por escalares.

Este es un ejemplo simple, pero los espacios vectoriales pueden ser más complejos y existir en dimensiones superiores. Los espacios vectoriales son fundamentales en álgebra lineal y tienen aplicaciones en diversas áreas de las matemáticas y la física.

## **Cálculo Diferencial e Integral**

### **Derivadas y reglas de derivación**

**Derivadas:**

La derivada de una función describe cómo cambia la función con respecto a su variable independiente. Formalmente, si $f(x)$ es una función de $(x)$, la derivada de $(f)$ con respecto a $(x)$, denotada como $f'(x)$ o $(\frac{df}{dx})$, se define como el límite:

$f'(x) = \lim_{h \to 0} \frac{f(x + h) - f(x)}{h}$

Si este límite existe, la función $f(x)$ se considera diferenciable en el punto $(x)$. La derivada proporciona la pendiente de la recta tangente a la curva de la función en un punto dado.

**Reglas de Derivación:**

**1. Derivada de una constante:**

   $\frac{d}{dx} (c) = 0$

   Donde $(c)$ es una constante.

**2. Regla de Potencia:**

   $\frac{d}{dx} (x^n) = n \cdot x^{n-1}$

   Donde $(n)$ es un número real.

**3. Regla de la Suma y Resta:**

   $\frac{d}{dx} (f(x) + g(x)) = \frac{d}{dx} f(x) + \frac{d}{dx} g(x)$

   $\frac{d}{dx} (f(x) - g(x)) = \frac{d}{dx} f(x) - \frac{d}{dx} g(x)$

**4. Regla del Producto:**

   $\frac{d}{dx} (f(x) \cdot g(x)) = f'(x) \cdot g(x) + f(x) \cdot g'(x)$

**5. Regla del Cociente:**

   $\frac{d}{dx} \left(\frac{f(x)}{g(x)}\right) = \frac{f'(x) \cdot g(x) - f(x) \cdot g'(x)}{(g(x))^2}$

**Ejemplo con Gráfico en Python:**

````python
import numpy as np
import matplotlib.pyplot as plt

# Definir una función
def f(x):
    return x**3 - 3*x**2 + 2*x + 1

# Calcular la derivada
def df(x):
    return 3*x**2 - 6*x + 2

# Crear datos para graficar la función y su derivada
x = np.linspace(-2, 4, 400)
y = f(x)
dy = df(x)

# Graficar la función y su derivada
plt.plot(x, y, label='$f(x)$')
plt.plot(x, dy, label="$f'(x)$", linestyle='--')
plt.title('Función y su Derivada')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
````

En este ejemplo, la función $f(x) = x^3 - 3x^2 + 2x + 1$ y su derivada $f'(x) = 3x^2 - 6x + 2$ están graficadas. Puedes observar cómo la derivada representa la pendiente de la función en cada punto. La derivada es cero en los puntos críticos y toma valores positivos o negativos según la pendiente de la función en otros puntos.

Estas son solo algunas reglas básicas de derivación, y existen reglas más avanzadas para funciones más complejas. Las derivadas tienen aplicaciones fundamentales en cálculo y son esenciales para comprender el cambio y la tasa de cambio en matemáticas y ciencias aplicadas.

### **Integrales definidas e indefinidas**

Las integrales son operaciones inversas de las derivadas. Existen dos tipos principales: las integrales indefinidas y las integrales definidas.

**Integral Indefinida:**

La integral indefinida de una función $f(x)$ con respecto a $(x)$ se denota como $(\int f(x) \, dx)$ y se refiere a la familia de funciones cuyas derivadas son iguales a $f(x)$. Matemáticamente, se expresa como:

$\int f(x) \, dx = F(x) + C$

Donde $F(x)$ es la función antiderivada de $f(x)$ y $(C)$ es una constante de integración.

**Integral Definida:**

La integral definida de una función $f(x)$ en un intervalo $[a, b]$ se denota como:

$\int_{a}^{b} f(x) \, dx$

Representa el área bajo la curva de $f(x)$ en el intervalo $[a, b]$ en el plano xy. Puede ser interpretada como la acumulación total de cambios infinitesimales de la función a lo largo del intervalo.

**Ejemplo con Gráfico en Python:**

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

# Definir una función
def f(x):
    return x**2 - 2*x + 1

# Definir los límites de integración
a = 0
b = 2

# Graficar la función y la región bajo la curva
x = np.linspace(a, b, 100)
y = f(x)

fig, ax = plt.subplots()
ax.plot(x, y, 'r', linewidth=2)
ax.fill_between(x, y, alpha=0.2)

# Calcular e imprimir el valor de la integral definida
area, _ = quad(f, a, b)
print(f"Valor de la integral definida: {area}")

# Configuración del gráfico
ax.text(0.5, 1, f'$\int_{{0}}^{{2}} (x^2 - 2x + 1)dx = {area:.2f}$', horizontalalignment='center', verticalalignment='center', fontsize=12)
ax.set_title('Integral Definida')
ax.set_xlabel('x')
ax.set_ylabel('y')

plt.show()
````

En este ejemplo, la función $f(x) = x^2 - 2x + 1$ está graficada en el intervalo $[0, 2]$, y se sombrea el área bajo la curva. La integral definida $\int_{0}^{2} (x^2 - 2x + 1) \, dx$ representa el área de esa región y se calcula utilizando la función `quad` de SciPy.

Este es un caso simple, pero las integrales tienen aplicaciones extensas en cálculo, física y estadísticas para calcular áreas, volúmenes, probabilidades, entre otros.
