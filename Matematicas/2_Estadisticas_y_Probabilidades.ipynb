# **2. Estad铆sticas y Probabilidades**

## **Estad铆sticas Descriptivas**

### **Medidas de tendencia central y dispersi贸n**

Las medidas de tendencia central y dispersi贸n son herramientas estad铆sticas que resumen y describen conjuntos de datos.

**Medidas de Tendencia Central:**

**Media Aritm茅tica:**

La media aritm茅tica es el promedio de un conjunto de datos y se calcula sumando todos los valores y dividiendo por el n煤mero de observaciones.

$\text{Media} = \frac{\sum_{i=1}^{n} x_i}{n}$

**Mediana:**

La mediana es el valor que se encuentra en el centro del conjunto de datos ordenado. Si el conjunto tiene un n煤mero impar de elementos, la mediana es el valor central; si es par, es el promedio de los dos valores centrales.

**Moda:**

La moda es el valor que aparece con mayor frecuencia en un conjunto de datos.

**Medidas de Dispersi贸n:**

**Varianza:**

La varianza mide cu谩nto se dispersan los datos con respecto a la media. Se calcula como la media de los cuadrados de las desviaciones de cada punto respecto a la media.

$\text{Varianza} = \frac{\sum_{i=1}^{n} (x_i - \text{Media})^2}{n}$

**Desviaci贸n Est谩ndar:**

La desviaci贸n est谩ndar es simplemente la ra铆z cuadrada de la varianza. Proporciona una medida de dispersi贸n en las mismas unidades que los datos originales.

$\text{Desviaci贸n Est谩ndar} = \sqrt{\text{Varianza}}$

**Ejemplo con Gr谩fico en Python:**

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos aleatorios
np.random.seed(42)
data = np.random.normal(loc=10, scale=5, size=1000)

# Asegurarse de que los datos sean no negativos
data = np.abs(data)

# Medidas de Tendencia Central
media = np.mean(data)
mediana = np.median(data)

# Calcular la moda
counts, bins = np.histogram(data, bins=30)
modo_index = np.argmax(counts)
moda = (bins[modo_index] + bins[modo_index + 1]) / 2

# Medidas de Dispersi贸n
varianza = np.var(data)
desviacion_estandar = np.std(data)

# Graficar el histograma
plt.hist(data, bins=30, color='lightblue', edgecolor='black')
plt.axvline(media, color='red', linestyle='dashed', linewidth=2, label=f'Media: {media:.2f}')
plt.axvline(mediana, color='green', linestyle='dashed', linewidth=2, label=f'Mediana: {mediana:.2f}')
plt.axvline(moda, color='blue', linestyle='dashed', linewidth=2, label=f'Moda: {moda:.2f}')
plt.legend()
plt.title('Histograma y Medidas de Tendencia Central')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.show()

# Imprimir medidas de dispersi贸n
print(f"Varianza: {varianza:.2f}")
print(f"Desviaci贸n Est谩ndar: {desviacion_estandar:.2f}")
```

En este ejemplo, se generan datos aleatorios con una distribuci贸n normal y se calculan varias medidas de tendencia central y dispersi贸n. Luego, se grafica un histograma de los datos junto con l铆neas verticales que representan la media, mediana y moda.

Este es solo un ejemplo, y las medidas de tendencia central y dispersi贸n son 煤tiles para resumir y entender la distribuci贸n de un conjunto de datos.

### **Histogramas y gr谩ficos de dispersi贸n**

**Histogramas:**

Un histograma es una representaci贸n gr谩fica de la distribuci贸n de un conjunto de datos. Se utiliza para visualizar la frecuencia de ocurrencia de diferentes intervalos (o bins) en los datos. La altura de cada barra en el histograma indica la frecuencia o la densidad de datos en ese intervalo.

**F贸rmula Matem谩tica:**

La f贸rmula para calcular la frecuencia de un intervalo en un histograma es:

$\text{Frecuencia} = \frac{\text{N煤mero de datos en el intervalo}}{\text{Ancho del intervalo}}$

**Ejemplo con Gr谩fico en Python:**

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos aleatorios
np.random.seed(42)
data = np.random.normal(loc=10, scale=5, size=1000)

# Graficar el histograma
plt.hist(data, bins=30, color='lightblue', edgecolor='black')
plt.title('Histograma de Datos Aleatorios')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.show()
```

Este c贸digo utiliza NumPy para generar datos aleatorios con una distribuci贸n normal, y luego crea un histograma con Matplotlib. El histograma divide el rango de datos en 30 intervalos (bins) y muestra la frecuencia de datos en cada intervalo.

**Gr谩ficos de Dispersi贸n:**

Un gr谩fico de dispersi贸n es una representaci贸n visual de la relaci贸n entre dos variables. Cada punto en el gr谩fico representa un par de valores, uno para cada variable. Estos gr谩ficos son 煤tiles para identificar patrones, tendencias o correlaciones en los datos.

**F贸rmula Matem谩tica:**

No hay una f贸rmula espec铆fica para un gr谩fico de dispersi贸n, ya que simplemente muestra la relaci贸n entre dos conjuntos de datos. Sin embargo, la correlaci贸n entre las variables puede calcularse mediante diferentes m茅todos, como el coeficiente de correlaci贸n de Pearson.

**Ejemplo con Gr谩fico en Python:**

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos para un gr谩fico de dispersi贸n
np.random.seed(42)
x = np.random.normal(loc=10, scale=5, size=100)
y = 2 * x + np.random.normal(loc=0, scale=5, size=100)

# Graficar el gr谩fico de dispersi贸n
plt.scatter(x, y, color='coral')
plt.title('Gr谩fico de Dispersi贸n')
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.show()
```

En este ejemplo, se generan datos para dos variables $(X)$ e $(Y)$ y se crea un gr谩fico de dispersi贸n con Matplotlib. Puedes observar c贸mo los puntos se distribuyen en el plano, y en este caso, existe una relaci贸n lineal entre $(X)$ e $(Y)$ debido al t茅rmino $(2X)$ en la generaci贸n de datos.

Tanto los histogramas como los gr谩ficos de dispersi贸n son herramientas valiosas para explorar y visualizar conjuntos de datos en estad铆sticas y an谩lisis de datos.

## **Probabilidades**

### **Espacio muestral y eventos**

En teor铆a de probabilidad, el espacio muestral $(S)$ es el conjunto de todos los posibles resultados de un experimento aleatorio. Los eventos son subconjuntos del espacio muestral que describen posibles resultados o combinaciones de resultados del experimento.

**Espacio Muestral:**

El espacio muestral se denota como $(S)$ y contiene todos los resultados individuales posibles del experimento aleatorio. Puede ser finito o infinito, dependiendo de la naturaleza del experimento.

**Eventos:**

Un evento $(E)$ es cualquier subconjunto del espacio muestral $(S)$. Pueden clasificarse como eventos simples (que contienen un solo resultado) o eventos compuestos (que contienen m谩s de un resultado). La probabilidad de un evento se expresa como la proporci贸n del n煤mero de resultados favorables al evento con respecto al n煤mero total de resultados posibles.

**F贸rmulas Matem谩ticas:**

1. **Espacio Muestral:**

   $S = \{x_1, x_2, \ldots, x_n\}$

2. **Probabilidad de un Evento:**

   $P(E) = \frac{\text{N煤mero de resultados favorables a } E}{\text{N煤mero total de resultados en } S}$

**Ejemplo con Gr谩fico en Python:**

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir el espacio muestral para lanzar un dado
espacio_muestral_dado = {1, 2, 3, 4, 5, 6}

# Convertir el conjunto a una lista
espacio_muestral_lista = list(espacio_muestral_dado)

# Graficar el espacio muestral
plt.bar(espacio_muestral_lista, [1] * len(espacio_muestral_lista), color='lightblue', edgecolor='black')
plt.title('Espacio Muestral para Lanzar un Dado')
plt.xlabel('Resultado del dado')
plt.ylabel('Probabilidad')
plt.show()
```

En este ejemplo, el espacio muestral para lanzar un dado es el conjunto de n煤meros ${1, 2, 3, 4, 5, 6}$. La barra en el gr谩fico representa la probabilidad uniforme asociada con cada resultado.

**Ejemplo de Evento Compuesto:**

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir el espacio muestral para lanzar un dado
espacio_muestral_dado = {1, 2, 3, 4, 5, 6}

# Convertir el conjunto a una lista
espacio_muestral_lista = list(espacio_muestral_dado)

# Definir un evento compuesto: obtener un n煤mero par
evento_par = {2, 4, 6}

# Convertir el conjunto a una lista
evento_par_lista = list(evento_par)

# Graficar el espacio muestral y el evento compuesto
plt.bar(espacio_muestral_lista, [1] * len(espacio_muestral_lista), color='lightblue', edgecolor='black', label='Espacio Muestral')
plt.bar(evento_par_lista, [1] * len(evento_par_lista), color='coral', edgecolor='black', label='Evento: N煤mero Par')
plt.title('Espacio Muestral y Evento Compuesto para Lanzar un Dado')
plt.xlabel('Resultado del dado')
plt.ylabel('Probabilidad')
plt.legend()
plt.show()
```

En este ejemplo, el evento compuesto es obtener un n煤mero par al lanzar un dado. El gr谩fico muestra c贸mo el evento compuesto (n煤meros pares) es un subconjunto del espacio muestral.

Estos conceptos son fundamentales en probabilidad y son esenciales para comprender y modelar la incertidumbre en diversos contextos.

### **Distribuciones de probabilidad**

En teor铆a de probabilidad y estad铆sticas, una distribuci贸n de probabilidad describe la probabilidad de ocurrencia de cada valor posible en un conjunto de datos. Hay varias distribuciones de probabilidad comunes, cada una con propiedades y caracter铆sticas espec铆ficas.

**Distribuci贸n Discreta:**

**Distribuci贸n de Probabilidad Binomial:**

La distribuci贸n binomial modela el n煤mero de 茅xitos en un n煤mero fijo de ensayos independientes, donde cada ensayo tiene la misma probabilidad de 茅xito.

**F贸rmula Matem谩tica:**

$P(X = k) = \binom{n}{k} p^k (1-p)^{n-k}$

donde $(n)$ es el n煤mero de ensayos, $(k)$ es el n煤mero de 茅xitos, $(p)$ es la probabilidad de 茅xito en un solo ensayo y $\binom{n}{k}$ es el coeficiente binomial.

**Ejemplo con Gr谩fico en Python:**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Par谩metros de la distribuci贸n binomial
n = 10  # n煤mero de ensayos
p = 0.5  # probabilidad de 茅xito

# Generar valores de la variable aleatoria
x = np.arange(0, n+1)
y = binom.pmf(x, n, p)

# Graficar la distribuci贸n binomial
plt.bar(x, y, color='lightblue', edgecolor='black')
plt.title('Distribuci贸n Binomial')
plt.xlabel('N煤mero de xitos')
plt.ylabel('Probabilidad')
plt.show()
```

**Distribuci贸n Continua:**

**Distribuci贸n Normal (Gaussiana):**

La distribuci贸n normal es una de las distribuciones m谩s importantes en estad铆sticas. Es sim茅trica y se caracteriza completamente por su media $(\mu)$ y desviaci贸n est谩ndar $(\sigma)$.

**F贸rmula Matem谩tica:**

$f(x) = \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{(x-\mu)^2}{2\sigma^2}}$

**Ejemplo con Gr谩fico en Python:**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Par谩metros de la distribuci贸n normal
mu = 0    # media
sigma = 1  # desviaci贸n est谩ndar

# Generar valores de la variable aleatoria
x = np.linspace(-5, 5, 1000)
y = norm.pdf(x, mu, sigma)

# Graficar la distribuci贸n normal
plt.plot(x, y, color='blue')
plt.title('Distribuci贸n Normal')
plt.xlabel('Variable Aleatoria')
plt.ylabel('Densidad de Probabilidad')
plt.show()
```

Estos son solo ejemplos b谩sicos de dos distribuciones de probabilidad comunes. Hay muchas otras distribuciones (exponencial, Poisson, etc.) y cada una se utiliza en situaciones espec铆ficas. La elecci贸n de la distribuci贸n depende de la naturaleza de los datos y del problema que se est谩 abordando.
