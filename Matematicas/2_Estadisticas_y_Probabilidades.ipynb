# **2. Estad√≠sticas y Probabilidades**

## **Estad√≠sticas Descriptivas**

### **Medidas de tendencia central y dispersi√≥n**

Las medidas de tendencia central y dispersi√≥n son herramientas estad√≠sticas que resumen y describen conjuntos de datos.

**Medidas de Tendencia Central:**

**Media Aritm√©tica:**

La media aritm√©tica es el promedio de un conjunto de datos y se calcula sumando todos los valores y dividiendo por el n√∫mero de observaciones.

$\text{Media} = \frac{\sum_{i=1}^{n} x_i}{n}$

**Mediana:**

La mediana es el valor que se encuentra en el centro del conjunto de datos ordenado. Si el conjunto tiene un n√∫mero impar de elementos, la mediana es el valor central; si es par, es el promedio de los dos valores centrales.

**Moda:**

La moda es el valor que aparece con mayor frecuencia en un conjunto de datos.

**Medidas de Dispersi√≥n:**

**Varianza:**

La varianza mide cu√°nto se dispersan los datos con respecto a la media. Se calcula como la media de los cuadrados de las desviaciones de cada punto respecto a la media.

$\text{Varianza} = \frac{\sum_{i=1}^{n} (x_i - \text{Media})^2}{n}$

**Desviaci√≥n Est√°ndar:**

La desviaci√≥n est√°ndar es simplemente la ra√≠z cuadrada de la varianza. Proporciona una medida de dispersi√≥n en las mismas unidades que los datos originales.

$\text{Desviaci√≥n Est√°ndar} = \sqrt{\text{Varianza}}$

**Ejemplo con Gr√°fico en Python:**

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos aleatorios
np.random.seed(42)
data = np.random.normal(loc=10, scale=5, size=1000)

# Asegurarse de que los datos sean no negativos
data = np.abs(data)

# Medidas de Tendencia Central
media = np.mean(data)
mediana = np.median(data)

# Calcular la moda
counts, bins = np.histogram(data, bins=30)
modo_index = np.argmax(counts)
moda = (bins[modo_index] + bins[modo_index + 1]) / 2

# Medidas de Dispersi√≥n
varianza = np.var(data)
desviacion_estandar = np.std(data)

# Graficar el histograma
plt.hist(data, bins=30, color='lightblue', edgecolor='black')
plt.axvline(media, color='red', linestyle='dashed', linewidth=2, label=f'Media: {media:.2f}')
plt.axvline(mediana, color='green', linestyle='dashed', linewidth=2, label=f'Mediana: {mediana:.2f}')
plt.axvline(moda, color='blue', linestyle='dashed', linewidth=2, label=f'Moda: {moda:.2f}')
plt.legend()
plt.title('Histograma y Medidas de Tendencia Central')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.show()

# Imprimir medidas de dispersi√≥n
print(f"Varianza: {varianza:.2f}")
print(f"Desviaci√≥n Est√°ndar: {desviacion_estandar:.2f}")
```

En este ejemplo, se generan datos aleatorios con una distribuci√≥n normal y se calculan varias medidas de tendencia central y dispersi√≥n. Luego, se grafica un histograma de los datos junto con l√≠neas verticales que representan la media, mediana y moda.

Este es solo un ejemplo, y las medidas de tendencia central y dispersi√≥n son √∫tiles para resumir y entender la distribuci√≥n de un conjunto de datos.

### **Histogramas y gr√°ficos de dispersi√≥n**

**Histogramas:**

Un histograma es una representaci√≥n gr√°fica de la distribuci√≥n de un conjunto de datos. Se utiliza para visualizar la frecuencia de ocurrencia de diferentes intervalos (o bins) en los datos. La altura de cada barra en el histograma indica la frecuencia o la densidad de datos en ese intervalo.

**F√≥rmula Matem√°tica:**

La f√≥rmula para calcular la frecuencia de un intervalo en un histograma es:

$\text{Frecuencia} = \frac{\text{N√∫mero de datos en el intervalo}}{\text{Ancho del intervalo}}$

**Ejemplo con Gr√°fico en Python:**

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos aleatorios
np.random.seed(42)
data = np.random.normal(loc=10, scale=5, size=1000)

# Graficar el histograma
plt.hist(data, bins=30, color='lightblue', edgecolor='black')
plt.title('Histograma de Datos Aleatorios')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.show()
```

Este c√≥digo utiliza NumPy para generar datos aleatorios con una distribuci√≥n normal, y luego crea un histograma con Matplotlib. El histograma divide el rango de datos en 30 intervalos (bins) y muestra la frecuencia de datos en cada intervalo.

**Gr√°ficos de Dispersi√≥n:**

Un gr√°fico de dispersi√≥n es una representaci√≥n visual de la relaci√≥n entre dos variables. Cada punto en el gr√°fico representa un par de valores, uno para cada variable. Estos gr√°ficos son √∫tiles para identificar patrones, tendencias o correlaciones en los datos.

**F√≥rmula Matem√°tica:**

No hay una f√≥rmula espec√≠fica para un gr√°fico de dispersi√≥n, ya que simplemente muestra la relaci√≥n entre dos conjuntos de datos. Sin embargo, la correlaci√≥n entre las variables puede calcularse mediante diferentes m√©todos, como el coeficiente de correlaci√≥n de Pearson.

**Ejemplo con Gr√°fico en Python:**

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos para un gr√°fico de dispersi√≥n
np.random.seed(42)
x = np.random.normal(loc=10, scale=5, size=100)
y = 2 * x + np.random.normal(loc=0, scale=5, size=100)

# Graficar el gr√°fico de dispersi√≥n
plt.scatter(x, y, color='coral')
plt.title('Gr√°fico de Dispersi√≥n')
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.show()
```

En este ejemplo, se generan datos para dos variables $(X)$ e $(Y)$ y se crea un gr√°fico de dispersi√≥n con Matplotlib. Puedes observar c√≥mo los puntos se distribuyen en el plano, y en este caso, existe una relaci√≥n lineal entre $(X)$ e $(Y)$ debido al t√©rmino $(2X)$ en la generaci√≥n de datos.

Tanto los histogramas como los gr√°ficos de dispersi√≥n son herramientas valiosas para explorar y visualizar conjuntos de datos en estad√≠sticas y an√°lisis de datos.

## **Probabilidades**

### **Espacio muestral y eventos**

En teor√≠a de probabilidad, el espacio muestral $(S)$ es el conjunto de todos los posibles resultados de un experimento aleatorio. Los eventos son subconjuntos del espacio muestral que describen posibles resultados o combinaciones de resultados del experimento.

**Espacio Muestral:**

El espacio muestral se denota como $(S)$ y contiene todos los resultados individuales posibles del experimento aleatorio. Puede ser finito o infinito, dependiendo de la naturaleza del experimento.

**Eventos:**

Un evento $(E)$ es cualquier subconjunto del espacio muestral $(S)$. Pueden clasificarse como eventos simples (que contienen un solo resultado) o eventos compuestos (que contienen m√°s de un resultado). La probabilidad de un evento se expresa como la proporci√≥n del n√∫mero de resultados favorables al evento con respecto al n√∫mero total de resultados posibles.

**F√≥rmulas Matem√°ticas:**

1. **Espacio Muestral:**

   $S = \{x_1, x_2, \ldots, x_n\}$

2. **Probabilidad de un Evento:**

   $P(E) = \frac{\text{N√∫mero de resultados favorables a } E}{\text{N√∫mero total de resultados en } S}$

**Ejemplo con Gr√°fico en Python:**

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir el espacio muestral para lanzar un dado
espacio_muestral_dado = {1, 2, 3, 4, 5, 6}

# Convertir el conjunto a una lista
espacio_muestral_lista = list(espacio_muestral_dado)

# Graficar el espacio muestral
plt.bar(espacio_muestral_lista, [1] * len(espacio_muestral_lista), color='lightblue', edgecolor='black')
plt.title('Espacio Muestral para Lanzar un Dado')
plt.xlabel('Resultado del dado')
plt.ylabel('Probabilidad')
plt.show()
```

En este ejemplo, el espacio muestral para lanzar un dado es el conjunto de n√∫meros ${1, 2, 3, 4, 5, 6}$. La barra en el gr√°fico representa la probabilidad uniforme asociada con cada resultado.

**Ejemplo de Evento Compuesto:**

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir el espacio muestral para lanzar un dado
espacio_muestral_dado = {1, 2, 3, 4, 5, 6}

# Convertir el conjunto a una lista
espacio_muestral_lista = list(espacio_muestral_dado)

# Definir un evento compuesto: obtener un n√∫mero par
evento_par = {2, 4, 6}

# Convertir el conjunto a una lista
evento_par_lista = list(evento_par)

# Graficar el espacio muestral y el evento compuesto
plt.bar(espacio_muestral_lista, [1] * len(espacio_muestral_lista), color='lightblue', edgecolor='black', label='Espacio Muestral')
plt.bar(evento_par_lista, [1] * len(evento_par_lista), color='coral', edgecolor='black', label='Evento: N√∫mero Par')
plt.title('Espacio Muestral y Evento Compuesto para Lanzar un Dado')
plt.xlabel('Resultado del dado')
plt.ylabel('Probabilidad')
plt.legend()
plt.show()
```

En este ejemplo, el evento compuesto es obtener un n√∫mero par al lanzar un dado. El gr√°fico muestra c√≥mo el evento compuesto (n√∫meros pares) es un subconjunto del espacio muestral.

Estos conceptos son fundamentales en probabilidad y son esenciales para comprender y modelar la incertidumbre en diversos contextos.

### **Distribuciones de probabilidad**

En teor√≠a de probabilidad y estad√≠sticas, una distribuci√≥n de probabilidad describe la probabilidad de ocurrencia de cada valor posible en un conjunto de datos. Hay varias distribuciones de probabilidad comunes, cada una con propiedades y caracter√≠sticas espec√≠ficas.

**Distribuci√≥n Discreta:**

**Distribuci√≥n de Probabilidad Binomial:**

La distribuci√≥n binomial modela el n√∫mero de √©xitos en un n√∫mero fijo de ensayos independientes, donde cada ensayo tiene la misma probabilidad de √©xito.

**F√≥rmula Matem√°tica:**

$P(X = k) = \binom{n}{k} p^k (1-p)^{n-k}$

donde $(n)$ es el n√∫mero de ensayos, $(k)$ es el n√∫mero de √©xitos, $(p)$ es la probabilidad de √©xito en un solo ensayo y $\binom{n}{k}$ es el coeficiente binomial.

**Ejemplo con Gr√°fico en Python:**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Par√°metros de la distribuci√≥n binomial
n = 10  # n√∫mero de ensayos
p = 0.5  # probabilidad de √©xito

# Generar valores de la variable aleatoria
x = np.arange(0, n+1)
y = binom.pmf(x, n, p)

# Graficar la distribuci√≥n binomial
plt.bar(x, y, color='lightblue', edgecolor='black')
plt.title('Distribuci√≥n Binomial')
plt.xlabel('N√∫mero de √âxitos')
plt.ylabel('Probabilidad')
plt.show()
```

**Distribuci√≥n Continua:**

**Distribuci√≥n Normal (Gaussiana):**

La distribuci√≥n normal es una de las distribuciones m√°s importantes en estad√≠sticas. Es sim√©trica y se caracteriza completamente por su media $(\mu)$ y desviaci√≥n est√°ndar $(\sigma)$.

**F√≥rmula Matem√°tica:**

$f(x) = \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{(x-\mu)^2}{2\sigma^2}}$

**Ejemplo con Gr√°fico en Python:**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Par√°metros de la distribuci√≥n normal
mu = 0    # media
sigma = 1  # desviaci√≥n est√°ndar

# Generar valores de la variable aleatoria
x = np.linspace(-5, 5, 1000)
y = norm.pdf(x, mu, sigma)

# Graficar la distribuci√≥n normal
plt.plot(x, y, color='blue')
plt.title('Distribuci√≥n Normal')
plt.xlabel('Variable Aleatoria')
plt.ylabel('Densidad de Probabilidad')
plt.show()
```

Estos son solo ejemplos b√°sicos de dos distribuciones de probabilidad comunes. Hay muchas otras distribuciones (exponencial, Poisson, etc.) y cada una se utiliza en situaciones espec√≠ficas. La elecci√≥n de la distribuci√≥n depende de la naturaleza de los datos y del problema que se est√° abordando.
