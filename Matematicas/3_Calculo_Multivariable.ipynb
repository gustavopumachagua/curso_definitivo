# **3. C√°lculo Multivariable**

## **Derivadas Parciales**

### **Derivadas parciales de funciones multivariables**

Las derivadas parciales son un concepto fundamental en c√°lculo y an√°lisis matem√°tico, especialmente cuando se trabaja con funciones de varias variables. Mientras que la derivada de una funci√≥n de una variable mide la tasa de cambio con respecto a una variable, las derivadas parciales miden la tasa de cambio con respecto a una variable espec√≠fica, manteniendo las otras variables constantes.

**F√≥rmula Matem√°tica:**

Supongamos que tenemos una funci√≥n $f(x, y)$ de dos variables. La derivada parcial de $(f)$ con respecto a $(x)$ se denota como $\frac{\partial f}{\partial x}$ y se calcula manteniendo $(y)$ constante:

$\frac{\partial f}{\partial x} = \lim_{h \to 0} \frac{f(x + h, y) - f(x, y)}{h}$

An√°logamente, la derivada parcial de $(f)$ con respecto a $(y)$ se denota como $\frac{\partial f}{\partial y}$ y se calcula manteniendo $(x)$ constante:

$\frac{\partial f}{\partial y} = \lim_{h \to 0} \frac{f(x, y + h) - f(x, y)}{h}$

**Ejemplo con Gr√°fico en Python:**

Vamos a considerar la funci√≥n $f(x, y) = x^2 + 2y^2$ y calcular sus derivadas parciales con respecto a $(x)$ e $(y)$.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la funci√≥n
def f(x, y):
    return x**2 + 2*y**2

# Calcular las derivadas parciales
def df_dx(x, y, h=1e-5):
    return (f(x + h, y) - f(x, y)) / h

def df_dy(x, y, h=1e-5):
    return (f(x, y + h) - f(x, y)) / h

# Generar datos para el gr√°fico
x_vals = np.linspace(-5, 5, 100)
y_vals = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x_vals, y_vals)
Z = f(X, Y)

# Calcular las derivadas parciales en un punto espec√≠fico
x_point = 2
y_point = 1
df_dx_point = df_dx(x_point, y_point)
df_dy_point = df_dy(x_point, y_point)

# Graficar la funci√≥n y las derivadas parciales
fig = plt.figure(figsize=(12, 5))

# Gr√°fico de la funci√≥n
ax1 = fig.add_subplot(121, projection='3d')
ax1.plot_surface(X, Y, Z, cmap='viridis')
ax1.set_title('Funci√≥n $f(x, y) = x^2 + 2y^2$')

# Gr√°fico de las derivadas parciales en el punto espec√≠fico
ax2 = fig.add_subplot(122)
ax2.quiver(x_point, y_point, df_dx_point, df_dy_point, scale=100, color='red', label='Derivadas Parciales')
ax2.set_title('Derivadas Parciales en el Punto Espec√≠fico')
ax2.contour(X, Y, Z, levels=20, cmap='viridis', alpha=0.5)
ax2.legend()

plt.show()
```

En este ejemplo, calculamos las derivadas parciales de la funci√≥n $f(x, y) = x^2 + 2y^2$ con respecto a $(x)$ e $(y)$. Luego, graficamos la funci√≥n tridimensional y las derivadas parciales en un punto espec√≠fico en el plano $(xy)$. Los vectores rojos en el gr√°fico de flechas indican las direcciones y magnitudes de las derivadas parciales en ese punto.

## **Integrales M√∫ltiples**

### **Integrales dobles y triples**

Las integrales dobles y triples son extensiones de la integraci√≥n de una variable a funciones de dos y tres variables, respectivamente. Estas herramientas son fundamentales en c√°lculo y an√°lisis matem√°tico, y se utilizan para calcular √°reas, vol√∫menes, masas y otros conceptos en diversas disciplinas.

**Integral Doble:**

**F√≥rmula Matem√°tica:**

Para una funci√≥n $f(x, y)$ sobre una regi√≥n $(R)$ en el plano $(xy)$, la integral doble se expresa como:

$\iint_R f(x, y) \,dx\,dy$

Donde $(dx\,dy)$ indica la integraci√≥n respecto a $(x)$ y luego respecto a $(y)$. La regi√≥n $(R)$ define los l√≠mites de integraci√≥n.

**Ejemplo con Gr√°fico en Python:**

Vamos a calcular la integral doble de $f(x, y) = x^2 + y^2$ sobre el rect√°ngulo $R = [0, 2] \times [0, 1]$.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import dblquad

# Definir la funci√≥n
def f(x, y):
    return x**2 + y**2

# Definir los l√≠mites de integraci√≥n
x_min, x_max = 0, 2
y_min, y_max = 0, 1

# Calcular la integral doble
result, error = dblquad(f, x_min, x_max, y_min, y_max)

# Imprimir el resultado
print(f"Resultado de la integral doble: {result}")

# Graficar la regi√≥n R y la funci√≥n
x_vals = np.linspace(x_min, x_max, 100)
y_vals = np.linspace(y_min, y_max, 100)
X, Y = np.meshgrid(x_vals, y_vals)
Z = f(X, Y)

plt.contourf(X, Y, Z, levels=20, cmap='viridis')
plt.colorbar()
plt.title('Regi√≥n de Integraci√≥n y Funci√≥n $f(x, y) = x^2 + y^2$')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

En este ejemplo, calculamos la integral doble de $f(x, y) = x^2 + y^2$ sobre el rect√°ngulo $R = [0, 2] \times [0, 1]$. El gr√°fico muestra la regi√≥n de integraci√≥n y la funci√≥n.

**Integral Triple:**

**F√≥rmula Matem√°tica:**

Para una funci√≥n $g(x, y, z)$ sobre un s√≥lido $(S)$ en el espacio tridimensional, la integral triple se expresa como:

$\iiint_S g(x, y, z) \,dx\,dy\,dz$

Donde $(dx\,dy\,dz)$ indica la integraci√≥n respecto a $(x)$, $(y)$, y luego respecto a $(z)$. El s√≥lido $(S)$ define los l√≠mites de integraci√≥n.

**Ejemplo con Gr√°fico en Python:**

Vamos a calcular la integral triple de $g(x, y, z) = xyz$ sobre el cubo $S = [0, 1] \times [0, 1] \times [0, 1]$.

```python
from scipy.integrate import tplquad

# Definir la funci√≥n
def g(x, y, z):
    return x * y * z

# Definir los l√≠mites de integraci√≥n
x_min, x_max = 0, 1
y_min, y_max = 0, 1
z_min, z_max = 0, 1

# Calcular la integral triple
result, error = tplquad(g, x_min, x_max, y_min, y_max, z_min, z_max)

# Imprimir el resultado
print(f"Resultado de la integral triple: {result}")
```

En este ejemplo, calculamos la integral triple de $g(x, y, z) = xyz$ sobre el cubo $S = [0, 1] \times [0, 1] \times [0, 1]$. No se proporciona un gr√°fico en este caso debido a la complejidad de representar un s√≥lido tridimensional.

Estos son ejemplos b√°sicos para ilustrar las integrales dobles y triples. Las aplicaciones en ciencia, ingenier√≠a y otras disciplinas son numerosas, ya que estas integrales permiten calcular propiedades f√≠sicas y geom√©tricas de objetos en m√∫ltiples dimensiones.

### **Coordenadas polares y esf√©ricas**

Las coordenadas polares y esf√©ricas son sistemas alternativos para describir posiciones en el plano y en el espacio tridimensional, respectivamente. Mientras que las coordenadas cartesianas utilizan ejes rectangulares (x, y, z), las coordenadas polares y esf√©ricas utilizan conceptos radiales y angulares.

**Coordenadas Polares:**

En el plano, las coordenadas polares $(r, \theta)$ se definen por la distancia radial $(r)$ desde el origen y el √°ngulo polar $(\theta)$ entre la l√≠nea que conecta el punto con el origen y el eje x.

**F√≥rmulas Matem√°ticas:**

$x = r \cdot \cos(\theta)$

$y = r \cdot \sin(\theta)$

**Ejemplo con Gr√°fico en Python:**

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir coordenadas polares
r = 2
theta = np.pi / 4  # 45 grados

# Convertir a coordenadas cartesianas
x = r * np.cos(theta)
y = r * np.sin(theta)

# Graficar punto en coordenadas polares
plt.polar(theta, r, 'ro')  # Gr√°fico polar
plt.text(theta, r, f'  ({x}, {y})', color='red')
plt.title('Coordenadas Polares')
plt.show()
```

En este ejemplo, se define un punto en coordenadas polares $(r, \theta) = (2, \frac{\pi}{4})$ y se convierte a coordenadas cartesianas para graficar el punto en un sistema polar.

**Coordenadas Esf√©ricas:**

En el espacio tridimensional, las coordenadas esf√©ricas $(r, \theta, \phi)$ se definen por la distancia radial $(r)$ desde el origen, el √°ngulo polar $(\theta)$ (como en coordenadas polares) y el √°ngulo azimutal $(\phi)$ que mide la inclinaci√≥n desde el eje z.

**F√≥rmulas Matem√°ticas:**

$x = r \cdot \sin(\phi) \cdot \cos(\theta)$

$y = r \cdot \sin(\phi) \cdot \sin(\theta)$

$z = r \cdot \cos(\phi)$

**Ejemplo con Gr√°fico en Python:**

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Definir coordenadas esf√©ricas
r = 2
theta = np.pi / 4  # √Ångulo polar
phi = np.pi / 3    # √Ångulo azimutal

# Convertir a coordenadas cartesianas
x = r * np.sin(phi) * np.cos(theta)
y = r * np.sin(phi) * np.sin(theta)
z = r * np.cos(phi)

# Graficar punto en coordenadas esf√©ricas
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z, color='red', s=100, label=f'({x}, {y}, {z})')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.legend()
plt.title('Coordenadas Esf√©ricas')
plt.show()
```

En este ejemplo, se define un punto en coordenadas esf√©ricas $(r, \theta, \phi) = (2, \frac{\pi}{4}, \frac{\pi}{3})$ y se convierte a coordenadas cartesianas para graficar el punto en un sistema tridimensional.
