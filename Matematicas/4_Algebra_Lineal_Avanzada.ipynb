## **Transformaciones Lineales**

### **N煤cleo y rango**

En 谩lgebra lineal, las transformaciones lineales son funciones entre espacios vectoriales que respetan la estructura lineal. El n煤cleo y el rango son conceptos asociados a estas transformaciones.

**Transformaciones Lineales:**

Una transformaci贸n lineal $(T: V \rightarrow W)$ entre dos espacios vectoriales $(V)$ y $(W)$ cumple dos propiedades fundamentales:

1. **Aditividad:** $(T(u + v) = T(u) + T(v))$ para cualquier $(u, v \in V)$.

2. **Homogeneidad:** $(T(cu) = cT(u))$ para cualquier escalar $(c)$ y $(u \in V)$.

**N煤cleo (Kernel):**

El n煤cleo de una transformaci贸n lineal $(T: V \rightarrow W)$, denotado como $(\text{ker}(T))$, es el conjunto de vectores $(v)$ en $(V)$ que mapean a $(0)$ en $(W)$.

**F贸rmula Matem谩tica:**

$\text{ker}(T) = \{v \in V \mid T(v) = 0\}$

**Rango (Imagen):**

El rango de una transformaci贸n lineal $(T: V \rightarrow W)$, denotado como $(\text{im}(T))$ o $(\text{ran}(T))$, es el conjunto de vectores en $(W)$ que son im谩genes de alg煤n vector en $(V)$.

**F贸rmula Matem谩tica:**

$\text{im}(T) = \{T(v) \mid v \in V\}$

**Ejemplo con Gr谩fico en Python:**

Vamos a considerar una transformaci贸n lineal en el plano 2D. Supongamos $(T: \mathbb{R}^2 \rightarrow \mathbb{R}^2)$ dada por $(T(x, y) = (2x, y))$.


```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la transformaci贸n lineal
def linear_transformation(v):
    x, y = v
    return np.array([2*x, y])

# Generar puntos en el plano 2D
points = np.array([[1, 1], [2, 2], [3, 1]])

# Aplicar la transformaci贸n a los puntos
transformed_points = np.array([linear_transformation(point) for point in points])

# Graficar los puntos originales y transformados
plt.figure(figsize=(8, 4))

plt.subplot(121)
plt.scatter(points[:, 0], points[:, 1], color='blue', label='Original')
plt.title('Puntos Originales')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()

plt.subplot(122)
plt.scatter(transformed_points[:, 0], transformed_points[:, 1], color='red', label='Transformado')
plt.title('Puntos Transformados')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

En este ejemplo, la transformaci贸n lineal $T(x, y) = (2x, y)$ multiplica la coordenada x por 2, manteniendo la coordenada y constante. Los puntos originales (en azul) y los puntos transformados (en rojo) se grafican para ilustrar la transformaci贸n lineal.

**C谩lculo del N煤cleo y el Rango:**

En Python, puedes utilizar bibliotecas como NumPy para calcular el n煤cleo y el rango de una matriz que representa la transformaci贸n lineal. Aqu铆 hay un ejemplo:

```python
import numpy as np

# Definir la matriz que representa la transformaci贸n lineal
matrix = np.array([[2, 0], [0, 1]])

# Calcular el n煤cleo y el rango
kernel = np.linalg.solve(matrix, np.zeros_like(matrix[0]))
range_space = np.linalg.matrix_rank(matrix)

print(f"N煤cleo de la matriz: {kernel}")
print(f"Rango de la matriz: {range_space}")
```

Este c贸digo utiliza la funci贸n `np.linalg.solve` para encontrar el n煤cleo (soluci贸n del sistema lineal $(Ax = 0)$ y `np.linalg.matrix_rank` para calcular el rango de la matriz, que es igual al rango de la transformaci贸n lineal.

### **Eigenvalores y eigenvectores**

Las transformaciones lineales, eigenvalores y eigenvectores son conceptos fundamentales en 谩lgebra lineal que se aplican en diversas 谩reas como f铆sica, estad铆sticas, aprendizaje autom谩tico y m谩s.

**Transformaciones Lineales:**

Una transformaci贸n lineal $(T: V \rightarrow W)$ entre dos espacios vectoriales $(V)$ y $(W)$ cumple dos propiedades fundamentales:

1. **Aditividad:** $(T(u + v) = T(u) + T(v))$ para cualquier $(u, v \in V)$.

2. **Homogeneidad:** $(T(cu) = cT(u))$ para cualquier escalar $(c)$ y $(u \in V)$.

**Eigenvalores y Eigenvectores:**

Dada una matriz cuadrada $(A)$, un eigenvalor $(\lambda)$ y un eigenvector $(v)$ satisfacen la ecuaci贸n:

$Av = \lambda v$

Esto significa que la acci贸n de la matriz $(A)$ sobre el eigenvector $(v)$ solo escala el vector por el factor $(\lambda)$.

**F贸rmulas Matem谩ticas:**

1. **Eigenvalor:**

$\text{det}(A - \lambda I) = 0$

donde $(I)$ es la matriz identidad.

2. **Eigenvector:**

$(A - \lambda I) \mathbf{v} = \mathbf{0}$

La soluci贸n no trivial de esta ecuaci贸n es el eigenvector.

**Ejemplo con Gr谩fico en Python:**

Consideremos una matriz de transformaci贸n $(A)$ y encontremos sus eigenvalores y eigenvectores.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la matriz de transformaci贸n
A = np.array([[3, 1],
              [1, 2]])

# Calcular eigenvalores y eigenvectores
eigenvalues, eigenvectors = np.linalg.eig(A)

# Graficar los eigenvectores
origin = np.zeros_like(eigenvectors[:, 0])

plt.quiver(*origin, *eigenvectors[:, 0], angles='xy', scale_units='xy', scale=1, color='blue', label=f'Eigenvalue: {eigenvalues[0]}')
plt.quiver(*origin, *eigenvectors[:, 1], angles='xy', scale_units='xy', scale=1, color='red', label=f'Eigenvalue: {eigenvalues[1]}')

plt.xlim(-1, 4)
plt.ylim(-1, 2)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.title('Eigenvalues and Eigenvectors')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

En este ejemplo, la matriz $(A)$ tiene los eigenvalores $(\lambda_1 = 4)$ y $(\lambda_2 = 1)$, con eigenvectores asociados. Estos eigenvectores est谩n representados en el plano con flechas de colores, y cada flecha est谩 escalada por su eigenvalor asociado.

**Aplicaci贸n de Transformaci贸n Lineal:**

Vamos a aplicar la transformaci贸n lineal $(A)$ a un conjunto de puntos y observar c贸mo cambian bajo la acci贸n de la matriz.

```python
# Puntos originales
points = np.array([[1, 0],
                   [0, 1]])

# Aplicar transformaci贸n lineal
transformed_points = A.dot(points.T).T

# Graficar puntos originales y transformados
plt.scatter(points[:, 0], points[:, 1], color='blue', label='Original')
plt.scatter(transformed_points[:, 0], transformed_points[:, 1], color='red', label='Transformado')

plt.xlim(-1, 4)
plt.ylim(-1, 2)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.title('Transformaci贸n Lineal')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

Aqu铆, los puntos originales (en azul) y los puntos transformados (en rojo) ilustran c贸mo la transformaci贸n lineal afecta al conjunto de puntos bajo la acci贸n de la matriz $(A)$.

## **Descomposici贸n de Valores Singulares (SVD)**

### **Aplicaciones en reducci贸n de dimensionalidad**

La Descomposici贸n de Valores Singulares (SVD, por sus siglas en ingl茅s) es una factorizaci贸n matricial fundamental que se aplica a cualquier matriz, no solo a matrices cuadradas. La SVD tiene diversas aplicaciones, y una de las m谩s comunes es la reducci贸n de dimensionalidad.

**Descomposici贸n de Valores Singulares (SVD):**

Dada una matriz $(A)$ de dimensiones $(m \times n)$, la SVD de $(A)$ es expresada como:

$A = U \Sigma V^T$

Donde:

- $(U)$ es una matriz ortogonal $(m \times m)$.
- $(\Sigma)$ es una matriz diagonal $(m \times n)$ con valores singulares en la diagonal principal.
- $(V^T)$ es la traspuesta de una matriz ortogonal $(n \times n)$.

Los valores singulares en $(\Sigma)$ est谩n ordenados de mayor a menor, y son las ra铆ces cuadradas de los autovalores de $(A^T A)$ o $(A A^T)$.

**Reducci贸n de Dimensionalidad con SVD:**

La reducci贸n de dimensionalidad implica truncar la SVD eliminando algunos de los valores singulares y las columnas correspondientes en $(U)$ y $(V)$. Esto es 煤til para aproximaciones de baja dimensionalidad y reducci贸n de ruido.

**F贸rmulas Matem谩ticas:**

Si $(A)$ tiene rango $(r)$, la aproximaci贸n de rango $(k < r)$ se obtiene truncando la SVD como sigue:

$A_k = U_k \Sigma_k V_k^T$

Donde:

- $(U_k)$ es la matriz $(U)$ truncada a las primeras $(k)$ columnas.
- $(\Sigma_k)$ es la matriz diagonal $(\Sigma)$ truncada a las primeras $(k)$ filas y $(k)$ columnas.
- $(V_k^T)$ es la traspuesta de la matriz $(V)$ truncada a las primeras $(k)$ filas.

**Ejemplo con Gr谩fico en Python:**

Vamos a generar una matriz aleatoria y aplicar la SVD para reducci贸n de dimensionalidad.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar una matriz aleatoria
np.random.seed(42)
A = np.random.rand(2, 2)

# Aplicar la SVD
U, Sigma, Vt = np.linalg.svd(A)

# Reducci贸n de dimensionalidad a rango 1
k = 1
Ak = np.dot(U[:, :k], np.dot(np.diag(Sigma[:k]), Vt[:k, :]))

# Graficar la matriz original y la aproximaci贸n
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

# Matriz Original
axes[0].imshow(A, cmap='Blues')
axes[0].set_title('Matriz Original')
axes[0].grid(False)

# Aproximaci贸n de rango 1
axes[1].imshow(Ak, cmap='Blues')
axes[1].set_title('Aproximaci贸n de rango 1')
axes[1].grid(False)

plt.show()
```

En este ejemplo, generamos una matriz aleatoria $(A)$ y aplicamos la SVD. Luego, reducimos la dimensionalidad a rango 1 y comparamos la matriz original con la aproximaci贸n de rango 1.

La reducci贸n de dimensionalidad con SVD es 煤til en tareas como compresi贸n de im谩genes, procesamiento de texto, y en general, cuando se busca representar la informaci贸n de manera m谩s eficiente manteniendo la informaci贸n m谩s relevante.
